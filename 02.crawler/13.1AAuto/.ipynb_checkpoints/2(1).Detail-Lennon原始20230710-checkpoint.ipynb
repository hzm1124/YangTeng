```python
import pandas as pd
```

```python
df_menu = pd.read_excel('./menu.xlsx', header=0)
list_url = df_menu['Url'].to_list()

print('总数量：' + str(len(list_url)))
list_url
```

```python
df = pd.DataFrame(columns=['Url',
                           'vehicle_url',
                           'Src',
                           'status'])

df
```

```python
import requests
```

```python
i = 0

resp = requests.get(list_url[i]).text

resp
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(resp, 'lxml')

print(soup.prettify())
```

```python
with open('./part.html', 'w', encoding='utf-8') as file:
    file.write(str(soup))
```

```python
product_id = str(soup).split('window.productId = ')[1].split(';')[0]

product_id
```

```python
vehicle_url = 'https://www.1aauto.com/catalog/product/fit?skuId=' + product_id + '&pathName=%2F' + list_url[i].split('/')[3] + '%2F' + list_url[i].split('/')[4] + '%2F' + list_url[i].split('/')[5]

vehicle_url
```

```python
from lxml import etree
```

```python
html = etree.HTML(str(soup))

src = html.xpath('//img[@id="main-img"]/@src')[0]

src
```

```python
df_temp = pd.DataFrame([{'Url': list_url[i],
                         'vehicle_url': vehicle_url,
                         'Src': src,
                         'status': 'ok'}])

df = pd.concat([df, df_temp], ignore_index=True).fillna('')

df
```
