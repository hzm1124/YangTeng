```python
import pandas as pd

output_correct, output_error = pd.DataFrame(), pd.DataFrame()
```

```python
input_ = pd.read_excel(f'./file/5.item.xlsx',
                       header=0,
                       dtype=str).fillna('')

input_['Year'] = input_['Year'].astype(int)
input_['SKU Count'] = input_['SKU Count'].astype(int)
input_['Page'] = input_['Page'].astype(int)

print(f'总数量：{len(input_)}')
input_['Item'].tolist()
```

```python
import requests

import sys
sys.path.append('../00.Tools')
from crawler_configuration import get_header, get_proxy

b = 0
while True:
    b += 1
    print(b)
    try:
        resp = requests.get(f'''https://www.auveco.com/api/cacheable/items?country=US&currency=USD&custitem_uc_ymm_make={input_.loc[a, 'Make']}&custitem_uc_ymm_model={input_.loc[a, 'Model']}&custitem_uc_ymm_year={input_.loc[a, 'Year']}&fieldset=search&id={input_.loc[a, 'Item']}&include=facets&language=en&limit=100&offset=0&pricelevel=1&sort=relevance%3Adesc&use_pcv=T''',
                            headers=get_header(),
                            proxies=get_proxy(),
                            timeout=(10, 10))

        if resp.status_code == 404 or resp.status_code == 200:
            json_ = resp.json()

            if json_['code'] == 404 and json_['errors'] == 'URL not found':
                break
                
            list_urlcomponent = [dict_['urlcomponent'] for dict_ in json_['items']]
            if json_['code'] == 200 and list_urlcomponent:
                break
    except KeyboardInterrupt:
        break
    except:
        continue

json_
```

```python
list_url = [f'''https://www.auveco.com/{urlcomponent.strip()}''' for urlcomponent in list_urlcomponent]

list_url
```

```python
df_temp = pd.DataFrame({'Make': input_.loc[a, 'Make'],
                        'Model': input_.loc[a, 'Model'],
                        'Year': input_.loc[a, 'Year'],
                        'SKU Count': input_.loc[a, 'SKU Count'],
                        'Page': input_.loc[a, 'Page'],
                        'Item': input_.loc[a, 'Item'],
                        'No': [input_.loc[a, 'Page']*100 - 100 + i + 1 for i in range(len(list_url))],
                        'Url': list_url})

output_correct = pd.concat([output_correct, df_temp], ignore_index=True).fillna('')

output_correct
```
