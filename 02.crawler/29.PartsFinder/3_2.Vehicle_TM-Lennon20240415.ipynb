```python
import pandas as pd
```

```python
df = pd.read_excel('./part.xlsx',
                   header=0,
                   dtype=str).fillna('')[['Url', 'Make']]

df['list_make'] = [make.split(';') for make in df['Make'].tolist()]

df = df.explode('list_make').reset_index(drop=True)

print('总数量：' + str(len(df)))
df['Url'].tolist()
```

```python
list_url = [url+'&make=' + make + '&Lennon_Index='+str(i+1) if '?' in url else url+'?make=' + make + '&Lennon_Index='+str(i+1) for url, make, i in zip(df['Url'].tolist(), df['list_make'].tolist(), range(len(df)))]

list_url
```

```python
import pyperclip
```

```python
for i in range(len(list_url)):
    if i == 0:
        data = 'let dict_url = {"' + str(i+1) + '": "' + list_url[i] + '", '
    else:
        data += '"' + str(i+1) + '": "' + list_url[i] + '", '

data = data[:-2] + '};\n'
        
pyperclip.copy(data)

data
```
