```python
import pandas as pd

df = pd.DataFrame()
```

```python
import os
```

```python
list_file = list(os.walk('./model'))[0][2]
if '.DS_Store' in list_file:
    list_file.remove('.DS_Store')
if '0.null.txt' in list_file:
    list_file.remove('0.null.txt')
    
list_file.sort()
    
len(list_file)
```

```python
from tqdm import tqdm
from bs4 import BeautifulSoup
from lxml import etree
```

```python
for a in tqdm(range(len(list_file)), desc='Progress', ncols=77):
    with open('./model/' + list_file[a], 'r', encoding='UTF-8') as file:
        resp = file.read()
        
    # = = = = = = = = = = = = = = =
    
    soup = BeautifulSoup(resp, 'lxml')
    html = etree.HTML(str(soup))
    
    # = = = = = = = = = = = = = = =
    
    list_model = html.xpath('//div[@class="compatibility__model-title"]')
    
    # = = = = = = = = = = = = = = =
    
    product_id = list_model[0].xpath('./@data-article-id')[0].strip()
    
    # = = = = = = = = = = = = = = =
    
    maker_id = list_model[0].xpath('./@data-maker-id')[0].strip()
    
    # = = = = = = = = = = = = = = =
    
    list_model_id = [model.xpath('./@data-model-id')[0].strip() for model in list_model]
    
    # = = = = = = = = = = = = = = =
    
    list_vehicle_1 = [model.xpath('./text()')[0].strip() for model in list_model]
    
    # = = = = = = = = = = = = = = =
    
    df_temp = pd.DataFrame({'No.': int(list_file[a].replace('.txt', '')),
                            'Product_Id': product_id,
                            'Maker_Id': maker_id,
                            'Model_Id': list_model_id,
                            'Vehicle_1': list_vehicle_1})

    df = pd.concat([df, df_temp], ignore_index=True).fillna('')

# = = = = = = = = = = = = = = =
    
df = df.drop_duplicates(ignore_index=True)
df = df.sort_values(by=['No.'], ascending=[True]).reset_index(drop=True)
df.to_excel('./model_1.xlsx', index=False)

df
```

```python
df = df[['Product_Id', 'Vehicle_1']]
df = df.groupby('Product_Id', as_index=False).agg({'Vehicle_1': '\n'.join})
df.to_excel('./model.xlsx', index=False)

df
```
