```python
param_amount = 2064
```

```python
list_url = []
c = 1
while len(list_url) != param_amount:
    print(str(c) + ': # = = = = = = = = = = = = = = =')

    from gevent import monkey
    monkey.patch_all(thread=False)
    from gevent.queue import Queue
    import gevent

    # = = = = = = = = = = = = = = =

    import pandas as pd

    input_ = pd.read_excel('./menu.xlsx',
                           header=0,
                           dtype=str).fillna('')

    print('总数量：' + str(len(input_)))
    print()

    # = = = = = = = = = = = = = = =

    import requests

    import sys
    sys.path.append('../00.Tools')
    import UA
    import Proxy

    # = = = = = = = = = = = = = = =

    from bs4 import BeautifulSoup
    from lxml import etree

    # = = = = = = = = = = = = = = =

    import time
    from datetime import datetime

    # = = = = = = = = = = = = = = =

    work = Queue()
    for a in range(len(input_)):
        work.put_nowait(a)

    # = = = = = = = = = = = = = = =

    def crawler():
        global list_url

        while not work.empty():
            a = work.get_nowait()

            # = = = = = = = = = = = = = = =

            crawler_status = 'error'
            for b in range(666):
                try:
                    resp = requests.get(input_.loc[a, 'Url_1'],
                                        headers=UA.get_User_Agent_Requests(),
                                        proxies=Proxy.get_Proxy_Requests()).text

                    # = = = = = = = = = = = = = = =

                    soup = BeautifulSoup(resp, 'lxml')
                    html = etree.HTML(str(soup))

                    # = = = = = = = = = = = = = = =

                    list_url += ['https://www.dormanproducts.com/' + href.strip() for href in html.xpath('//div[@class="searchItems-img"]/a/@href')]

                    list_url = sorted(list(set(list_url)))

                    # = = = = = = = = = = = = = = =

                    crawler_status = 'ok'

                    break
                except:
                    time.sleep(0.3)
                    continue

            # = = = = = = = = = = = = = = =

            print('[' + crawler_status + '] - ' + input_.loc[a, 'Url_1'] +  '\n[尝试次数：' + str(b+1) + '] - [剩余数量：' + str(work.qsize()) + '] - [当前时间：' + datetime.now().strftime('%H:%M:%S') + ']\n[第' + str(c) + '轮] - [length：' + str(len(list_url)) + ']\n')

    # = = = = = = = = = = = = = = =

    list_task = []
    for _ in range(10):
        task = gevent.spawn(crawler)
        list_task.append(task)
    gevent.joinall(list_task)
    
    c += 1
    
    # = = = = = = = = = = = = = = =

print('输出ing...')
print()
df = pd.DataFrame({'Url': list_url})
df['No.'] = [i+1 for i in range(len(df))]
df.to_excel('./url.xlsx', index=False)
print('Done ~')
```
