```python
param_url = 'https://www.rockauto.com/en/parts/melling,timing+chain+&+component+kit,5756'
```

```python
import requests

import sys
sys.path.append('../00.Tools')
import UA
import Proxy
```

```python
resp = requests.get(param_url,
                    headers=UA.get_User_Agent_Requests(),
                    proxies=Proxy.get_Proxy_Requests()).text

resp
```

```python
from bs4 import BeautifulSoup
from lxml import etree
```

```python
soup = BeautifulSoup(resp, 'lxml')
html = etree.HTML(str(soup))

# with open('./part_number.html', 'w', encoding='UTF-8') as file:
#     file.write(str(soup))

print(soup.prettify())
```

```python
manufacturer = html.xpath('//select[@id="manufacturer_partsearch_007"]/option[@selected]/text()')[0].strip()

manufacturer
```

```python
category = html.xpath('//select[@id="partgroup_partsearch_007"]/option[@selected]/text()')[0].strip()

category
```

```python
type_ = html.xpath('//select[@id="parttype_partsearch_007"]/option[@selected]/text()')[0].strip()

type_
```

```python
title = html.xpath('//div[contains(@style, "display: inline-block; padding: 0em 5em 0em 5em;")]/text()')[0]

title
```

```python
manufacturer not in title or type_ not in title or 'Part Numbers' not in title
```

```python
list_part = html.xpath('//div[@class="ranavnode"]/input[contains(@id, "jsn")]/@value')[1:]

list_part
```

```python
import json
```

```python
list_no = [int(json.loads(part)['groupindex'].strip())-1 for part in list_part]

list_no
```

```python
list_type_code = [json.loads(part)['parttype'].strip() for part in list_part]

list_type_code
```

```python
list_manufacturer = [json.loads(part)['catalogname'].strip() for part in list_part]

list_manufacturer
```

```python
list_part_number = [json.loads(part)['partnumber'].strip() for part in list_part]

list_part_number
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({'No.': list_no,
                   'Manufacturer': list_manufacturer,
                   'Category': category,
                   'Type': type_,
                   'Type_Code': list_type_code,
                   'Part_Number': list_part_number})

df = df.drop_duplicates(ignore_index=True)
df = df.sort_values(by=['No.'], ascending=[True]).reset_index(drop=True)
df.to_excel('./menu.xlsx', index=False)
    
df
```
