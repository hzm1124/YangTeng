```python
import pandas as pd

output_correct, output_error = pd.DataFrame(), pd.DataFrame()
```

```python
input_ = pd.read_excel(f'./1.page.xlsx',
                       header=0,
                       dtype=str).fillna('')

input_['Page'] = input_['Page'].astype(int)

print(f'总数量：{len(input_)}')
input_['Page Url'].tolist()
```

```python
a = 0

a
```

```python
crawler_status = 'error'

crawler_status
```

```python
import requests

import sys
sys.path.append('../00.Tools')
from crawler_configuration import get_header, get_proxy

resp = requests.get(input_.loc[a, 'Page Url'],
                    headers=get_header(),
                    proxies=get_proxy(),
                    timeout=(10, 10)).text

resp
```

```python
from bs4 import BeautifulSoup
from lxml import etree

soup = BeautifulSoup(resp.strip(), 'lxml')
html = etree.HTML(str(soup))

with open('./menu.html', 'w', encoding='UTF-8') as file:
    file.write(str(soup))

print(soup.prettify())
```

```python
list_url = [f'https://www.1aauto.com{href.strip()}' for href in html.xpath('//div[@class="desktop-product-results d-none d-lg-block"]/descendant::*/h4/a[@data-referrer="SearchResults"]/@href')]

list_url
```

```python
len(list_url) == 0
```

```python
df_temp = pd.DataFrame({'Page': input_.loc[a, 'Page'],
                        'Page Url': input_.loc[a, 'Page Url'],
                        'No': [i+1 for i in range(len(list_url))],
                        'Url': list_url})

output_correct = pd.concat([output_correct, df_temp], ignore_index=True).fillna('')

output_correct
```

```python
crawler_status = 'ok'

crawler_status
```

```python
if crawler_status == 'error':
    df_temp = pd.DataFrame([{'Page': input_.loc[a, 'Page'],
                             'Page Url': input_.loc[a, 'Page Url']}])

    output_error = pd.concat([output_error, df_temp], ignore_index=True).fillna('')

output_error
```

```python
if len(output_correct) != 0:
    output_correct = output_correct.sort_values(by=['Page', 'No'],
                                                ascending=[True, True],
                                                ignore_index=True)
    output_correct = output_correct.drop_duplicates(['Url'],
                                                    keep='first',
                                                    ignore_index=True)
    output_correct['No'] = [i+1 for i in range(len(output_correct))]
    output_correct.to_excel('./test_menu.xlsx', index=False)

output_correct
```

```python
if len(output_error) != 0:
    output_error = output_error.sort_values(by=['Page'],
                                            ascending=[True, True],
                                            ignore_index=True)
    output_error.to_excel('./test_menu_error.xlsx', index=False)

output_error
```
