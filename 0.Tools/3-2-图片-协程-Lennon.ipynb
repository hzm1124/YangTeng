# 0. 配置参数

- `file`：要下载图片的文件的路径

```python
file = './crawler-LS_Oil_Pan-Summit-20230423.xlsx'

# = = = = = = = = = = = = = = = = = =

from gevent import monkey
monkey.patch_all()
from gevent.queue import Queue
import gevent

# = = = = = = = = = = = = = = = = = =

from openpyxl import load_workbook
import requests
from io import BytesIO
from openpyxl.drawing.image import Image
from openpyxl.utils import get_column_letter

# = = = = = = = = = = = = = = = = = =

proxies = {'http': 'http://t18126899091866:kh8hjqtl@j933.kdltps.com:15818',
           'https': 'http://t18126899091866:kh8hjqtl@j933.kdltps.com:15818'}

# = = = = = = = = = = = = = = = = = =

wb = load_workbook(file)
ws = wb.active

for i in range(1, ws.max_column+1):
    if ws.cell(1, i).value == '图片链接':
        index_url = i
    elif ws.cell(1, i).value == '图片':
        index_img = i

# = = = = = = = = = = = = = = = = = =

work = Queue()
for i in range(2, ws.max_row+1):
    work.put_nowait(i)

# = = = = = = = = = = = = = = = = = =

def crawler():
    global ws
    
    while not work.empty():
        row = work.get_nowait()
        url = ws.cell(row, index_url).value
        
        for i in range(31):
            try:
                img = Image(BytesIO(requests.get(url, proxies=proxies).content))
                
                # = = = = = = = = = = = = = = = = = =
                
                if img.width >= img.height:
                    img.height = 18.0 / img.width * img.height
                    img.width = 18.0
                else:
                    img.width = 18.0 / img.height * img.width
                    img.height = 18.0
                    
                # = = = = = = = = = = = = = = = = = =
                
                ws.add_image(img, get_column_letter(index_img) + str(row))
            
                break
            
            except:
                
                continue
                
        # = = = = = = = = = = = = = = = = = =
                
        print(url, ' [' + str(i) + '] - 剩余数量：' + str(work.qsize()))

# = = = = = = = = = = = = = = = = = =

list_task = []
for _ in range(10):
    task = gevent.spawn(crawler)
    list_task.append(task)
gevent.joinall(list_task)

wb.save(file.replace('.xlsx', '-pic.xlsx'))
print()
print('搞定')
```
