```python
from selenium.webdriver import ChromeOptions
import random
```

```python
option = ChromeOptions()
option.add_experimental_option('useAutomationExtension', False)
# option.add_argument('blink-settings=imagesEnabled=false')
option.add_argument('--disable-blink-features=AutomationControlled')
option.add_argument('--disable-dev-shm-usage')
option.add_argument('--disable-gpu')
option.add_experimental_option('excludeSwitches', ['enable-automation'])
# option.add_argument('--headless')
option.add_argument('--no-sandbox')
# option.add_argument('--proxy-server=http://u431.kdltps.com:15818')
option.add_argument('user-agent=' + random.choice(list_user_agent))
option.page_load_strategy = 'eager'
```

```python
from selenium.webdriver import Chrome
from selenium.webdriver.chrome.service import Service
```

```python
browser = Chrome(service=Service('../../../chromedriver'), options=option)

browser.maximize_window()
```

```python
import time
```

```python
i = 0

browser.get(list_url[i])

time.sleep(5)
```

```python
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.common.by import By
```

```python
wait = WebDriverWait(browser, 60)
wait.until(EC.presence_of_element_located((By.XPATH, '//span[@itemprop="name"]')))
title = browser.find_element(by=By.XPATH, value='//span[@itemprop="name"]').text

title
```

```python
wait.until(EC.presence_of_element_located((By.XPATH, '//span[@class="ti-our-price"]/following-sibling::i')))
price = browser.find_element(by=By.XPATH, value='//span[@class="ti-our-price"]/following-sibling::i').text

price
```

```python
vehicle = browser.find_element(by=By.XPATH, value='//div[@id="country3-head"]/following-sibling::div[@class="pro_tab_container"]').text.replace('\nClick here for a full list \n+', '')
list_li = browser.find_elements(by=By.XPATH, value='//div[@id="country3-head"]/following-sibling::div[@class="pro_tab_container"]/descendant::li')
list_li = [_.get_attribute('textContent') for _ in list_li]
for _ in list_li:
    vehicle += '\n' + _
    
print(vehicle)
```

```python
wait.until(EC.presence_of_element_located((By.XPATH, '//img[@itemprop="image"]')))
src = browser.find_element(by=By.XPATH, value='//img[@itemprop="image"]').get_attribute('src')

src
```

```python
df_temp = pd.DataFrame([{'Title': title,
                         'Price': price,
                         'Vehicle': vehicle,
                         'Url': list_url[i],
                         'Src': src,
                         'status': 'ok'}])

df_temp
```

```python
list_name = browser.find_elements(by=By.XPATH, value='//td[@class="pro_tab_left_para even-dark"]')
list_value = browser.find_elements(by=By.XPATH, value='//td[@class="pro_tab_left_para even-dark"]/following-sibling::td[@class="pro_tab_right_para"]')

for name, value in zip(list_name, list_value):
    df_temp[name.text.lower()] = value.text
    
df_temp
```

```python
list_vehicle = browser.find_elements(by=By.XPATH, value='//div[@id="country3-head"]/following-sibling::div[@class="pro_tab_container"]')

if len(list_vehicle) == 0:
    vehicle = ''
else:
    vehicle = list_vehicle[0].text.replace('\nClick here for a full list \n+', '')
    list_li = browser.find_elements(by=By.XPATH, value='//div[@id="country3-head"]/following-sibling::div[@class="pro_tab_container"]/descendant::li')
    list_li = [_.get_attribute('textContent') for _ in list_li]
    for _ in list_li:
        vehicle += '\n' + _
    
print(vehicle)
```

```python
wait.until(EC.presence_of_element_located((By.XPATH, '//div[@id="tabDescription-content"]/descendant::tr')))
list_information = browser.find_elements(by=By.XPATH, value='//div[@id="tabDescription-content"]/descendant::tr')

for information in list_information:
    df_temp[information.find_element(by=By.XPATH, value='./td[1]').text.lower()] = information.find_element(by=By.XPATH, value='./td[2]').text
    
browser.quit()

df = pd.concat([df, df_temp], ignore_index=True).fillna('')

df
```
