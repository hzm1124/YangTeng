```python
import requests

import sys
sys.path.append('../00.Tools')
from crawler_configuration import get_header, get_proxy

b = 0
while True:
    b += 1
    print(b)
    try:
        resp = requests.get('https://ecat.spectrapremium.com/en',
                             headers=get_header(),
                             proxies=get_proxy(),
                             timeout=(10, 10))
        
        if resp.status_code == 200:
            break
    except KeyboardInterrupt:
        break
    except:
        continue

resp
```

```python
from bs4 import BeautifulSoup
from lxml import etree

soup = BeautifulSoup(resp.text, 'lxml')
html = etree.HTML(str(soup))

with open('./input.html', 'w', encoding='UTF-8') as file:
    file.write(str(soup))

print(soup.prettify())
```

```python
list_url = [f'''https://ecat.spectrapremium.com/en/parts?line={value.strip()}''' for value in html.xpath('//optgroup/option/@value')]

list_url
```

```python
import pandas as pd

output = pd.DataFrame({'No': [_+1 for _ in range(len(list_url))],
                       'Url': list_url})

output.to_excel('./file/input.xlsx', index=False)

output
```
