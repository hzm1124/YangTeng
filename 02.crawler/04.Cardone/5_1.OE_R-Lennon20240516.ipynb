```python
import pandas as pd

output_correct = pd.DataFrame()
output_error = pd.DataFrame()
```

```python
input_ = pd.read_excel('./auth.xlsx',
                       header=0,
                       dtype=str).fillna('')

print('总数量：' + str(len(input_)))
input_['Auth'].tolist()
```

```python
crawler_status = 'error'

print(crawler_status)
```

```python
import requests

import sys
sys.path.append('../00.Tools')
import UA
import Proxy
```

```python
a = 0

resp = requests.get('https://productdesk-api.cellacore.net/legacy/cardone/FinishedGood/?q=productDeskItemNumber.keyword:' + input_.loc[a, 'Part_Number'] + '&size=1&useCache=true&auth=' + input_.loc[a, 'Auth'],
                    headers=UA.get_User_Agent_Requests(),
                    proxies=Proxy.get_Proxy_Requests()).text

resp
```

```python
import json
```

```python
json_data = json.loads(resp)

json_data
```

```python
list_oe = []
for reference in json_data[0]['CrossReferences']:
    if reference['referenceType'].strip() == 'OE Part':
        list_oe.append(reference['code'].strip())

oe = ';'.join(sorted(list(set(list_oe))))

oe
```

```python
df_temp = pd.DataFrame([{'Part_Number': input_.loc[a, 'Part_Number'],
                         'OE': oe}])

output_correct = pd.concat([output_correct, df_temp], ignore_index=True).fillna('')

len(output_correct)
```

```python
crawler_status = 'ok'

print(crawler_status)
```

```python
if crawler_status == 'error':
    df_temp = pd.DataFrame([{'Part_Number': input_.loc[a, 'Part_Number'],
                             'Auth': input_.loc[a, 'Auth']}])
    
    output_error = pd.concat([output_error, df_temp], ignore_index=True).fillna('')

len(output_error)
```

```python
output_correct = output_correct.drop_duplicates(ignore_index=True)
output_correct = output_correct.sort_values(by=['Part_Number'], ascending=[True]).reset_index(drop=True)
output_correct.to_excel('./test_oe.xlsx', index=False)

output_correct
```

```python
if len(output_error) != 0:
    output_error = output_error.drop_duplicates(ignore_index=True)
    output_error = output_error.sort_values(by=['Part_Number'], ascending=[True]).reset_index(drop=True)
    output_error.to_excel('./test_oe_error.xlsx', index=False)

output_error
```
