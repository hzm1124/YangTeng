```python
param_date = '20240726'
```

```python
import requests

import sys
sys.path.append('../../00.Tools')
from crawler_configuration import get_header, get_proxy

b = 0
while True:

    while True:
        b += 1
        print(b)
        try:
            resp = requests.get('https://www.rockauto.com/en/partsearch',
                                headers=get_header(),
                                proxies=get_proxy(),
                                timeout=(10, 10))
            break
        except:
            continue

    if resp.status_code == 200:
        break

resp
```

```python
from bs4 import BeautifulSoup
from lxml import etree

soup = BeautifulSoup(resp.text, 'lxml')
html = etree.HTML(str(soup))

with open('./manufacturer.html', 'w', encoding='UTF-8') as file:
    file.write(str(soup))

print(soup.prettify())
```

```python
list_manufacturer = html.xpath('//select[@id="manufacturer_partsearch_007"]/option/@value')[1:]

list_manufacturer
```

```python
len(list_manufacturer) == 0
```

```python
import pandas as pd

output = pd.DataFrame({'Manufacturer No': [i+1 for i in range(len(list_manufacturer))],
                       'Manufacturer': list_manufacturer})

output.to_excel(f'./{param_date}/1.manufacturer.xlsx', index=False)

output
```
