```python
import pandas as pd

df_menu = pd.read_excel('./item.xlsx', header=0)
list_item = df_menu['Item'].to_list()

print(len(list_item))
list_item
```

```python
df = pd.DataFrame(columns=['Item',
                           'Category',
                           'POP',
                           'Per Car',
                           'Vehicle',
                           'Url',
                           'Src',
                           'status'])
```

```python
import requests
```

```python
i = 0

resp = requests.get('https://ecatalog.smpcorp.com/V2/STD/api/part/partselect?func=PART&part=' + str(list_item[i])).text

resp
```

```python
import json
```

```python
content = json.loads(resp)

content
```

```python
list_category = content['pp']['categories']
list_category = [category['categoryName_en'] for category in list_category]

category = ';'.join(list_category)

category
```

```python
list_vehicle = content['pp']['buyersGuides']
list_vehicle = [vehicle['buyersGuideDesc'] for vehicle in list_vehicle]

vehicle = '\n'.join(list_vehicle)

vehicle
```

```python
resp = requests.get('https://ecatalog.smpcorp.com/V2/STD/api/image/getallimages?partNum=S20001&brand=STD&zoomFactor_sm=75&zoomFactor_md=360&zoomFactor_bg=960').text

resp
```

```python
df_temp = pd.DataFrame({'Item': str(list_item[i]),
                        'Category': category,
                        'POP': content['pd']['pop'],
                        'Per Car': content['pd']['qtyEach'],
                        'Vehicle': vehicle,
                        'Url': 'https://www.standardbrand.com/en/ecatalog?partdetail=' + str(list_item[i]),
                        'Src': json.loads(resp)[0]['image_BG_Url'],
                        'status': ['ok']})

df_temp
```

```python
list_specification = content['pp']['partSpecs']
for specification in list_specification:
    df_temp[specification['attributeName_en'].lower()] = specification['siteValue_en']
    
df_temp
```

```python
df = pd.concat([df, df_temp], ignore_index=True).fillna('')

df
```
