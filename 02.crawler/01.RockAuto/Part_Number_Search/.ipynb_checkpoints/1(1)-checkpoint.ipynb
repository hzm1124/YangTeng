```python
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.common.by import By
```

```python
if len(browser.find_elements(by=By.XPATH, value='//span[contains(@class, "navlabellink") and contains(@class, "nvoffset") and contains(@class, "nnormal")]')) != 0:
    print('No Parts Found')
```

```python
wait = WebDriverWait(browser, 60)
wait.until(EC.presence_of_element_located((By.XPATH, '//tbody[contains(@class, "listing-inner") and contains(@class, "altrow-a-1")]')))
list_part = browser.find_elements(by=By.XPATH, value='//tbody[contains(@class, "listing-inner") and (contains(@class, "altrow-a-1") or contains(@class, "altrow-a-0"))]')

list_part
```

```python
list_part_manufacturer = [_.find_element(by=By.XPATH, value='./descendant::div[contains(@class, "listing-text-row-moreinfo-truck")]/span[contains(@class, "listing-final-manufacturer")]') for _ in list_part]
list_part_manufacturer = [_.text for _ in list_part_manufacturer]

list_part_manufacturer
```

```python
list_part_number = [_.find_element(by=By.XPATH, value='./descendant::div[contains(@class, "listing-text-row-moreinfo-truck")]/span[contains(@title, "Buyer\'s Guide")]') for _ in list_part]
list_part_number = [_.text for _ in list_part_number]

list_part_number
```

```python
list_info = [_.find_element(by=By.XPATH, value='./descendant::div[contains(@class, "listing-text-row-moreinfo-truck")]/a[contains(@class, "ra-btn") and contains(@class, "ra-btn-moreinfo")]') for _ in list_part]
list_info = [_.get_attribute('href') for _ in list_info]

list_info
```

```python
list_note_2 = [_.find_element(by=By.XPATH, value='./descendant::div[contains(@class, "listing-text-row") and not(contains(@class, "moreinfo-truck"))]') for _ in list_part]
list_note_2 = [_.text for _ in list_note_2]

list_note_2
```

```python
list_oe = [_.find_element(by=By.XPATH, value='./descendant::span[contains(@title, "Replaces these Alternate/ OE Part Numbers")]') for _ in list_part]
list_oe = [_.get_attribute('textContent') for _ in list_oe]
list_oe = [_.replace(', ', ';') for _ in list_oe]

list_oe
```
