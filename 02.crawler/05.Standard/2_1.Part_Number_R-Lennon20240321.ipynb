```python
import pandas as pd

output_correct = pd.DataFrame()
output_error = pd.DataFrame()
```

```python
input_ = pd.read_excel('./menu.xlsx',
                       header=0,
                       dtype=str).fillna('')

print('总数量：' + str(len(input_)))
input_['Url'].tolist()
```

```python
crawler_status = 'error'

crawler_status
```

```python
import requests

import sys
sys.path.append('../00.Tools')
import UA
import Proxy
```

```python
a = 0

resp = requests.get(input_.loc[a, 'Url'],
                    headers=UA.get_User_Agent_Requests(),
                    proxies=Proxy.get_Proxy_Requests()).text

resp
```

```python
import json
```

```python
output_correct = pd.concat([output_correct, pd.DataFrame(json.loads(resp))], ignore_index=False).fillna('')

len(output_correct)
```

```python
crawler_status = 'ok'

crawler_status
```

```python
if crawler_status == 'error':
    df_temp = pd.DataFrame([{'No.': int(input_.lenoc[a, 'No.']),
                             'Url': input_.lenoc[a, 'Url']}])
    
    output_error = pd.concat([output_error, df_temp], ignore_index=True).fillna('')

len(output_error)
```

```python
output_correct = output_correct.drop_duplicates(ignore_index=True)
output_correct = output_correct.sort_values(by=['rowId'], ascending=[True]).reset_index(drop=True)
output_correct.to_excel('./test_part_number.xlsx', index=False)

output_correct
```

```python
if len(output_error) != 0:
    output_error = output_error.drop_duplicates(ignore_index=True)
    output_error = output_error.sort_values(by=['No.'], ascending=[True]).reset_index(drop=True)
    output_error.to_excel('./test_part_number_error.xlsx', index=False)

output_error
```
