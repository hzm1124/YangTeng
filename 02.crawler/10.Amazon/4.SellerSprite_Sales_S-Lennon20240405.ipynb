```python
from datetime import datetime
```

```python
for a in range(len(input_)):
    crawler_status = 'error'
    
    # = = = = = = = = = = = = = = =
    
    browser.get('https://www.sellersprite.com/v2/tools/sales-estimator?market=US&asin=' + input_.loc[a, 'ASIN'])
    
    # = = = = = = = = = = = = = = =
    
    list_td = browser.find_elements(by=By.XPATH, value='//div[@id="container-sales-overflow"]/table/tbody/tr[1]/td')
    list_td_2 = browser.find_elements(by=By.XPATH, value='//div[@id="container-sales-overflow"]/table/tbody/tr[2]/td')

    # = = = = = = = = = = = = = = =
    
    not_found = browser.find_elements(by=By.XPATH, value='//div[@class="bg-white py-5 px-4 no-collection"]/p')

    # = = = = = = = = = = = = = = =
    
    for b in range(100):
        if len(list_td) == 0 and len(not_found) == 0:
            time.sleep(1)
            list_td = browser.find_elements(by=By.XPATH, value='//div[@id="container-sales-overflow"]/table/tbody/tr[1]/td')
            list_td_2 = browser.find_elements(by=By.XPATH, value='//div[@id="container-sales-overflow"]/table/tbody/tr[2]/td')
            not_found = browser.find_elements(by=By.XPATH, value='//div[@class="bg-white py-5 px-4 no-collection"]/p')
        else:
            break
    
    if len(list_td) == 0 and len(not_found) == 0:
        df_temp = pd.DataFrame([{'ASIN': input_.loc[a, 'ASIN']}])
        output_error = pd.concat([output_error, df_temp], ignore_index=True).fillna('') 
        print('[' + crawler_status + '] - ' + input_.loc[a, 'ASIN'] + '\n[等待时间：' + str(b+1) + '] - [完成数量：' + str(a+1) + '] - [当前时间：' + datetime.now().strftime('%H:%M:%S') + ']\n')
        continue
    
    # = = = = = = = = = = = = = = =
    
    if len(not_found) != 0:
        df_temp = pd.DataFrame([{'ASIN': input_.loc[a, 'ASIN']}])
        output_correct = pd.concat([output_correct, df_temp], ignore_index=True).fillna('') 
        print('[ok] - ' + input_.loc[a, 'ASIN'] + '\n[等待时间：' + str(b+1) + '] - [完成数量：' + str(a+1) + '] - [当前时间：' + datetime.now().strftime('%H:%M:%S') + ']\n')
        continue
        
    # = = = = = = = = = = = = = = =
    
    df_temp = pd.DataFrame([{'ASIN': input_.loc[a, 'ASIN']}])

    for td, td_2 in zip(list_td, list_td_2):
        df_temp.loc[0, td.text.strip()] = td_2.text.strip()

    output_correct = pd.concat([output_correct, df_temp], ignore_index=True).fillna('') 

    # = = = = = = = = = = = = = = =
    
    crawler_status = 'ok'
    
    print('[' + crawler_status + '] - ' + input_.loc[a, 'ASIN'] + '\n[等待时间：' + str(b+1) + '] - [完成数量：' + str(a+1) + '] - [当前时间：' + datetime.now().strftime('%H:%M:%S') + ']\n')
    
    # = = = = = = = = = = = = = = =
    
browser.quit()

print('输出ing...')
output_correct = output_correct.drop_duplicates(ignore_index=True)
output_correct = output_correct.sort_values(by=['ASIN'], ascending=[True]).reset_index(drop=True)
output_correct.to_excel('./sellersprite_sales-' + datetime.now().strftime('%Y%m%d_%H%M%S') + '.xlsx', index=False)
print('Done ~')
```
