# 0. 配置参数

- `price`：改价表的相关信息
- `alternate`：替代料总表路径
- `out`：保存结果的路径

```python
dict_price = {'file_name': '../../file/0203/更新数据/改价表 2-3.xlsx',
              'sheet_name': 'YDK2库存 02-03',
              'col_name': '商品编码 2-3'}
alternate = '../../file/0203/更新数据/替代料总表.xlsx'

out = '../../file/Lennon_US库存匹配_NOW().xlsx'
```

```python
import pandas as pd
import re
```

```python
df_price = pd.read_excel(dict_price['file_name'], header=0, sheet_name=dict_price['sheet_name'])[[dict_price['col_name'], '可售']]
df_price = df_price[df_price['可售'] != 0].reset_index(drop=True)

list_sku = df_price[dict_price['col_name']].tolist()
for i in range(len(list_sku)):
    list_sku[i] = re.sub(r'^USCAB3376KITD$', 'USCAB3376KitD', list_sku[i])
    list_sku[i] = re.sub(r'-N$', '', list_sku[i])

df_price['商品编码'] = list_sku
df_price = df_price[['商品编码', '可售']].groupby(['商品编码'], as_index=False).sum()

df_price
```

```python
df_alternate_mrp = pd.read_excel(alternate, sheet_name='MRP SKU库存', header=0).fillna('')

df_alternate_mrp
```

```python
df_alternate_NOTmrp = pd.read_excel(alternate, sheet_name='MRP外SKU', header=0).fillna('')

df_alternate_NOTmrp
```

```python
df_alternate_mrp = pd.merge(left=df_alternate_mrp, right=df_price, how='left', left_on='主料SKU', right_on='商品编码').fillna(0)
df_alternate_mrp = pd.merge(left=df_alternate_mrp, right=df_price, how='left', left_on='替换料1', right_on='商品编码', suffixes=('', '_1')).fillna(0)
df_alternate_mrp = pd.merge(left=df_alternate_mrp, right=df_price, how='left', left_on='替换料2', right_on='商品编码', suffixes=('', '_2')).fillna(0)
df_alternate_mrp = pd.merge(left=df_alternate_mrp, right=df_price, how='left', left_on='替换料3', right_on='商品编码', suffixes=('', '_3')).fillna(0)
df_alternate_mrp = pd.merge(left=df_alternate_mrp, right=df_price, how='left', left_on='替换料4', right_on='商品编码', suffixes=('', '_4')).fillna(0)
df_alternate_mrp = pd.merge(left=df_alternate_mrp, right=df_price, how='left', left_on='替换料5', right_on='商品编码', suffixes=('', '_5')).fillna(0)
df_alternate_mrp = pd.merge(left=df_alternate_mrp, right=df_price, how='left', left_on='替换料6', right_on='商品编码', suffixes=('', '_6')).fillna(0)

df_alternate_mrp
```

```python
df_alternate_mrp['库存'] = df_alternate_mrp['可售'] + df_alternate_mrp['可售_1'] + df_alternate_mrp['可售_2'] + df_alternate_mrp['可售_3'] + df_alternate_mrp['可售_4'] + df_alternate_mrp['可售_5'] + df_alternate_mrp['可售_6']
df_alternate_mrp = df_alternate_mrp[['主料SKU', '替换料1', '替换料2', '替换料3', '替换料4', '替换料5', '替换料6', 'PT SKU', '数量', 'SKU1', 'SKU2', 'SKU3', 'SKU4', 'SKU5', '库存']]

df_alternate_mrp
```

```python
df_combination = df_alternate_mrp[(df_alternate_mrp['数量'] != '') & (df_alternate_mrp['数量'] != 1)].reset_index(drop=True)[['主料SKU', '数量', 'SKU1', 'SKU2', 'SKU3', 'SKU4', 'SKU5', '库存']]

df_combination
```

```python
df_combination = pd.merge(left=df_combination, right=df_alternate_mrp[['主料SKU', '数量', '库存']], how='left', left_on='SKU1', right_on='主料SKU', suffixes=('', '_1')).fillna(0)
df_combination = pd.merge(left=df_combination, right=df_alternate_mrp[['主料SKU', '数量', '库存']], how='left', left_on='SKU2', right_on='主料SKU', suffixes=('', '_2')).fillna(0)
df_combination = pd.merge(left=df_combination, right=df_alternate_mrp[['主料SKU', '数量', '库存']], how='left', left_on='SKU3', right_on='主料SKU', suffixes=('', '_3')).fillna(0)
df_combination = pd.merge(left=df_combination, right=df_alternate_mrp[['主料SKU', '数量', '库存']], how='left', left_on='SKU4', right_on='主料SKU', suffixes=('', '_4')).fillna(0)
df_combination = pd.merge(left=df_combination, right=df_alternate_mrp[['主料SKU', '数量', '库存']], how='left', left_on='SKU5', right_on='主料SKU', suffixes=('', '_5')).fillna(0)

df_combination
```

```python
list_combined_stock = []
for i in range(len(df_combination)):
    list_stock = []
    list_amount = str(df_combination.loc[i, '数量']).split('+')
    for j in range(len(list_amount)):
        list_stock.append(df_combination.loc[i, '库存_' + str(j+1)] / int(list_amount[j]))
    list_combined_stock.append(int(min(list_stock)))

df_combination['组合库存'] = list_combined_stock

df_combination
```

```python
df_alternate_mrp = pd.merge(left=df_alternate_mrp, right=df_combination[['主料SKU', '组合库存']], how='left', left_on='主料SKU', right_on='主料SKU').fillna(0)

df_alternate_mrp
```

```python
df_alternate_NOTmrp = pd.read_excel(alternate, sheet_name='MRP外SKU', header=0).fillna('')

df_alternate_NOTmrp
```

```python
df_alternate_NOTmrp['库存'] = [0] * len(df_alternate_NOTmrp)
df_alternate_NOTmrp = pd.merge(left=df_alternate_NOTmrp, right=df_alternate_mrp[['主料SKU', '库存']], how='left', left_on='SKU1', right_on='主料SKU', suffixes=('', '_1')).fillna(0)
df_alternate_NOTmrp = pd.merge(left=df_alternate_NOTmrp, right=df_alternate_mrp[['主料SKU', '库存']], how='left', left_on='SKU2', right_on='主料SKU', suffixes=('', '_2')).fillna(0)
df_alternate_NOTmrp = pd.merge(left=df_alternate_NOTmrp, right=df_alternate_mrp[['主料SKU', '库存']], how='left', left_on='SKU3', right_on='主料SKU', suffixes=('', '_3')).fillna(0)
df_alternate_NOTmrp = pd.merge(left=df_alternate_NOTmrp, right=df_alternate_mrp[['主料SKU', '库存']], how='left', left_on='SKU4', right_on='主料SKU', suffixes=('', '_4')).fillna(0)
df_alternate_NOTmrp = pd.merge(left=df_alternate_NOTmrp, right=df_alternate_mrp[['主料SKU', '库存']], how='left', left_on='SKU5', right_on='主料SKU', suffixes=('', '_5')).fillna(0)
df_alternate_NOTmrp = pd.merge(left=df_alternate_NOTmrp, right=df_alternate_mrp[['主料SKU', '库存']], how='left', left_on='SKU6', right_on='主料SKU', suffixes=('', '_6')).fillna(0)

df_alternate_NOTmrp
```

```python
list_combined_stock = []
for i in range(len(df_alternate_NOTmrp)):
    list_stock = []
    list_amount = str(df_alternate_NOTmrp.loc[i, '数量']).split('+')
    for j in range(len(list_amount)):
        list_stock.append(df_alternate_NOTmrp.loc[i, '库存_' + str(j+1)] / int(list_amount[j]))
    list_combined_stock.append(int(min(list_stock)))

df_alternate_NOTmrp['组合库存'] = list_combined_stock

df_alternate_NOTmrp
```

```python
list_columns = ['主料SKU', '替换料1', '替换料2', '替换料3', '替换料4', '替换料5', '替换料6', 'PT SKU', '数量', 'SKU1', 'SKU2', 'SKU3', 'SKU4', 'SKU5', '库存', '组合库存']
df = pd.concat([df_alternate_mrp[list_columns], df_alternate_NOTmrp[list_columns]]).reset_index(drop=True)
df['总库存'] = df['库存'] + df['组合库存']

df
```

```python
import datetime

writer = pd.ExcelWriter(out.replace('NOW()', datetime.datetime.now().strftime("%Y%m%d_%H%M%S")))
df.to_excel(writer, index=False)
writer.close()
```
