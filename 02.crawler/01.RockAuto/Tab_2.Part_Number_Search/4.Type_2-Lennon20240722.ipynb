```python
param_date = '20240726'
```

```python
import os

os.makedirs(f'./{param_date}/4.type_2')
os.makedirs(f'./{param_date}/5.part_number')
os.makedirs(f'./{param_date}/5.part_number/0.error')
```

```python
import pandas as pd

input_ = pd.read_excel(f'./{param_date}/3.type.xlsx',
                       header=0,
                       dtype=str).fillna('')

input_['Manufacturer No'] = input_['Manufacturer No'].astype(int)
input_['Group No'] = input_['Group No'].astype(int)
input_['Type No'] = input_['Type No'].astype(int)

print(f'总数量：{len(input_)}')
input_['Type'].tolist()
```

```python
input_2 = input_[['Group']].copy()
input_2 = input_2.drop_duplicates(ignore_index=True)
input_2 = input_2.sort_values(by=['Group'],
                              ascending=[True],
                              ignore_index=True)
input_2['Group No'] = [i+1 for i in range(len(input_2))]
input_2['File Group'] = [group.replace('/', '%2F').replace('\\', '%5C').strip() for group in input_2['Group'].tolist()]

input_2
```

```python
from tqdm import tqdm

for i in tqdm(range(len(input_2)), desc='Progress', ncols=77):
    output = input_[input_['Group'] == input_2.loc[i, 'Group']].reset_index(drop=True)
    output['Group No'] = input_2.loc[i, 'Group No']
    output = output.sort_values(by=['Type', 'Type Code', 'Manufacturer'],
                                ascending=[True, True, True],
                                ignore_index=True)

    # = = = = = = = = = = = = = = =

    output.loc[0, 'Type No'], output.loc[0, 'Manufacturer No'] = 1, 1
    for j in range(1, len(output)):
        if output.loc[j, 'Type Code'] == output.loc[j-1, 'Type Code']:
            output.loc[j, 'Type No'] = output.loc[j-1, 'Type No']
        else:
            output.loc[j, 'Type No'] = output.loc[j-1, 'Type No'] + 1

        if output.loc[j, 'Type No'] == output.loc[j-1, 'Type No']:
            output.loc[j, 'Manufacturer No'] = output.loc[j-1, 'Manufacturer No'] + 1
        else:
            output.loc[j, 'Manufacturer No'] = 1

    # = = = = = = = = = = = = = = =

    output = output[['Group No', 'Group', 'Type No', 'Type', 'Type Code', 'Manufacturer No', 'Manufacturer', 'Tab 2 Url']]
    output.to_excel(f'''./{param_date}/4.type_2/group_{input_2.loc[i, 'Group No']}.{input_2.loc[i, 'File Group']}.xlsx''', index=False)

print('Done ~')
```
