```python
import pandas as pd
```

```python
df = pd.read_excel('./manufacturer.xlsx',
                   header=0,
                   dtype=str).fillna('')

list_manufacturer = df['Manufacturer'].tolist()
list_manufacturer_number = df['No.'].tolist()

print('总数量：' + str(len(list_manufacturer)))
list_manufacturer
```

```python
df = pd.DataFrame()

df
```

```python
import requests
import json
```

```python
import sys
sys.path.append('../../../00.module')
import UA
import Proxy

i = 0

manufacturer = list_manufacturer[i]
manufacturer_number = int(list_manufacturer_number[i])

data = 'func=partsearchgetpartgroupsdd&mfr=' + manufacturer.replace('+', '%2B').replace(' ', '+').replace('&', '%26').replace('/', '%2F').replace(',', '%2C').replace('Ö', '%C3%96').replace('Ü', '%C3%9C') + '&id=partsearch_007&api_json_request=1&sctchecked=1&scbeenloaded=true&curCartGroupID='

ua = UA.get_User_Agent_Requests()
ua['Content-Type'] = 'application/x-www-form-urlencoded; charset=UTF-8'

resp = requests.post('https://www.rockauto.com/catalog/catalogapi.php',
                     data=data,
                     headers=ua,
                     proxies=Proxy.get_Proxy_Requests()).text

resp = json.loads(resp)

resp
```

```python
from bs4 import BeautifulSoup
```

```python
html = resp['html_replace_sections']['ddrepl[partgroup_partsearch_007][container]']

soup = BeautifulSoup(html, 'lxml')

with open('./category.html', 'w', encoding='UTF-8') as file:
    file.write(str(soup))

print(soup.prettify())
```

```python
from lxml import etree
```

```python
html = etree.HTML(str(soup))

list_category = html.xpath('//select[@id="partgroup_partsearch_007"]/option/text()')

list_category
```

```python
list_category_code = html.xpath('//select[@id="partgroup_partsearch_007"]/option/@value')

list_category_code
```

```python
df_temp = pd.DataFrame({'status': 'ok',
                        'Manufacturer_No.': manufacturer_number,
                        'Manufacturer': manufacturer,
                        'Category_No.': [_+1 for _ in range(len(list_category)-1)],
                        'Category': list_category[1:],
                        'Category_Code': list_category_code[1:]})

df = pd.concat([df, df_temp], ignore_index=True).fillna('')

df
```
