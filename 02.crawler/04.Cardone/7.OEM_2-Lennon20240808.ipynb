```python
import os

os.makedirs('./7.OEM_2')
os.makedirs('./8.vehicle')
os.makedirs('./8.vehicle/0.error')
```

```python
import pandas as pd

input_ = pd.read_excel('./6.OEM.xlsx',
                       header=0,
                       dtype=str).fillna('')

input_['No'] = input_['No'].astype(int)

print(f'总数量：{len(input_)}')
input_['Part Number'].tolist()
```

```python
input_2 = input_.drop_duplicates(['Type'],
                                 keep='first',
                                 ignore_index=True)
input_2 = input_2.sort_values(by=['Type'],
                              ascending=[True],
                              ignore_index=True)

input_2['File Type'] = [type_.replace('/', '%2F').replace('\\', '%5C').strip() for type_ in input_2['Type'].tolist()]

input_2
```

```python
from tqdm import tqdm

for i in tqdm(range(len(input_2)), desc='Progress', ncols=77):
    output = input_[input_['Type'] == input_2.loc[i, 'Type']].reset_index(drop=True)
    output = output.sort_values(by=['No'],
                                ascending=[True],
                                ignore_index=True)

    # = = = = = = = = = = = = = = =

    list_output, index_start, index_end = [], 0, 300
    while True:
        if index_end < len(output):
            list_output.append(output[index_start:index_end].reset_index(drop=True))
            index_start, index_end = index_end, index_end + 300
        else:
            list_output.append(output[index_start:].reset_index(drop=True))
            break

    if len(list_output) == 1:
        output.to_excel(f'''./7.OEM_2/type_{i+1}.{input_2.loc[i, 'File Type']}.xlsx''', index=False)
    else:
        for j, output in zip(range(len(list_output)), list_output):
            output.to_excel(f'''./7.OEM_2/type_{i+1}.{input_2.loc[i, 'File Type']}-serial_{j+1}.xlsx''', index=False)

print('Done ~')
```
