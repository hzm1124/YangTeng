# 0. 配置参数

- `ajax`：AJAX(ROCKAUTO)
- `rockauto`：ROCKAUTO.COM

```python
ajax = 'https://www.rockauto.com/catalog/catalogapi.php'

rockauto = 'https://www.rockauto.com'
```

```python
headers = {'Content-Type': 'application/x-www-form-urlencoded; charset=UTF-8'}

data = {'func': 'navnode_fetch',
        'payload': {'jsn': {'tab': 'catalog',
                            'make': 'ABARTH'}},
        'api_json_request': '1'}
```

```python
import json
```

```python
data['payload'] = json.dumps(data['payload'])

data
```

```python
import requests
```

```python
resp = requests.post(ajax, headers=headers, data=data).text

resp
```

```python
html = json.loads(resp)['html_fill_sections']['navchildren[]']

html
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html, 'lxml')

print(soup.prettify())
```

```python
with open('./year.html', 'w', encoding='utf-8') as file:
    file.write(str(soup))
```

```python
from lxml import etree
```

```python
html = etree.HTML(str(soup))
```

```python
list_year = html.xpath('//td[@class="nlabel"]/a/text()')

list_year
```

```python
list_url = [rockauto + i for i in html.xpath('//td[@class="niconspace ncollapsedicon"]/a/@href')]

list_url
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({'year': list_year,
                   'url': list_url})

df.to_excel('./menu-year.xlsx', index=False)

df
```
