# 1. Manufacturer分割

```python
import pandas as pd
```

```python
df = pd.read_excel('./3.type.xlsx',
                   header=0,
                   dtype=str).fillna('')

list_manufacturer = list(set(df['Manufacturer'].tolist()))
list_manufacturer.sort()

print('总数量：' + str(len(list_manufacturer)))
list_manufacturer
```

```python
import os
```

```python
param_dir = 'data_20240207'

for manufacturer in list_manufacturer:
    os.makedirs('./' + param_dir + '/' + manufacturer.replace('/', '~'))
    df_temp = df[df['Manufacturer']==manufacturer]
    df_temp.to_excel('./' + param_dir + '/' + manufacturer.replace('/', '~') + '/3.type.xlsx', index=False)
    os.makedirs('./' + param_dir + '/' + manufacturer.replace('/', '~') + '/4.part_number')
```

# 3. 数量统计

```python
import pandas as pd
```

```python
output = pd.DataFrame(columns=['JOIN_MT', 'Amount'])

output
```

```python
import os
```

```python
param_dir = 'data_20240207'

list_dir = list(os.walk('./' + param_dir))[0][1]
if '.DS_Store' in list_dir:
    list_dir.remove('.DS_Store')
if '0.null.txt' in list_dir:
    list_dir.remove('0.null.txt')
    
list_dir.sort()
    
print(len(list_dir))
list_dir
```

```python
for dir_ in list_dir:
    list_file = list(os.walk('./' + param_dir + '/' + dir_ + '/4.part_number'))[0][2]
    if '.DS_Store' in list_file:
        list_file.remove('.DS_Store')
    if '0.null.txt' in list_file:
        list_file.remove('0.null.txt')
    
    list_file.sort()
    
    for file in list_file:
        df = pd.read_excel('./' + param_dir + '/' + dir_ + '/4.part_number/' + file,
                           header=0,
                           dtype=str).fillna('')
        if len(df) > 0:
            df_temp = pd.DataFrame([{'JOIN_MT': df.loc[0, 'Manufacturer'] + ';' + df.loc[0, 'Type'],
                                     'Amount': len(df)}])
            output = pd.concat([output, df_temp], ignore_index=True).fillna('')
        else:
            os.remove('./' + param_dir + '/' + dir_ + '/4.part_number/' + file)
        
output.to_excel('amount.xlsx', index=False)

output
```
