```python
soup = BeautifulSoup(response.text, 'html5lib')

with open('./5.specification.html', 'w', encoding='utf-8') as file:
    file.write(str(soup))

print(soup.prettify())
```

```python
from parsel import Selector
```

```python
selector = Selector(text=str(soup))

selector
```

```python
dict_src, list_src = {}, selector.xpath('//ul[@class="MobileHide DesktopPDPImagesContainer"]/li[@class="DesktopPDPImages"]/a/@href').getall()
for i, src in enumerate(list_src):
    dict_src[str(i)] = src.strip()

dict_src
```

```python
dict_key, list_div = {}, selector.xpath('//div[@id="tab-specs"]/div[@class="pdmfld"]/div[@class="productView-table general" or @class="productView-table technical"]/div[@class="productView-table-row"]')
for div in list_div:
    key = div.xpath('./div[@class="productView-table-header"]/text()').get().removeprefix('pdm:').strip().removesuffix(':').strip()

    if key in dict_key:
        dict_key[key] += 1
    else:
        dict_key[key] = 1

dict_specification, dict_count = {}, {}
for div in list_div:
    key, value = div.xpath('./div[@class="productView-table-header"]/text()').get().removeprefix('pdm:').strip().removesuffix(':').strip(), div.xpath('./div[@class="productView-table-data"]/text()').get().strip()

    if dict_key[key] == 1:
        dict_specification[key] = 'nan' if value == '' else value
    else:
        if key in dict_count:
            dict_count[key] += 1
        else:
            dict_count[key] = 1
        dict_specification[f'{key}_{dict_count[key]}'] = 'nan' if value == '' else value

dict_specification
```
