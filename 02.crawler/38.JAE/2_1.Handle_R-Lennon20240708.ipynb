```python
import pandas as pd

output_correct, output_error = pd.DataFrame(), pd.DataFrame()
```

```python
input_ = pd.read_excel(f'./1.menu.xlsx',
                       header=0,
                       dtype=str).fillna('')

input_['Page'] = input_['Page'].astype(int)

print(f'总数量：{len(input_)}')
input_['Page'].tolist()
```

```python
a = 0

a
```

```python
crawler_status = 'error'

crawler_status
```

```python
import requests

import sys
sys.path.append('../00.Tools')
from crawler_configuration import get_header, get_proxy

resp = requests.get(input_.loc[a, 'Page Url'],
                    headers=get_header(),
                    # proxies=get_proxy()).text
                    timeout=(10, 10)).text

resp
```

```python
from bs4 import BeautifulSoup
from lxml import etree

soup = BeautifulSoup(resp.strip(), 'lxml')
html = etree.HTML(str(soup))

with open('./handle.html', 'w', encoding='UTF-8') as file:
    file.write(str(soup))

print(soup.prettify())
```

```python
list_script = [text.strip() for text in html.xpath('//script[@data-product-compare-data]/text()')]

list_script
```

```python
import json

list_json = [json.loads(script.strip()) for script in list_script]

list_json
```

```python
len(list_json) == 0
```

```python
df_temp = pd.DataFrame({'Page': input_.loc[a, 'Page'],
                        'Page Url': input_.loc[a, 'Page Url'],
                        'No': [i+1 for i in range(len(list_json))],
                        'Handle': [f'''https://jaeautoelectrical.com.au/products/{json_['handle'].strip()}.js''' for json_ in list_json]})

output_correct = pd.concat([output_correct, df_temp], ignore_index=True).fillna('')

output_correct
```

```python
crawler_status = 'ok'

crawler_status
```

```python
if crawler_status == 'error':
    df_temp = pd.DataFrame([{'Page': input_.loc[a, 'Page'],
                             'Page Url': input_.loc[a, 'Page Url']}])

    output_error = pd.concat([output_error, df_temp], ignore_index=True).fillna('')

output_error
```

```python
if not output_correct.empty:
    output_correct = output_correct.sort_values(by=['Page', 'No'],
                                                ascending=[False, True],
                                                ignore_index=True)
    output_correct.to_excel('./test_url.xlsx', index=False)

output_correct
```

```python
if not output_error.empty:
    output_error = output_error.sort_values(by=['Page'],
                                            ascending=[False],
                                            ignore_index=True)
    output_error.to_excel('./test_url_error.xlsx', index=False)

output_error
```
