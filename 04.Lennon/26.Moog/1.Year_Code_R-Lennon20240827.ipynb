```python
import requests

import sys
sys.path.append('../00.Tools')
from crawler_configuration import get_header, get_proxy

b = 0
while True:
    b += 1
    print(b)
    try:
        resp = requests.get('https://www.moogparts.com/driv/partfinder/api.catalog.years?brand=moog&locale=en_US&country_code=US&vehicle_group_ids=2',
                            headers=get_header(),
                            proxies=get_proxy(),
                            timeout=(10, 10))
        
        if resp.status_code == 200:
            json_ = resp.json()
            
            list_year_code = [year['id'] for year in json_['years']]
            if list_year_code:
                break
    except KeyboardInterrupt:
        break
    except:
        continue

resp
```

```python
import pandas as pd

output = pd.DataFrame({'Vehicle Type Code': 2,
                       'Year Code': list_year_code})

output
```
