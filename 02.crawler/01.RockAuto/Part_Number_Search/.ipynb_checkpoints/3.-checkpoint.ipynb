```python
import pandas as pd
```

```python
df = pd.read_excel('./menu.xlsx').fillna('')

df
```

```python
for i in range(len(df)):
    if df.loc[i, 'Manufacturer'] == '':
        df.loc[i, 'new_Url'] = ''
    else:
        df.loc[i, 'new_Url'] = 'https://www.rockauto.com/en/parts/' + df.loc[i, 'Manufacturer'].lower().replace(' ', '+') + ',' + df.loc[i, 'Part_Number'] + ',fuel+tank,6268'
        
df
```

```python
df.to_excel('new_url.xlsx', index=False)
```
