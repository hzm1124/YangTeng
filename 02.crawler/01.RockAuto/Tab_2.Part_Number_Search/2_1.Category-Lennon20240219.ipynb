```python
import pandas as pd

output_correct = pd.DataFrame()

output_error = pd.DataFrame()
```

```python
input_ = pd.read_excel('./1.manufacturer.xlsx',
                       header=0,
                       dtype=str).fillna('')

length = len(input_)

print('总数量：' + str(length))
input_['Manufacturer'].tolist()
```

```python
crawler_status = 'error'

print(crawler_status)
```

```python
import requests
import json
```

```python
i = 0

payload = {'func': 'partsearchgetpartgroupsdd',
           'mfr': input_.loc[i, 'Manufacturer'],
           'api_json_request': '1'}

resp = requests.post('https://www.rockauto.com/catalog/catalogapi.php',
                     data=payload).text

resp = json.loads(resp)

resp
```

```python
from bs4 import BeautifulSoup
```

```python
html = resp['html_replace_sections']['ddrepl[partgroup_][container]']

soup = BeautifulSoup(html, 'lxml')

with open('./category.html', 'w', encoding='UTF-8') as file:
    file.write(str(soup))

print(soup.prettify())
```

```python
from lxml import etree
```

```python
html = etree.HTML(str(soup))

security_code = html.xpath('//div[@class="center"]/div')

security_code
```

```python
len(security_code) != 0 and 'security code' in security_code[0].text
```

```python
list_category = [category.strip() for category in html.xpath('//select[@id="partgroup_"]/option/text()')[1:]]

list_category
```

```python
df_temp = pd.DataFrame({'Tab_Label': input_.loc[i, 'Tab_Label'],
                        'No.': int(input_.loc[i, 'No.']),
                        'Manufacturer': input_.loc[i, 'Manufacturer'],
                        'Category': list_category})

output_correct = pd.concat([output_correct, df_temp], ignore_index=True).fillna('')

output_correct
```
