```python
import os

list_file = sorted(list(os.walk('./file/5.type'))[0][2])
while '.DS_Store' in list_file:
    list_file.remove('.DS_Store')
while '0.null.txt' in list_file:
    list_file.remove('0.null.txt')

list_file
```

```python
import pandas as pd
from tqdm import tqdm

output = pd.DataFrame()
for file in tqdm(list_file, desc='Progress', ncols=77):
    input_ = pd.read_excel(f'./file/5.type/{file}',
                           header=0,
                           dtype=str).fillna('')

    input_['Year'] = input_['Year'].astype(int)

    output = pd.concat([output, input_], ignore_index=True).fillna('')

output = output.drop_duplicates(['Type', 'Url'],
                                keep='first',
                                ignore_index=True)
output = output.sort_values(by=['Type'],
                            ascending=[True],
                            ignore_index=True)

output['No'] = [i+1 for i in range(len(output))]
output[['No', 'Url']].to_excel('./file/input.xlsx', index=False)

output
```
