```python
import requests

import sys
sys.path.append('../../00.Tools')
import UA
import Proxy
```

```python
a = 0

resp = requests.get(input_.loc[a, 'Tab_Url'],
                    headers=UA.get_User_Agent_Requests(),
                    proxies=Proxy.get_Proxy_Requests()).text

resp
```

```python
from bs4 import BeautifulSoup
from lxml import etree
```

```python
soup = BeautifulSoup(resp, 'lxml')
html = etree.HTML(str(soup))

with open('./' + param_date + '/' + input_.loc[a, 'Type_Code'] + '/part_number.html', 'w', encoding='UTF-8') as file:
    file.write(str(soup))

print(soup.prettify())
```

```python
import re
```

```python
title = html.xpath('//div[contains(@style, "display: inline-block; padding: 0em 5em 0em 5em;")]/text()')[0].strip()

title = re.sub(' +', ' ', title)

title
```

```python
title == input_.loc[a, 'Manufacturer'] + ' ' + input_.loc[a, 'Type'] + ' Part Numbers'
```

```python
list_part = html.xpath('//input[contains(@id, "jsn")]/@value')[1:]

list_part
```

```python
import json
```

```python
list_part_number = [json.loads(part)['partnumber'].strip() for part in list_part]

list_part_number
```

```python
import os
```

```python
df = pd.DataFrame({'Tab_No.': 2,
                   'Tab': 'Part Number Search',
                   'Manufacturer_No.': int(input_.loc[a, 'Manufacturer_No.']),
                   'Manufacturer': input_.loc[a, 'Manufacturer'],
                   'Category_No.': int(input_.loc[a, 'Category_No.']),
                   'Category': input_.loc[a, 'Category'],
                   'Type_No.': int(input_.loc[a, 'Type_No.']),
                   'Type': input_.loc[a, 'Type'],
                   'Type_Code': input_.loc[a, 'Type_Code'],
                   'Tab_Url': input_.loc[a, 'Tab_Url'],
                   'Part_No.': [i+1 for i in range(len(list_part))],
                   'Part_Number': list_part_number})

df = df.drop_duplicates(ignore_index=True)
df = df.sort_values(by=['Manufacturer_No.', 'Category_No.', 'Type_No.', 'Part_No.'], ascending=[True, True, True, True]).reset_index(drop=True)
if not os.path.exists('./' + param_date + '/' + input_.loc[a, 'Type_Code']):
    os.makedirs('./' + param_date + '/' + input_.loc[a, 'Type_Code'])
df.to_excel('./test_part_number.xlsx', index=False)

df
```
