```python
import os
```

```python
list_file = list(os.walk('./union'))[0][2]
if '.DS_Store' in list_file:
    list_file.remove('.DS_Store')
if '0.null.txt' in list_file:
    list_file.remove('0.null.txt')
    
len(list_file)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame()

df
```

```python
from tqdm import tqdm
```

```python
for file in tqdm(list_file, desc='进度', ncols=77):
    df = pd.concat([df, pd.read_excel('./union/' + file)], ignore_index=False)

df = df.drop_duplicates(ignore_index=True)
    
df.to_excel('./crawler.xlsx', index=False)

df
```
