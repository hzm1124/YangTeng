```python
import requests

import sys
sys.path.append('../../00.Tools')
import UA
import Proxy
```

```python
resp = requests.get('https://www.rockauto.com/en/tools/',
                    headers=UA.get_User_Agent_Requests(),
                    proxies=Proxy.get_Proxy_Requests()).text

resp
```

```python
from bs4 import BeautifulSoup
from lxml import etree
```

```python
soup = BeautifulSoup(resp, 'lxml')
html = etree.HTML(str(soup))

with open('./groupname.html', 'w', encoding='UTF-8') as file:
    file.write(str(soup))

print(soup.prettify())
```

```python
list_json = html.xpath('//input[contains(@id, "jsn[")]/@value')

list_json
```

```python
import json
```

```python
list_groupname = [json.loads(json_)['groupname'].strip() for json_ in list_json]

list_groupname
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({'Tab_No.': 3,
                   'Tab': 'Tools & Universal Parts',
                   'GroupName_No.': [i+1 for i in range(len(list_groupname))],
                   'GroupName': list_groupname})

df = df.sort_values(by=['Tab_No.', 'GroupName_No.'], ascending=[True, True]).reset_index(drop=True)
df.to_excel(f'./{param_date}/1.groupname.xlsx', index=False)

df
```
