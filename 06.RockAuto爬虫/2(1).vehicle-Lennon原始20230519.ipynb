```python
site_code= '10401'
```

```python
import pandas as pd

df_menu = pd.read_excel('./part_out.xlsx', header=0)
list_info = df_menu['Info'].to_list()

print(len(list_info))
list_info
```

```python
df = pd.DataFrame(columns=['Info',
                           'Vehicle',
                           'status'])
```

```python
from urllib.parse import quote
```

```python
i = 0

headers = {'Content-Type': 'application/x-www-form-urlencoded; charset=UTF-8'}

func = r'getbuyersguide'
payload = r'{"partData":{"listing_data_essential":{"parttype":"' + site_code + '","partkey":"' + list_info[i].split('?')[1].split('pk=')[1].split('&')[0] + '"}}}'
api_json_request = r'1'

data = 'func=' + func + '&payload=' + quote(payload, safe=' ').replace(' ', '+') + '&api_json_request=' + api_json_request

data
```

```python
import requests
```

```python
resp = requests.post('https://www.rockauto.com/catalog/catalogapi.php', headers=headers, data=data).text

resp
```

```python
import json
```

```python
html = json.loads(resp)['buyersguidepieces']['body']

html
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html, 'lxml')

print(soup.prettify())
```

```python
with open('./vehicle.html', 'w', encoding='utf-8') as file:
    file.write(str(soup))
```

```python
from lxml import etree
```

```python
html = etree.HTML(str(soup))

list_tr = html.xpath('//table[@class="nobmp"]/tr')

list_tr
```

```python
list_vehicle = []
for tr in list_tr:
    list_vehicle.append(' '.join(tr.xpath('./td/text()')))

vehicle = '\n'.join(list_vehicle)

vehicle
```

```python
df_temp = pd.DataFrame({'Info': list_info[i],
                        'Vehicle': vehicle,
                        'status': ['ok']})

df = pd.concat([df, df_temp], ignore_index=True).fillna('')

df
```
