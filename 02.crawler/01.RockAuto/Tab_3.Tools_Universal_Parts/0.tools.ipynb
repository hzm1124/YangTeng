```python
param_date = '20240623'
```

```python
import pandas as pd
```

```python
df_part = pd.read_excel(f'./{param_date}/part.xlsx',
                        header=0,
                        dtype=str).fillna('')

df_part
```

```python
df_vehicle = pd.read_excel(f'./{param_date}/vehicle.xlsx',
                           header=0,
                           dtype=str).fillna('')

df_vehicle
```

```python
df_info = pd.read_excel(f'./{param_date}/info.xlsx',
                        header=0,
                        dtype=str).fillna('')

df_info
```

```python
df = pd.merge(left=df_part,
              right=df_vehicle,
              how='left',
              left_on='JOIN_PTPKPNCN',
              right_on='JOIN_PTPKPNCN',
              suffixes=('_part', '_vehicle')).fillna('N/A')

df = pd.merge(left=df,
              right=df_info,
              how='left',
              left_on='JOIN_PTPKPNCN',
              right_on='JOIN_PTPKPNCN',
              suffixes=('_part', '_info')).fillna('N/A')

df
```

```python
list_column = df_part.columns.tolist()

list_column[list_column.index('Json_Vehicle')] = 'Json_Vehicle_vehicle'
list_column[list_column.index('Vehicle')] = 'Vehicle_vehicle'
list_column[list_column.index('Info')] = 'Info_part'
list_column[list_column.index('Json_Specification')] = 'Json_Specification_info'
list_column[list_column.index('Json_Kit')] = 'Json_Kit_info'

list_column
```

```python
df = df[list_column]

list_column[list_column.index('Json_Vehicle_vehicle')] = 'Json_Vehicle'
list_column[list_column.index('Vehicle_vehicle')] = 'Vehicle'
list_column[list_column.index('Info_part')] = 'Info'
list_column[list_column.index('Json_Specification_info')] = 'Json_Specification'
list_column[list_column.index('Json_Kit_info')] = 'Json_Kit'

df.columns = list_column

df['Tab No'] = df['Tab No'].astype(int)
df['GroupName No'] = df['GroupName No'].astype(int)
df['SubGroupName No'] = df['SubGroupName No'].astype(int)
df['PartType No'] = df['PartType No'].astype(int)
df['Part No'] = df['Part No'].astype(int)

df.to_excel(f'''./{param_date}/4.part.xlsx''', index=False)

df
```
