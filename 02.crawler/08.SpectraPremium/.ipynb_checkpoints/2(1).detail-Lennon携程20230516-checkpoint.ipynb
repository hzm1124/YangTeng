```python
from gevent import monkey
monkey.patch_all()
from gevent.queue import Queue
import gevent

# = = = = = = = = = = = = = = = = = =

import pandas as pd
import requests
from bs4 import BeautifulSoup
from lxml import etree
import re
import time

# = = = = = = = = = = = = = = = = = =

proxies = {'http': 'http://t18126899091866:kh8hjqtl@j933.kdltps.com:15818',
           'https': 'http://t18126899091866:kh8hjqtl@j933.kdltps.com:15818'}

# = = = = = = = = = = = = = = = = = =

df = pd.DataFrame(columns=['Item',
                           '链接',
                           '图片链接',
                           'status'])

# = = = = = = = = = = = = = = = = = =

df_menu = pd.read_excel('./item.xlsx', header=0)
list_product = df_menu['item'].to_list()[:10]
print('总数量：' + str(len(list_product)))
print()

# = = = = = = = = = = = = = = = = = =

work = Queue()
for product in list_product:
    work.put_nowait(product)

# = = = = = = = = = = = = = = = = = =

def crawler():
    global df
    
    while not work.empty():
        product = work.get_nowait()
        url = 'https://ecat.spectrapremium.com/en/parts/' + str(product)
        
        for i in range(31):
            try:
                resp = requests.get(url).text

                # = = = = = = = = = = = = = = = = = =
                
                soup = BeautifulSoup(resp, 'lxml')
                
                # = = = = = = = = = = = = = = = = = =
                
                html = etree.HTML(str(soup))
                
                pic = 'https://ecat.spectrapremium.com' + html.xpath('//a[@id="magic-zoom-plus"]')[0].get('href')

                # = = = = = = = = = = = = = = = = = =
                
                df_temp = pd.DataFrame({'Item': str(product),
                                        '链接': url,
                                        '图片链接': pic,
                                        'status': ['ok']})
                
                # = = = = = = = = = = = = = = = = = =
                
                list_specification = html.xpath('//div[@class="part__nom-spec-valeur-spec"]')
                for specification in list_specification:
                    if specification.xpath('./dt[@class="part__nom-spec"]')[0].text == 'Thickness':
                        df_temp['Thickness'] = specification.xpath('./dd[@class="part__valeur-spec"]')[0].text
                
                break
            
            except:
                if i == 30:
                    df_temp = pd.DataFrame({'Item': str(product),
                                            '链接': '',
                                            '图片链接': '',
                                            'status': ['ok']})
                    time.sleep(3)
                    continue
                    
            # = = = = = = = = = = = = = = = = = =
                
            df = pd.concat([df, df_temp], ignore_index=True).fillna('')
            
            print(str(product) +  '  <->  [' + str(i) + '] - 剩余：' + str(work.qsize()))

# = = = = = = = = = = = = = = = = = =

list_task = []
for _ in range(5):
    task = gevent.spawn(crawler)
    list_task.append(task)
gevent.joinall(list_task)

df.to_excel('./detail.xlsx', index=False)
print()
print('搞定')
```
