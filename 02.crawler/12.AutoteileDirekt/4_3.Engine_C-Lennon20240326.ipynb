```python
import pandas as pd

df = pd.DataFrame()
```

```python
input_ = pd.read_excel('./model_1.xlsx',
                       header=0,
                       dtype=str).fillna('')

print('总数量：' + str(len(input_)))
input_['Model_Id'].tolist()
```

```python
import os
```

```python
list_file = list(os.walk('./engine'))[0][2]
if '.DS_Store' in list_file:
    list_file.remove('.DS_Store')
if '0.null.txt' in list_file:
    list_file.remove('0.null.txt')
    
list_file.sort()
    
len(list_file)
```

```python
import json
```

```python
a = 0

with open('./engine/' + list_file[a], 'r', encoding='UTF-8') as file:
    resp = file.read()
    
json_data = json.loads(resp)

json_data
```

```python
vehicle_2 = '\n'.join([input_.loc[int(list_file[a].replace('.txt', ''))-1, 'Vehicle_1'] + ' ' + engine['title'].strip() for engine in json_data['vehicles']])

print(vehicle_2)
```

```python
df_temp = pd.DataFrame([{'No.': int(list_file[a].replace('.txt', '')),
                         'Product_Id': input_.loc[int(list_file[a].replace('.txt', ''))-1, 'Product_Id'],
                         'Vehicle_2': vehicle_2}])

df = pd.concat([df, df_temp], ignore_index=True).fillna('')
    
len(df)
```

```python
df = df.drop_duplicates(ignore_index=True)
df = df.sort_values(by=['No.'], ascending=[True]).reset_index(drop=True)

df
```
