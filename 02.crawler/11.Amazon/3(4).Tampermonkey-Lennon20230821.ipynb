```python
import os
```

```python
list_file = list(os.walk('./crawler'))[0][2]
if '.DS_Store' in list_file:
   list_file.remove('.DS_Store')

len(list_file)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame()

df
```

```python
import json
```

```python
for file in list_file:
    print(file)
    with open('./crawler/' + file, 'r') as f:
        df = pd.concat([df, pd.DataFrame(json.loads(f.read()))], ignore_index=True)
```

```python
df.to_excel('./crawler.xlsx', index=False)

df
```
