```python
from gevent import monkey
monkey.patch_all(thread=False)
from gevent.queue import Queue
import gevent

import sys
sys.path.append('../../../00.module')
import Proxy

# = = = = = = = = = = = = = = =

import pandas as pd

df = pd.read_excel('./menu.xlsx',
                   header=0,
                   dtype=str).fillna('')

list_manufacturer = df['Manufacturer'].tolist()
list_part_number = df['Part_Number'].tolist()
list_type_code = df['Type_Code'].tolist()

print('总数量：' + str(len(list_part_number)))
print()

# = = = = = = = = = = = = = = =

df = pd.DataFrame()

# = = = = = = = = = = = = = = =

from urllib.parse import quote
import requests
import json
from bs4 import BeautifulSoup
from lxml import etree
import numpy as np

# = = = = = = = = = = = = = = =

import time
from datetime import datetime

# = = = = = = = = = = = = = = =

work = Queue()
for manufacturer, part_number, type_code in zip(list_manufacturer, list_part_number, list_type_code):
    work.put_nowait((manufacturer, part_number, type_code))

def crawler():
    global df
    
    while not work.empty():
        manufacturer, part_number, type_code = work.get_nowait()
        
        # = = = = = = = = = = = = = = =
        
        status = 'error'
        for _ in range(11):
            try:
                data = 'dopartsearch=1&partsearch%5Bpartnum%5D%5Bpartsearch_007%5D=' + part_number + '&partsearch%5Bmanufacturer%5D%5Bpartsearch_007%5D=' + manufacturer.replace('+', '%2B').replace(' ', '+').replace('&', '%26').replace('/', '%2F').replace(',', '%2C').replace('Ö', '%C3%96').replace('Ü', '%C3%9C') + '&partsearch%5Bparttype%5D%5Bpartsearch_007%5D=' + type_code + '&partsearch[do][partsearch_007]=1&func=sendparttabsearch&payload=%7B%7D&api_json_request=1&sctchecked=1&scbeenloaded=true&curCartGroupID='

                resp = requests.post('https://www.rockauto.com/catalog/catalogapi.php',
                                     data=data,
                                     headers={'Content-Type': 'application/x-www-form-urlencoded; charset=UTF-8'},
                                     proxies=Proxy.get_Proxy_Requests()).text

                resp = json.loads(resp)

                
                url = 'https://www.rockauto.com/en/partsearch/?mfr=' + manufacturer.replace('+', '%2B').replace(' ', '+').replace('&', '%26').replace('/', '%2F').replace('\'', '%27').replace(',', '%2C').replace('(', '%28').replace(')', '%29').replace('Ö', '%D6').replace('Ü', '%DC') + '&partnum=' + part_number

                # = = = = = = = = = = = = = = =

                
                html = resp['searchnoderesults']
                
                # = = = = = = = = = = = = = = =
                
                soup = BeautifulSoup(html, 'lxml')
                
                # = = = = = = = = = = = = = = =
                
                html = etree.HTML(str(soup))
                
                if len(html.xpath('//span[contains(@class, "navlabellink") and contains(@class, "nvoffset") and contains(@class, "nnormal")]')) != 0:
                    df_temp = pd.DataFrame([{'status': 'ok',
                                             'Part_Number': part_number,
                                             'Manufacturer': manufacturer,
                                             'Type_Code': type_code,
                                             'Part_Key': 'Nan',
                                             'Vehicle': '',
                                             'Alternate_OE_Part_Numbers': 'Nan',
                                             'Note_1': 'Nan',
                                             'Note_2': 'Nan',
                                             'Pic': '',
                                             'Url': url,
                                             'Info': 'Nan',
                                             'Src': 'Nan'}])
                    
                    status = 'ok'
                    
                    break
                
                # = = = = = = = = = = = = = = =
                
                list_part = html.xpath('//tbody[contains(@class, "listing-inner") and (contains(@class, "altrow-a-1") or contains(@class, "altrow-a-0"))]')
                
                # = = = = = = = = = = = = = = =
                
                list_part_number = [_.xpath('./descendant::div[contains(@class, "listing-text-row-moreinfo-truck")]/span[contains(@title, "Buyer\'s Guide")]')[0] for _ in list_part]

                # = = = = = = = = = = = = = = =
                
                array_part_number = np.array([_.text for _ in list_part_number])

                # = = = = = = = = = = = = = = =
                
                number_index = np.where(array_part_number == part_number)[0]

                # = = = = = = = = = = = = = = =
                
                list_manufacturer = [_.xpath('./descendant::div[contains(@class, "listing-text-row-moreinfo-truck")]/span[contains(@class, "listing-final-manufacturer")]')[0] for _ in list_part]

                # = = = = = = = = = = = = = = =
                
                array_manufacturer = np.array([_.text for _ in list_manufacturer])

                # = = = = = = = = = = = = = = =
                
                manufacturer_index = np.where(array_manufacturer == manufacturer)[0]
                
                # = = = = = = = = = = = = = = =
                
                index = list(np.intersect1d(number_index, manufacturer_index))
                
                # = = = = = = = = = = = = = = =
                
                if len(index) == 1:
                    index = index[0]
                    # = = = = = = = = = = = = = = =
                    note_1 = ';'.join(_.xpath('./text()')[0] for _ in list_part[index].xpath('./descendant::div[contains(@class, "listing-text-row-moreinfo-truck")]/span[contains(@class, "span-link-underline-remover")]'))
                    # = = = = = = = = = = = = = = =
                    list_note_2 = []
                    note_2 = [_ for _ in list_part[index].xpath('./descendant::div[contains(@class, "listing-text-row") and not(contains(@class, "moreinfo-truck"))]/descendant-or-self::span')]
                    for _ in note_2:
                        if _.xpath('./text()') != []:
                            list_note_2.append(_.xpath('./text()')[0])
                    note_2 = ';'.join(list_note_2)
                    # = = = = = = = = = = = = = = =
                    oe = ';'.join(_.xpath('./text()')[0] for _ in list_part[index].xpath('./descendant::span[contains(@title, "Replaces these Alternate/ OE Part Numbers")]'))
                    oe = ';'.join(oe.split(', '))
                    # = = = = = = = = = = = = = = =
                    src = ';'.join(_.xpath('./@src')[0] for _ in list_part[index].xpath('./descendant::img[contains(@id, "inlineimg_thumb")]'))
                    if src == '':
                        src = ''
                    else:
                        src = 'https://www.rockauto.com' + src.replace('__ra_m.', '.')
                    # = = = = = = = = = = = = = = =
                    part_key = json.loads(list_part[index].xpath('./descendant::input[contains(@name, "listing_data_essential")]/@value')[0])['partkey']
                    # = = = = = = = = = = = = = = =
                    df_temp = pd.DataFrame([{'status': 'ok',
                                             'Part_Number': part_number,
                                             'Manufacturer': manufacturer,
                                             'Type_Code': type_code,
                                             'Part_Key': part_key,
                                             'Vehicle': '',
                                             'Alternate_OE_Part_Numbers': oe,
                                             'Note_1': note_1,
                                             'Note_2': note_2,
                                             'Pic': '',
                                             'Url': url,
                                             'Info': 'https://www.rockauto.com/en/moreinfo.php?pt=' + type_code + '&pk=' + part_key,
                                             'Src': src}])
                    
                    status = 'ok'
                    
                break
                
            except:
                time.sleep(1.5)
                continue
                    
        # = = = = = = = = = = = = = = =

        if status == 'error':
            df_temp = pd.DataFrame([{'status': 'error',
                                     'Part_Number': part_number,
                                     'Manufacturer': manufacturer,
                                     'Type_Code': type_code,
                                     'Url': url}])

        # = = = = = = = = = = = = = = =

        df = pd.concat([df, df_temp], ignore_index=True).fillna('')

        print(manufacturer + ', ' + part_number + ', ' + type_code +  '  <->  [' + status + '] - 剩余数量：' + str(work.qsize()))

# = = = = = = = = = = = = = = =
                
list_task = []
for _ in range(10):
    task = gevent.spawn(crawler)
    list_task.append(task)
gevent.joinall(list_task)

df = df.drop_duplicates(ignore_index=True)
                
print()
print('爬虫输出中')
df.to_excel('./part_' + datetime.now().strftime('%Y%m%d_%H%M%S') + '.xlsx', index=False)
print()
print('爬虫结束')
```
