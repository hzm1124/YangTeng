```python
import os

list_file = sorted(list(os.walk('./file/2.url'))[0][2])
while '.DS_Store' in list_file:
    list_file.remove('.DS_Store')
while '0.null.txt' in list_file:
    list_file.remove('0.null.txt')

list_file
```

```python
file = list_file[0]

file
```

```python
import pandas as pd

output_correct, output_error = pd.DataFrame(), pd.DataFrame()
```

```python
input_ = pd.read_excel(f'./file/2.url/{file}',
                       header=0,
                       dtype=str).fillna('')

input_['No'] = input_['No'].astype(int)

print(f'总数量：{len(input_)}')
input_['Url'].tolist()
```

```python
a = 0

a
```

```python
crawler_status = 'error'

crawler_status
```

```python
import requests

import sys
sys.path.append('../00.Tools')
from crawler_configuration import get_header, get_proxy

b = 0
while True:
    b += 1
    print(b)
    try:
        resp = requests.get(input_.loc[a, 'Url'],
                            headers=get_header(),
                            proxies=get_proxy(),
                            timeout=(10, 10))

        if resp.status_code == 200:
            break
    except KeyboardInterrupt:
        break
    except:
        continue

resp
```

```python
from bs4 import BeautifulSoup
from lxml import etree

soup = BeautifulSoup(resp.text, 'lxml')
html = etree.HTML(str(soup))

with open('./part.html', 'w', encoding='UTF-8') as file:
    file.write(str(soup))

print(soup.prettify())
```

```python
type_ = html.xpath('//h1/span[@id="lblPartTypeName"]/text()')[0].strip()

type_
```

```python
i, dict_src = 0, {}
list_src = list(reversed([href.strip() for href in html.xpath('//a[@data-zoom-id="ProductPic"]/@href')]))
for src in list_src:
    if src.startswith('http'):
        dict_src[str(i)] = src
        i += 1

dict_src
```

```python
part_number = html.xpath('//span[@id="lblProductName"]/text()')[0].strip()

part_number
```

```python
description = html.xpath('//span[@id="lblProductDesc"]/text()')[0].strip()

description
```

```python
application_summary = '\n'.join(html.xpath('//div[@id="divAppSummary"]/p/text()')).strip()

application_summary
```

```python
brand = html.xpath('//span[@id="lblBrandName"]/text()')[0].strip()

brand
```

```python
vehicle_url = html.xpath('//input[@id="hidfDetailApp"]/@value')
vehicle_url = f'''https://www.dormanproducts.com/{vehicle_url[0].strip()}''' if vehicle_url else ''

vehicle_url
```

```python
dict_specification = {}
list_tr = html.xpath('//section[@id="productSpec"]/div/table/tr')
for i, tr in zip(range(len(list_tr)), list_tr):
    dict_specification[str(i)] = {tr.xpath('./th/text()')[0].strip().removesuffix(':').strip(): tr.xpath('./td/text()')[0].strip()}

dict_specification
```

```python
list_oem = [text.strip() for text in html.xpath('//section[@id="productOE"]/div/table/tbody/tr/th/text()')]
oem = ';'.join(sorted(list(set(list_oem)), key=list_oem.index))

oem
```

```python
import json

df_temp = pd.DataFrame([{'No': input_.loc[a, 'No'],
                         'Type': type_,
                         'Part Number': part_number,
                         'Brand': brand,
                         'Description': description,
                         'Application Summary': application_summary,
                         'OEM': oem,
                         'Picture': '',
                         'Url': input_.loc[a, 'Url'],
                         'Vehicle Url': vehicle_url,
                         'Json_Src': json.dumps(dict_src),
                         'Json_Specification': json.dumps(dict_specification)}])

output_correct = pd.concat([output_correct, df_temp], ignore_index=True).fillna('')

output_correct
```

```python
crawler_status = 'ok'

crawler_status
```

```python
if crawler_status == 'error':
    output_error = pd.concat([output_error, pd.DataFrame([input_.iloc[a]]).reset_index(drop=True)], ignore_index=True).fillna('')

output_error
```

```python
if not output_correct.empty:
    output_correct = output_correct.sort_values(by=['No'],
                                                ascending=[True],
                                                ignore_index=True)
    output_correct.to_excel('./test_part.xlsx', index=False)

output_correct
```

```python
if not output_error.empty:
    output_error = output_error.sort_values(by=['No'],
                                            ascending=[True],
                                            ignore_index=True)
    output_error.to_excel('./test_part_error.xlsx', index=False)

output_error
```
