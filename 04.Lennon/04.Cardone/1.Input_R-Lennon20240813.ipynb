```python
import requests

import sys
sys.path.append('../00.Tools')
from crawler_configuration import get_header, get_proxy

b = 0
while True:
    b += 1
    print(b)
    try:
        resp = requests.get('https://www.cardone.com/',
                             headers=get_header(),
                             proxies=get_proxy(),
                             timeout=(10, 10))
        
        if resp.status_code == 200:
            break
    except KeyboardInterrupt:
        break
    except:
        continue

resp
```

```python
from bs4 import BeautifulSoup
from lxml import etree

soup = BeautifulSoup(resp.text, 'lxml')
html = etree.HTML(str(soup))

with open('./input.html', 'w', encoding='UTF-8') as file:
    file.write(str(soup))

print(soup.prettify())
```

```python
list_li = html.xpath('//ul[@class="navPages-list navPages-list-depth-max simple"]/li[@class="navPages-item"][1]/div/ul[@class="navPage-subMenu-list"]/li[@class="navPages-item"]')[:-1]

list_li
```

```python
list_list_li_2 = [li.xpath('./div/ul[@class="navPage-subMenu-list"]/li[@class="navPage-subMenu-item-child"]') for li in list_li]

list_list_li_2
```

```python
list_list_list_li_3 = [[li.xpath('./div/ul[@class="navPage-subMenu-list"]/li[@class="navPage-subMenu-item-child"]') for li in list_li_3] for list_li_3 in list_list_li_2]
list_list_list_url_3 = [[[li.xpath('./a/@href')[0].strip() for li in list_li_3] for list_li_3 in list_list_li_3] for list_list_li_3 in list_list_list_li_3]

list_list_list_url_3
```

```python
list_url = [url for list_list_url_2 in list_list_list_url_3 for list_url in list_list_url_2 for url in list_url]

list_url
```

```python
import pandas as pd

list_url = sorted([url for list_list_url_2 in list_list_list_url_3 for list_url in list_list_url_2 for url in list_url])

output = pd.DataFrame({'No': [_+1 for _ in range(len(list_url))],
                       'Url': list_url})

output.to_excel('./file/input.xlsx', index=False)

output
```
