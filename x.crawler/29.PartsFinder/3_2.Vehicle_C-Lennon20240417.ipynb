```python
import pandas as pd
```

```python
df = pd.read_excel('./part.xlsx',
                   header=0,
                   dtype=str).fillna('')

print('总数量：' + str(len(df)))
df['list_make'].tolist()
```

```python
dict_url = {}
for i in range(len(df)):
    dict_url[str(i+1)] = 'https://partsfinder.bilsteingroup.com/rest/v1/article/applications?make=' + df.loc[i, 'list_make']
    
dict_url
```

```python
import os
```

```python
list_file = list(os.walk('./vehicle'))[0][2]
if '.DS_Store' in list_file:
    list_file.remove('.DS_Store')
if '0.null.txt' in list_file:
    list_file.remove('0.null.txt')
    
list_file.sort()
    
len(list_file)
```

```python
dict_url = dict(filter(lambda x: x[0] + '.txt' not in list_file, dict_url.items()))

dict_url
```

```python
import pyperclip
```

```python
for i, key in zip(range(len(dict_url)), dict_url):
    if i == 0:
        data_1 = 'let dict_url = {"' + key + '": "' + dict_url[key] + '", '
        data_2 = 'let list_key = ["' + key + '", '
        data_3 = 'let list_brand_id = ["' + df.loc[int(key)-1, 'Brand_Id'] + '", '
        data_4 = 'let list_master_id = ["' + df.loc[int(key)-1, 'Master_Id'] + '", '
    else:
        data_1 += '"' + key + '": "' + dict_url[key] + '", '
        data_2 += '"' + key + '", '
        data_3 += '"' + df.loc[int(key)-1, 'Brand_Id'] + '", '
        data_4 += '"' + df.loc[int(key)-1, 'Master_Id'] + '", '

data = data_1[:-2] + '};\n' + data_2[:-2] + '];\n' + data_3[:-2] + '];\n' + data_4[:-2] + '];\n'

pyperclip.copy(data)

data
```
