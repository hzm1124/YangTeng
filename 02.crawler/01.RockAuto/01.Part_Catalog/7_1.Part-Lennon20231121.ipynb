```python
import pandas as pd
```

```python
df = pd.read_excel('./' + param_make + '/6.type.xlsx',
                   header=0,
                   dtype=str).fillna('')

list_make = df['Make'].tolist()
list_year = df['Year'].tolist()
list_model = df['Model'].tolist()
list_engine = df['Engine'].tolist()
list_car_code = df['Car_Code'].tolist()
list_category = df['Category'].tolist()
list_category_code = df['Category_Code'].tolist()
list_type = df['Type'].tolist()
list_type_code = df['Type_Code'].tolist()
list_url = df['Url'].tolist()

print('总数量：' + str(len(list_type)))
list_type
```

```python
df = pd.DataFrame()

df
```

```python
from urllib.parse import quote
import requests
import json
```

```python
i = 1

make = list_make[0]
year = list_year[0]
model = list_model[0]
engine = list_engine[0]
car_code = list_car_code[0]
category = list_category[0]
category_code = list_category_code[0]
type_ = list_type[i]
type_code = list_type_code[i]
url = list_url[i]

if type_ == 'Nan' and type_code == 'Nan':
    print('fuck')

data = 'func=navnode_fetch&payload=' + quote('{"jsn":{"tab":"catalog","make":"') + make.replace(' ', '+') + quote('","year":"') + year + quote('","model":"') + model.replace('+', '%2B').replace(' ', '+').replace('&', '%26').replace('/', '%2F') + quote('","engine":"') + engine.replace(' ', '+').replace('/', '%2F') + quote('","carcode":"') + car_code + quote('","groupname":"') + category.replace(' ', '+').replace('&', '%26').replace('/', '%2F') + quote('","parttype":"') + type_code + quote('","nodetype":"parttype"}}') + '&api_json_request=1&sctchecked=1&scbeenloaded=true&curCartGroupID='

resp = requests.post('https://www.rockauto.com/catalog/catalogapi.php',
                     data=data,
                     headers={'Content-Type': 'application/x-www-form-urlencoded; charset=UTF-8'}).text

resp = json.loads(resp)

resp
```

```python
from bs4 import BeautifulSoup
```

```python
html = resp['html_fill_sections']['navchildren[]']

soup = BeautifulSoup(html, 'lxml')

with open('./part.html', 'w', encoding='UTF-8') as file:
    file.write(str(soup))

print(soup.prettify())
```

```python
from lxml import etree
```

```python
html = etree.HTML(str(soup))

list_part = html.xpath('//tbody[contains(@class, "listing-inner") and (contains(@class, "altrow-a-1") or contains(@class, "altrow-a-0"))]')

list_part
```

```python
list_manufacturer = [_.xpath('./descendant::div[contains(@class, "listing-text-row-moreinfo-truck")]/span[contains(@class, "listing-final-manufacturer")]')[0].text for _ in list_part]

list_manufacturer
```

```python
list_part_number = [_.xpath('./descendant::div[contains(@class, "listing-text-row-moreinfo-truck")]/span[contains(@title, "Buyer\'s Guide")]')[0].text for _ in list_part]

list_part_number
```

```python
list_note_1 = []
list_span = [_.xpath('./descendant::div[contains(@class, "listing-text-row-moreinfo-truck")]') for _ in list_part]
for _ in list_span:
    note_1 = _[0].xpath('./span[contains(@class, "span-link-underline-remover")]/text()')
    if len(note_1) == 0:
        list_note_1.append('')
    else:
        list_note_1.append(note_1[0].strip())

list_note_1
```

```python
list_note_2 = []
list_span = [_.xpath('./descendant::div[contains(@class, "listing-text-row") and not(contains(@class, "moreinfo-truck"))]') for _ in list_part]
for _ in list_span:
    note_2 = _[0].xpath('./span/span/text()')
    if len(note_2) == 0:
        list_note_2.append('')
    else:
        note_2 = [_.strip() for _ in note_2]
        while '' in note_2:
            note_2.remove('')
        note_2 = '\n'.join(note_2)
        list_note_2.append(note_2.strip())

list_note_2
```

```python
list_oe = []
list_span = [_.xpath('./descendant::span[contains(@title, "Replaces these Alternate/ OE Part Numbers")]') for _ in list_part]
for _ in list_span:
    if len(_) == 0:
        list_oe.append('')
    else:
        list_oe.append(_[0].xpath('./text()')[0].replace(', ', ';'))

list_oe
```

```python
list_src = []
list_span = [_.xpath('./descendant::img[contains(@id, "inlineimg_thumb")]') for _ in list_part]
for _ in list_span:
    if len(_) == 0:
        list_src.append('')
    else:
        list_src.append('https://www.rockauto.com' + _[0].xpath('./@src')[0].replace('__ra_m.', '.'))

list_src
```

```python
list_part_key = []
list_span = [_.xpath('./descendant::input[contains(@id, "listing_data_essential")]') for _ in list_part]
for _ in list_span:
    if len(_) == 0:
        list_part_key.append('')
    else:
        list_part_key.append(json.loads(_[0].xpath('./@value')[0])['partkey'])

list_part_key
```

```python
list_info = ['https://www.rockauto.com/en/moreinfo.php?pt=' + type_code + '&pk=' + _ for _ in list_part_key]

list_info
```

```python
list_tag = []
list_span = [_.xpath('./descendant::input[contains(@id, "listing_data_supplemental")]') for _ in list_part]
for _ in list_span:
    if len(_) == 0:
        list_tag.append('')
    else:
        list_tag.append(json.loads(_[0].xpath('./@value')[0])['sortgrouptext'])

list_tag
```

```python
df_temp = pd.DataFrame({'status': 'ok',
                        'Make': make,
                        'Year': year,
                        'Model': model,
                        'Engine': engine,
                        'Car_Code': car_code,
                        'Category': category,
                        'Category_Code': category_code,
                        'Type': type_,
                        'Type_Code': type_code,
                        'Tag': list_tag,
                        'Manufacturer': list_manufacturer,
                        'Part_Number': list_part_number,
                        'Part_Key': list_part_key,
                        'Vehicle': '',
                        'Alternate_OE_Part_Numbers': list_oe,
                        'Note_1': list_note_1,
                        'Note_2': list_note_2,
                        'Pic': '',
                        'Url': url,
                        'Info': list_info,
                        'Src': list_src})

df = pd.concat([df, df_temp], ignore_index=True).fillna('')

df
```
