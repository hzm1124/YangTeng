```python
import os

os.makedirs('./11.part_2')
os.makedirs('./12.vehicle')
os.makedirs('./12.vehicle/0.error')
```

```python
import pandas as pd

input_ = pd.read_excel('./10.part_1.xlsx',
                       header=0,
                       dtype=str).fillna('')

input_['No'] = input_['No'].astype(int)

input_
```

```python
input_ = input_.sort_values(by=['Type', 'Part Number'],
                            ascending=[True, True],
                            ignore_index=True)

input_['No'] = [_+1 for _ in range(len(input_))]

input_.to_excel('./11.part.xlsx', index=False)

input_
```

```python
input_2 = input_.drop_duplicates(['Type'],
                                 keep='first',
                                 ignore_index=True)
input_2 = input_2.sort_values(by=['Type'],
                              ascending=[True],
                              ignore_index=True)

input_2['Type No'] = [i+1 for i in range(len(input_2))]
input_2['File Type'] = [_.strip().replace('/', '%2F').replace('\\', '%5C').strip() for _ in input_2['Type'].tolist()]

input_2
```

```python
from tqdm import tqdm

for i in tqdm(range(len(input_2)), desc='Progress', ncols=77):
    output = input_[input_['Type'] == input_2.loc[i, 'Type']].reset_index(drop=True)
    output = output.sort_values(by=['No'],
                                ascending=[True],
                                ignore_index=True)

    # = = = = = = = = = = = = = = =

    input_2.loc[i, 'SKU Count'] = len(output)

    # = = = = = = = = = = = = = = =

    list_output, index_start, index_end = [], 0, 3_000
    while True:
        if index_end < len(output):
            list_output.append(output[index_start:index_end].reset_index(drop=True))
            index_start, index_end = index_end, index_end + 3_000
        else:
            list_output.append(output[index_start:].reset_index(drop=True))
            break
    
    if len(list_output) == 1:
        output.to_excel(f'''./11.part_2/type_{input_2.loc[i, 'Type No']}.{input_2.loc[i, 'File Type']}.xlsx''', index=False)
    else:
        for j, output in zip(range(len(list_output)), list_output):
            output.to_excel(f'''./11.part_2/type_{input_2.loc[i, 'Type No']}.{input_2.loc[i, 'File Type']}-serial_{j+1}.xlsx''', index=False)

input_2[['Type No',
         'Type',
         'SKU Count']].to_excel('./11.catalog.xlsx', index=False)

input_2
```
