```python
df = pd.DataFrame(columns=['Info',
                           'Alternate Inventory Number',
                           'OEM / Interchange Numbers',
                           'Src',
                           'status'])

df
```

```python
from urllib.parse import quote
```

```python
i = 0

url = 'https://www.rockauto.com/en/moreinfo.php?pk=' + list_info[i].split('?')[1].split('pk=')[1].split('&')[0] + '&pt=' + list_url[i].split(',')[-1]

url
```

```python
import requests
```

```python
resp = requests.get(url).text

resp
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(resp, 'lxml')

print(soup.prettify())
```

```python
with open('./detail.html', 'w', encoding='utf-8') as file:
    file.write(str(soup))
```

```python
from lxml import etree
```

```python
html = etree.HTML(str(soup))

dict_oe = {'alternate_inventory_number': '', 'oem_interchange_number': ''}
list_oe = []
list_section = html.xpath('//section[@aria-label="Alternate/OEM Part Number(s)"]')
for section in list_section:
    list_oe.append(''.join(section.xpath('./descendant-or-self::*/text()')))

list_oe
```

```python
for oe in list_oe:
    if '{Alternate Inventory Numbers: ' in oe:
        dict_oe['alternate_inventory_number'] = oe.replace('{Alternate Inventory Numbers: ', '').replace('}', '').replace(', ', ';')
    elif 'OEM / Interchange Numbers: ' in oe:
        dict_oe['oem_interchange_number'] = oe.replace('OEM / Interchange Numbers: ', '').replace(', ', ';')

dict_oe
```

```python
list_src = html.xpath('//img[@alt="Part image"]/@src')
if len(list_src) == 0:
    src = ''
else:
    src = 'https://www.rockauto.com' + list_src[0]

src
```

```python
df_temp = pd.DataFrame([{'Info': list_info[i],
                         'Alternate Inventory Number': dict_oe['alternate_inventory_number'],
                         'OEM / Interchange Numbers': dict_oe['oem_interchange_number'],
                         'Src': src,
                         'status': 'ok'}])

df = pd.concat([df, df_temp], ignore_index=True).fillna('')

df
```
