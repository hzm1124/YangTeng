```python
import json
import requests

import sys
sys.path.append('../00.Tools')
from configuration import generate_header, generate_proxy
```

```python
data = {'func': 'tab_fetch',
        'payload': json.dumps({'jstab': {'key': 'partsearch'}}),
        'api_json_request': '1'}
response = requests.post('https://www.rockauto.com/catalog/catalogapi.php',
                         data=data,
                         headers=generate_header(),
                         proxies=generate_proxy(),
                         timeout=(10, 15),
                         allow_redirects=False)

response.status_code
```

```python
dict_ = response.json()

dict_
```

```python
'redirect_to_url' in dict_ or dict_['jstab']['key'].strip() != 'partsearch'
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(dict_['html_fill_sections']['section-tab-panel-inner-dynamic[partsearch]'], 'html5lib')

with open('./1.manufacturer.html', 'w', encoding='utf-8') as file:
    file.write(str(soup))

print(soup.prettify())
```
