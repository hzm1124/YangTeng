```python
import sys
sys.path.append('../../00.module')
import UA
import Proxy

i = 0

resp = requests.get('https://ecatalog.smpcorp.com/v2/fs/api/part/partselect?part=' + input_.loc[i, 'Part_Number'] + '&func=PART&vid=').text

resp = json.loads(resp)

resp
```

```python
part_type = resp['pp']['categories'][0]['categoryName_en'].strip()

part_type
```

```python
pop = resp['pd']['pop'].strip()

pop
```

```python
per_car = resp['pd']['qtyEach']

per_car
```

```python
vehicle = '\n'.join([vehicle['buyersGuideDesc'].strip() for vehicle in resp['pp']['buyersGuides']])

print(vehicle)
```

```python
specifications = resp['pp']['partSpecs']

specifications
```

```python
brand = resp['pd']['brand'].strip()

brand
```

```python
df_temp = pd.DataFrame([{'序号': int(input_.loc[i, 'No.']),
                         'Part_Number': input_.loc[i, 'Part_Number'],
                         'Part_Type': part_type,
                         'Per_Car': per_car,
                         'Brand': brand,
                         'POP': pop,
                         'Vehicle': vehicle,
                         'Pic': '',
                         'Url': 'https://ecatalog.smpcorp.com/v2/fs/#/partdetail/part/'+input_.loc[i, 'Part_Number'],
                         'Src': ''}])

for key_value in specifications:
    df_temp[key_value['attributeName_en'].lower()] = key_value['siteValue_en']

output_correct = pd.concat([output_correct, df_temp], ignore_index=True).fillna('')

output_correct
```
