```python
import pandas as pd

input_ = pd.read_excel('./3.item.xlsx',
                       header=0,
                       dtype=str).fillna('')

input_['No'] = input_['No'].astype(int)

input_
```

```python
from tqdm import tqdm
import json

for i in tqdm(range(len(input_)), desc='Progress', ncols=77):
    dict_specific = json.loads(input_.loc[i, 'Json_Specific'])
    for value in list(dict_specific.values()):
        input_.loc[i, list(value.keys())[0]] = list(value.values())[0]

input_.to_excel('./7.specific.xlsx', index=False)

input_
```
