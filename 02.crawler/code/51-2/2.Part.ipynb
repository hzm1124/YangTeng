```python
item = {'no': 1,
 'brand': 'KYB',
 'part_number': '344115',
 'part_no': 1,
 'tag': 'KYB Shock absorber (344115)',
 'url': 'https://www.onlinecarparts.co.uk/kyb-828666.html',
 'article_id': 828666,
 'article_no': '344115'}

item
```

```python
import requests

import sys
sys.path.append('../00.Tools')
from configuration import generate_header, generate_proxy
```

```python
response = requests.get(item['url'],
                        headers=generate_header(**{'Referer': 'https://www.onlinecarparts.co.uk/'}),
                        # proxies=generate_proxy(),
                        timeout=(10, 15),
                        allow_redirects=False)

response.status_code
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(response.text, 'html5lib')

with open('./2.part.html', 'w', encoding='utf-8') as file:
    file.write(str(soup))

print(soup.prettify())
```

```python
from parsel import Selector
```

```python
selector = Selector(text=str(soup))

selector
```

```python
import json
```

```python
dict_ = json.loads(selector.xpath('//main[@id="main"]/script[@type="application/ld+json"]/text()').get())

dict_
```

```python
dict_src, list_src = {}, dict_['image'] if type(dict_['image']) == list else [dict_['image']]
for i, src in enumerate(list_src):
    dict_src[str(i)] = src.strip()

dict_src
```

```python
type_ = dict_['category'].strip()

type_
```

```python
price = f'''{dict_['offers']['priceCurrency'].strip()} {dict_['offers']['price']}'''

price
```

```python
list_li = selector.xpath('//div[@class="breadcrumbs"]/ul/li')
breadcrumbs, length = '', len(list_li)
for i in range(length - 1, 0, -1):
    breadcrumbs += f'''{i}. {list_li[i - 1].xpath('./a/span/text()').get().strip()} ({list_li[i - 1].xpath('./a/@href').get().strip()})\n'''

print(breadcrumbs.strip())
```

```python
title = selector.xpath('//div[@class="product__head"]/div[@class="product__title"]/h1[@class="product__title-link"]/text()').get().strip()

title
```

```python
subtitle = selector.xpath('//div[@class="product__head"]/div[@class="product__title"]/h1[@class="product__title-link"]/span[@class="product__subtitle"]/text()').get().strip()

subtitle
```

```python
dict_info, list_info =  {}, [info.strip() for info in selector.xpath('////div[@class="product__head"]/div[@class="product__artkl"]/text()').get().strip().split('\n')]
for info in list_info:
    dict_info[info.split(':')[0].strip()] = ':'.join(info.split(':')[1:]).strip()

dict_info
```

```python
dict_key, list_tr = {}, selector.xpath('//div[@class="product__middle d-flex flex-column"]/div[@class="product__description"]/div[@class="product__desc active"]/div[@class="product__desc-content"]/div[@class="product__desc-table"]/table[@class="product__table"]/tbody/tr')
for tr in list_tr:
    key = tr.xpath('./td[@class="product__table-col left"]/text()').get().strip().removesuffix(':').strip()

    if key in dict_key:
        dict_key[key] += 1
    else:
        dict_key[key] = 1

dict_detail, dict_count = {}, {}
for tr in list_tr:
    key, value = tr.xpath('./td[@class="product__table-col left"]/text()').get().strip().removesuffix(':').strip(), tr.xpath('./td[@class="product__table-col right"]/text()').get().strip()

    if dict_key[key] == 1:
        dict_detail[key] = 'nan' if value == '' else value
    else:
        if key in dict_count:
            dict_count[key] += 1
        else:
            dict_count[key] = 1
        dict_detail[f'{key}_{dict_count[key]}'] = 'nan' if value == '' else value

dict_detail
```

```python
import re
```

```python
list_li = selector.xpath('//div[@class="product-info-blocks"]/div[@class="row align-items-end"]/div[@data-show-list="data-product-accordion"]/div[@class="product-oem__content"]/div[@data-oem-list]/ul/li')
list_oem = [li.xpath('./a[@class="product-oem__link"]/text()').get().strip() if li.xpath('./a[@class="product-oem__link"]') else li.xpath('./text()').get().strip() for li in list_li]
oem = ';'.join(sorted(dict.fromkeys([re.sub(r'^.*?OE-', '', oem).strip() for oem in list_oem])))

oem
```

```python
list_div = selector.xpath('//div[@class="product-info-blocks"]/div[@class="row align-items-end"]/div[@data-show-list="data-product-accordion"]/div[@class="compatibility__content"]/div[@class="compatibility__maker-list"]/div[@class="compatibility__maker"]')

list_div
```

```python
dict_key, list_li = {}, selector.xpath('//div[@class="product-info-blocks"]/div[@class="row align-items-end"]/div[@data-show-list="data-product-accordion"]/div[@class="product-analogs__content"]/div[@class="product-analogs__wrapper"]/ul/li')
for li in list_li:
    key = li.xpath('./span[@class="product-analogs__name"]/text()').get().strip().removesuffix(':').strip()

    if key in dict_key:
        dict_key[key] += 1
    else:
        dict_key[key] = 1

dict_equivalent, dict_count = {}, {}
for li in list_li:
    key, value = li.xpath('./span[@class="product-analogs__name"]/text()').get().strip().removesuffix(':').strip(), li.xpath('./text()').getall()[1].strip()

    if dict_key[key] == 1:
        dict_equivalent[key] = 'nan' if value == '' else value
    else:
        if key in dict_count:
            dict_count[key] += 1
        else:
            dict_count[key] = 1
        dict_equivalent[f'{key}_{dict_count[key]}'] = 'nan' if value == '' else value

dict_equivalent
```
