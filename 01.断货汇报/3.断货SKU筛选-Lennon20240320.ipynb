```python
import os
```

```python
list_file = list(os.walk('./'))[0][2]
if '.DS_Store' in list_file:
    list_file.remove('.DS_Store')
if '0.null.txt' in list_file:
    list_file.remove('0.null.txt')
    
for file in list_file:
    if '计算表' in file:
        calculate = './' + file

print('calculate:     ' + calculate)
```

```python
import pandas as pd
```

```python
df = pd.read_excel(calculate,
                   sheet_name='附_1统计源（含新品、清仓、新品成长）',
                   header=3,
                   usecols='C, D, F, L, M')

df[df.columns[0]] = df[df.columns[0]].astype(str)
df[df.columns[1]] = df[df.columns[1]].astype(float)
df[df.columns[2]] = df[df.columns[2]].astype(str)
df[df.columns[3]] = df[df.columns[3]].astype(int)
df[df.columns[4]] = df[df.columns[4]].astype(int)

df
```

```python
import datetime
```

```python
df = df[(df['在库'] == 0) & (df['在途'] == 0) & (df['Level'] != '普通')].reset_index(drop=True)
df = df.sort_values(by=['Level', '目标\n周销', 'MRP SKU'], ascending=[True, False, True]).reset_index(drop=True)
df['序号'] = [i+1 for i in range(len(df))]
df = df[['序号', 'MRP SKU', 'Level', '目标\n周销', '在途']]
df.columns = ['序号', 'SKU', '层级', '周销', '在途']

df.to_excel('./筛选好的断货SKU_' + datetime.datetime.now().strftime("%Y%m%d_%H%M%S") + '.xlsx', index=False)

df
```
