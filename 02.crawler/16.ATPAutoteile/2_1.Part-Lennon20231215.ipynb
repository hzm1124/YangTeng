```python
import pandas as pd
```

```python
df = pd.read_excel('./menu.xlsx',
                   header=0,
                   dtype=str).fillna('')

list_url = df['Url'].tolist()

print('总数量：' + str(len(list_url)))
list_url
```

```python
df = pd.DataFrame()

df
```

```python
import requests
```

```python
import sys
sys.path.append('../../00.module')
import UA
import Proxy

i = 0

url = list_url[i]

resp = requests.get(url,
                    headers=UA.get_User_Agent_Requests(),
                    proxies=Proxy.get_Proxy_Requests()).text

resp
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(resp, 'lxml')

with open('./part.html', 'w', encoding='UTF-8') as file:
    file.write(str(soup))

print(soup.prettify())
```

```python
from lxml import etree
```

```python
html = etree.HTML(str(soup))

title = html.xpath('//h1[@data-scroll-trigger-for-sticky-product-header]/text()')[0].strip()

title
```

```python
list_oe = [_.strip() for _ in html.xpath('//div[@class="oe-number"]/a/text()')]

oe = ';'.join(list(set(list_oe)))

oe
```

```python
src = html.xpath('//div[@class="image product-main-image"]/a/@href')[0].strip()

src
```

```python
list_div = html.xpath('//div[@class="item-article-number"]')

list_div
```

```python
df_temp = pd.DataFrame([{'status': 'ok',
                         '序号': '',
                         'Title': title,
                         'OE': oe,
                         'Kit': '',
                         'Pic': '',
                         'Url': url,
                         'Src': src}])

for div in list_div:
    df_temp.loc[0, div.xpath('./span[1]/text()')[0].strip()[:-1].upper()] = div.xpath('./span[2]/text()')[0].strip()

df = pd.concat([df, df_temp], ignore_index=True).fillna('')

df
```
