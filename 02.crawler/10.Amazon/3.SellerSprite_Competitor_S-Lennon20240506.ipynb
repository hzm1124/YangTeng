```python
for a in range(len(input_)):
    crawler_status = 'error'
    
    # = = = = = = = = = = = = = = =
    
    browser.get('https://www.sellersprite.com/v3/competitor-lookup?market=US&monthName=bsr_sales_nearly&asins=%5B%22' + input_.loc[a, 'ASIN'] + '%22%5D&page=1&nodeIdPaths=%5B%5D&symbolFlag=false&size=100&order%5Bfield%5D=total_units&order%5Bdesc%5D=true')
    
    # = = = = = = = = = = = = = = =
    
    wait.until(EC.element_to_be_clickable((By.XPATH, '//div[@class="el-table__body-wrapper is-scrolling-none"]/table/tbody/tr')))
    tr, tr_2 = browser.find_elements(by=By.XPATH, value='//div[@class="el-table__body-wrapper is-scrolling-none"]/table/tbody/tr')
    
    # = = = = = = = = = = = = = = =
    
    list_tag = [div.text.strip() for div in tr.find_elements(by=By.XPATH, value='./td[3]/div/div[@class="tags"]')]
    while '' in list_tag:
        list_tag.remove('')

    tag = ';'.join(list_tag)
    
    # = = = = = = = = = = = = = = =
    
    parent_asin = tr.find_element(by=By.XPATH, value='./td[4]/div/div/div[3]/span[@class="text-muted"]').text.split(':')[1].strip()

    # = = = = = = = = = = = = = = =
    
    brand = '' if len(tr.find_elements(by=By.XPATH, value='./td[4]/div/div/div[4]/div')) == 0 else tr.find_elements(by=By.XPATH, value='./td[4]/div/div/div[4]/div')[0].text.strip()
    
    # = = = = = = = = = = = = = = =
    
    brand_url = '' if len(tr.find_elements(by=By.XPATH, value='./td[4]/div/div/div[4]/a[2]')) == 0 else tr.find_elements(by=By.XPATH, value='./td[4]/div/div/div[4]/a[2]')[0].get_attribute('href').strip()

    # = = = = = = = = = = = = = = =
    
    bsr_rank = tr.find_element(by=By.XPATH, value='./td[5]/div/div/div[1]').text.strip()

    # = = = = = = = = = = = = = = =
    
    bsr_increase = tr.find_element(by=By.XPATH, value='./td[5]/div/div/div[2]').text.strip()
    
    # = = = = = = = = = = = = = = =
    
    bsr_rate = tr.find_element(by=By.XPATH, value='./td[5]/div/div/div[3]').text.strip()
    
    # = = = = = = = = = = = = = = =
    
    parent_amount = tr.find_element(by=By.XPATH, value='./td[7]/div/div/div[1]').text.strip()
    
    # = = = = = = = = = = = = = = =
    
    parent_rate = tr.find_element(by=By.XPATH, value='./td[7]/div/div/div[2]').text.strip()
    
    # = = = = = = = = = = = = = = =
    
    parent_sale = tr.find_element(by=By.XPATH, value='./td[8]/div/div/div').text.strip()
    
    # = = = = = = = = = = = = = = =
    
    child_amount = tr.find_element(by=By.XPATH, value='./td[9]/div/div/span').text.strip()
    
    # = = = = = = = = = = = = = = =
    
    child_sale = tr.find_element(by=By.XPATH, value='./td[9]/div/div/div').text.strip()
    
    # = = = = = = = = = = = = = = =
    
    variant = tr.find_element(by=By.XPATH, value='./td[10]').text.strip()

    # = = = = = = = = = = = = = = =
    
    price_buybox = tr.find_element(by=By.XPATH, value='./td[11]/div/div/div/div[1]').text.strip()

    # = = = = = = = = = = = = = = =
    
    q_a = tr.find_element(by=By.XPATH, value='./td[11]/div/div/div/div[2]').text.strip()
    
    # = = = = = = = = = = = = = = =
    
    rating_number = tr.find_element(by=By.XPATH, value='./td[12]/div/div/div[1]').text.strip()

    # = = = = = = = = = = = = = = =
    
    new_rating_number = tr.find_element(by=By.XPATH, value='./td[12]/div/div/div[2]').text.strip()

    # = = = = = = = = = = = = = = =
    
    rating = tr.find_element(by=By.XPATH, value='./td[13]/div/div/div[1]').text.strip()

    # = = = = = = = = = = = = = = =
    
    retention_rate = tr.find_element(by=By.XPATH, value='./td[13]/div/div/div[2]').text.strip()

    # = = = = = = = = = = = = = = =
    
    fba_shipping = tr.find_element(by=By.XPATH, value='./td[14]/div/div/*[1]').text.strip()

    # = = = = = = = = = = = = = = =
    
    gross_margin = tr.find_element(by=By.XPATH, value='./td[14]/div/div/*[2]').text.strip()

    # = = = = = = = = = = = = = = =
    
    release_date = tr.find_element(by=By.XPATH, value='./td[15]/div/span').text.strip()

    # = = = = = = = = = = = = = = =
    
    release_date_2 = tr.find_element(by=By.XPATH, value='./td[15]/div/div/span').text.strip()
    
    # = = = = = = = = = = = = = = =
    
    sellers = tr.find_element(by=By.XPATH, value='./td[16]').text.strip()
    
    # = = = = = = = = = = = = = = =
    
    delivery = tr.find_element(by=By.XPATH, value='./td[17]').text.strip()
    
    # = = = = = = = = = = = = = = =
    
    buybox_seller = tr.find_element(by=By.XPATH, value='./td[18]').text.strip()
    
    # = = = = = = = = = = = = = = =
    
    lqs = tr.find_element(by=By.XPATH, value='./td[19]').text.strip()
    
    # = = = = = = = = = = = = = = =
    
    weight = tr.find_element(by=By.XPATH, value='./td[20]/div/div[1]').text.strip()
    
    # = = = = = = = = = = = = = = =
    
    volume = tr.find_element(by=By.XPATH, value='./td[20]/div/div[2]').text.strip()
    
    # = = = = = = = = = = = = = = =
    
    size = tr.find_element(by=By.XPATH, value='./td[20]/div/div[3]').text.strip()
    
    # = = = = = = = = = = = = = = =
    
    category_path = ' -> '.join([a.text.strip() for a in tr_2.find_elements(by=By.XPATH, value='./td/div[1]/div/div/div/div[1]/div[1]/span/span[not(@class)]/a')])

    # = = = = = = = = = = = = = = =
    
    category_path_2 = ' -> '.join([span.text.strip() for span in tr_2.find_elements(by=By.XPATH, value='./td/div[1]/div/div/div/div[1]/div[2]/div[1]/span/span[2]/span/span')])

    # = = = = = = = = = = = = = = =
    
    list_small_category_1 = [span.text.strip() for span in tr_2.find_elements(by=By.XPATH, value='./td/div[1]/div/div/div/div[1]/div[2]/div[2]/div/span')]

    # = = = = = = = = = = = = = = =
    
    list_small_category_2 = [a.text.strip() for a in tr_2.find_elements(by=By.XPATH, value='./td/div[1]/div/div/div/div[1]/div[2]/div[2]/div/a')]

    # = = = = = = = = = = = = = = =
    
    list_small_category_3 = [a.get_attribute('href').strip() for a in tr_2.find_elements(by=By.XPATH, value='./td/div[1]/div/div/div/div[1]/div[2]/div[2]/div/a')]

    # = = = = = = = = = = = = = = =
    
    small_category = ''
    for small_category_1, small_category_2, small_category_3 in zip(list_small_category_1, list_small_category_2, list_small_category_3):
        small_category += small_category_1 + ' - ' + small_category_2 + ' - ' + small_category_3 + '\n'

    small_category = small_category.strip()
    
    # = = = = = = = = = = = = = = =
    
    df_temp = pd.DataFrame([{'ASIN': input_.loc[a, 'ASIN'],
                             '标识': tag,
                             '父ASIN': parent_asin,
                             '品牌': brand,
                             '品牌链接': brand_url,
                             '大类BSR排名': bsr_rank,
                             '近7天BSR增长数': bsr_increase,
                             '近7天BSR增长率': bsr_rate,
                             '父体近30天总销量': parent_amount,
                             '父体近30天销量增长率': parent_rate,
                             '父体近30天总销售额': parent_sale,
                             '子体近一个月的销量': child_amount,
                             '子体近一个月的预估销售额': child_sale,
                             '变体数': variant,
                             '最新售价 (BuyBox)': price_buybox,
                             '最新Q&A数': q_a,
                             'Listing的最新评分数': rating_number,
                             'Listing近30天的新增评分数': new_rating_number,
                             'Listing的最新评分值': rating,
                             '月度留评率': retention_rate,
                             'FBA运费': fba_shipping,
                             '毛利率': gross_margin,
                             '上架时间_1': release_date,
                             '上架时间_2': release_date_2,
                             '卖家数': sellers,
                             '配送': delivery,
                             'BuyBox卖家': buybox_seller,
                             'LQS': lqs,
                             '重量': weight,
                             '体积': volume,
                             '尺寸': size,
                             '类目路径_1': category_path,
                             '类目路径_2': category_path_2,
                             '小类目': small_category}])
    
    output_correct = pd.concat([output_correct, df_temp], ignore_index=True).fillna('')
    
    # = = = = = = = = = = = = = = =
    
    crawler_status = 'ok'
    
    print('[' + crawler_status + '] - ' + input_.loc[a, 'ASIN'] + '\n[完成数量：' + str(a+1) + '] - [当前时间：' + datetime.now().strftime('%H:%M:%S') + ']\n')
    
    # = = = = = = = = = = = = = = =
    
browser.quit()

print('输出ing...')
output_correct = output_correct.drop_duplicates(ignore_index=True)
output_correct = output_correct.sort_values(by=['ASIN'], ascending=[True]).reset_index(drop=True)
output_correct.to_excel('./sellersprite_sales-' + datetime.now().strftime('%Y%m%d_%H%M%S') + '.xlsx', index=False)
print('Done ~')
```
