```python
import pandas as pd

# = = =  = = =  = = =

from tqdm import tqdm
import json
```

```python
print('Data Loading...')
df_input = pd.read_excel('./file/2.part.xlsx',
                         header=0,
                         dtype=str).fillna('')

print('Loading Done !')
print()

df_input['No'] = df_input['No'].astype(int)

df_input
```

```python
list_series, list_column, list_column_overview, list_column_information = [], df_input.columns.tolist(), [], []
for _, series in tqdm(df_input.iterrows(), total=len(df_input), desc='Progress', ncols=77):
    dict_dict, dict_dict_2, dict_dict_3, dict_dict_4, dict_dict_5, dict_dict_6 = json.loads(series['Json_Overview']), json.loads(series['Json_Technical']), json.loads(series['Json_Additional']), json.loads(series['Json_Technical 2']), json.loads(series['Json_Additional 2']), json.loads(series['Json_Information'])
    dict_, dict_2, dict_3, dict_4, dict_5, dict_6 = {k: v for dict_ in dict_dict.values() for k, v in dict_.items()}, {k: v for dict_ in dict_dict_2.values() for k, v in dict_.items()}, {k: v for dict_ in dict_dict_3.values() for k, v in dict_.items()}, {k: v for dict_ in dict_dict_4.values() for k, v in dict_.items()}, {k: v for dict_ in dict_dict_5.values() for k, v in dict_.items()}, {k: v for dict_ in dict_dict_6.values() for k, v in dict_.items()}

    dict_key = {}
    for key, value in dict_2.items():
        if key in dict_key:
            dict_key[key] += 1
        else:
            dict_key[key] = 1
    for key, value in dict_3.items():
        if key in dict_key:
            dict_key[key] += 1
        else:
            dict_key[key] = 1
    for key, value in dict_4.items():
        if key in dict_key:
            dict_key[key] += 1
        else:
            dict_key[key] = 1
    for key, value in dict_5.items():
        if key in dict_key:
            dict_key[key] += 1
        else:
            dict_key[key] = 1
    for key, value in dict_6.items():
        if key in dict_key:
            dict_key[key] += 1
        else:
            dict_key[key] = 1

    dict_info, dict_count = {}, {}
    for key, value in dict_2.items():
        if dict_key[key] == 1:
            dict_info[key] = value
        else:
            if key in dict_count:
                dict_count[key] += 1
            else:
                dict_count[key] = 1
            dict_info[f'{key}-{dict_count[key]}'] = value
    for key, value in dict_3.items():
        if dict_key[key] == 1:
            dict_info[key] = value
        else:
            if key in dict_count:
                dict_count[key] += 1
            else:
                dict_count[key] = 1
            dict_info[f'{key}-{dict_count[key]}'] = value
    for key, value in dict_4.items():
        if dict_key[key] == 1:
            dict_info[key] = value
        else:
            if key in dict_count:
                dict_count[key] += 1
            else:
                dict_count[key] = 1
            dict_info[f'{key}-{dict_count[key]}'] = value
    for key, value in dict_5.items():
        if dict_key[key] == 1:
            dict_info[key] = value
        else:
            if key in dict_count:
                dict_count[key] += 1
            else:
                dict_count[key] = 1
            dict_info[f'{key}-{dict_count[key]}'] = value
    for key, value in dict_6.items():
        if dict_key[key] == 1:
            dict_info[key] = value
        else:
            if key in dict_count:
                dict_count[key] += 1
            else:
                dict_count[key] = 1
            dict_info[f'{key}-{dict_count[key]}'] = value

    # = = =  = = =  = = =

    for key, value in dict_.items():
        key = f'[Overview] {key}'

        if key not in list_column_overview:
            list_column_overview.append(key)

        series[key] = value

    # = = =  = = =  = = =

    for key, value in dict_info.items():
        if key in list_column:
            key = f'{key} (Information)'

        if key not in list_column_information:
            list_column_information.append(key)

        series[key] = value

    list_series.append(series)

# = = =  = = =  = = =

print()
print('Data Outputting...')
df = pd.DataFrame(list_series).sort_values(by=['No'],
                                           ascending=[True],
                                           ignore_index=True)
df[list_column[:6] + sorted(list_column_overview) + sorted(list_column_information) + list_column[6:]].to_excel('./file/3.part.xlsx', index=False)
print('Output Finished !')
print()

df
```
