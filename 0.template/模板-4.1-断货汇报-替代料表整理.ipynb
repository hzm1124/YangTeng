# 0. 配置参数

- `alternate`：替代料总表路径
- `mrp`：MRP表路径
- `temp_out`：保存结果的路径
- `out`：保存结果的路径

```python
alternate = '../../file/替代料总表01-30.xlsx'
mrp = '../../file/US MRP20230130运营.xlsx'

temp_out = '../../file/Lennon_断货汇报_temp替代料表_NOW().xlsx'
out = '../../file/Lennon_断货汇报_整理好的替代料表_NOW().xlsx'
```

```python
import pandas as pd
```

```python
temp_df_alternate = pd.read_excel(alternate, sheet_name='MRP SKU库存', header=1)[['主料SKU', '替换料1', '替换料2', '替换料3', '替换料4', '替换料5', '替换料6']].fillna('')

temp_df_alternate
```

```python
df_mrp = pd.read_excel(mrp, header=1)[['MRP SKU']]

df_mrp
```

# 1. 调整替代料表

```python
temp_df_alternate = pd.merge(left=temp_df_alternate, right=df_mrp, how='left', left_on='主料SKU', right_on='MRP SKU').fillna('')
temp_df_alternate = pd.merge(left=temp_df_alternate, right=df_mrp, how='left', left_on='替换料1', right_on='MRP SKU', suffixes=('', '_1')).fillna('')
temp_df_alternate = pd.merge(left=temp_df_alternate, right=df_mrp, how='left', left_on='替换料2', right_on='MRP SKU', suffixes=('', '_2')).fillna('')
temp_df_alternate = pd.merge(left=temp_df_alternate, right=df_mrp, how='left', left_on='替换料3', right_on='MRP SKU', suffixes=('', '_3')).fillna('')
temp_df_alternate = pd.merge(left=temp_df_alternate, right=df_mrp, how='left', left_on='替换料4', right_on='MRP SKU', suffixes=('', '_4')).fillna('')
temp_df_alternate = pd.merge(left=temp_df_alternate, right=df_mrp, how='left', left_on='替换料5', right_on='MRP SKU', suffixes=('', '_5')).fillna('')
temp_df_alternate = pd.merge(left=temp_df_alternate, right=df_mrp, how='left', left_on='替换料6', right_on='MRP SKU', suffixes=('', '_6')).fillna('')
temp_df_alternate['MRP SKU'] = ['' if i == '' else '0' for i in temp_df_alternate['MRP SKU']]
temp_df_alternate['MRP SKU_1'] = ['' if i == '' else '1' for i in temp_df_alternate['MRP SKU_1']]
temp_df_alternate['MRP SKU_2'] = ['' if i == '' else '2' for i in temp_df_alternate['MRP SKU_2']]
temp_df_alternate['MRP SKU_3'] = ['' if i == '' else '3' for i in temp_df_alternate['MRP SKU_3']]
temp_df_alternate['MRP SKU_4'] = ['' if i == '' else '4' for i in temp_df_alternate['MRP SKU_4']]
temp_df_alternate['MRP SKU_5'] = ['' if i == '' else '5' for i in temp_df_alternate['MRP SKU_5']]
temp_df_alternate['MRP SKU_6'] = ['' if i == '' else '6' for i in temp_df_alternate['MRP SKU_6']]

temp_df_alternate
```

```python
import re
```

```python
list_note = ';' + temp_df_alternate['MRP SKU'] + ';' + temp_df_alternate['MRP SKU_1'] + ';' + temp_df_alternate['MRP SKU_2'] + ';' + temp_df_alternate['MRP SKU_3'] + ';' + temp_df_alternate['MRP SKU_4'] + ';' + temp_df_alternate['MRP SKU_5'] + ';' + temp_df_alternate['MRP SKU_6'] + ';'
temp_df_alternate['note'] = [re.sub(r';{2,}', ';', i)[1:-1] for i in list_note]
temp_df_alternate = temp_df_alternate[['主料SKU', '替换料1', '替换料2', '替换料3', '替换料4', '替换料5', '替换料6', 'note']]

temp_df_alternate
```

```python
df_0 = temp_df_alternate[temp_df_alternate['note'] == '0']

df_0
```

```python
df_1 = temp_df_alternate[(temp_df_alternate['note'] == '1') | (temp_df_alternate['note'] == '2') | (temp_df_alternate['note'] == '3') | (temp_df_alternate['note'] == '4') | (temp_df_alternate['note'] == '5') | (temp_df_alternate['note'] == '6')].reset_index(drop=True)

df_1
```

```python
df_2_1 = temp_df_alternate[(temp_df_alternate['note'] == '0;1') | (temp_df_alternate['note'] == '0;2')].reset_index(drop=True)
df_2_2 = df_2_1.copy()

df_2_1
```

```python
for i in range(len(df_2_1)):
    list_note = df_2_1.loc[i, 'note'].split(';')
    # 依次向前，去除这个SKU且保留后面的全部
    for j in range(int(list_note[1]), 6):
        df_2_1.loc[i, '替换料' + str(j)] = df_2_1.loc[i, '替换料' + str(j+1)]

df_0 = pd.concat([df_0, df_2_1]).reset_index(drop=True)

df_0
```

```python
for i in range(len(df_1)):
    if df_1.loc[i, '替换料' + df_1.loc[i, 'note']] not in df_0['主料SKU'].tolist():
        # 主元法：直接把不在MRP表中的主料SKU和在MRP表中的替换料x进行调换
        df_1.loc[i, '主料SKU'], df_1.loc[i, '替换料' + df_1.loc[i, 'note']] = df_1.loc[i, '替换料' + df_1.loc[i, 'note']], df_1.loc[i, '主料SKU']
        df_0 = pd.concat(df_0, df_1.loc[[i]]).reset_index(drop=True)

df_0
```

```python
for i in range(len(df_2_2)):
    list_note = df_2_2.loc[i, 'note'].split(';')
    # 先查重，不重复就添加
    if df_2_2.loc[i, '替换料' + list_note[1]] not in df_0['主料SKU'].tolist():
        df_0 = pd.concat(df_0, pd.DataFrame({'主料SKU': [df_2_2.loc[i, '替换料' + list_note[1]]],
                                             '替换料1': [''],
                                             '替换料2': [''],
                                             '替换料3': [''],
                                             '替换料4': [''],
                                             '替换料5': [''],
                                             '替换料6': ['']})).reset_index(drop=True)

df_0
```

```python
df_0 = df_0[['主料SKU', '替换料1', '替换料2', '替换料3', '替换料4', '替换料5', '替换料6']]

df_0
```

# 2. 保存

```python
import datetime

writer = pd.ExcelWriter(temp_out.replace('NOW()', datetime.datetime.now().strftime("%Y%m%d_%H%M%S")))
temp_df_alternate.to_excel(writer, index=False)
writer.close()
writer = pd.ExcelWriter(out.replace('NOW()', datetime.datetime.now().strftime("%Y%m%d_%H%M%S")))
df_0.to_excel(writer, index=False)
writer.close()
```
