```python
param_date = '20240706'
```

```python
import pandas as pd

input_ = pd.read_excel(f'./{param_date}/8.url_1.xlsx',
                       header=0,
                       dtype=str).fillna('')

input_ = input_[['Url']].drop_duplicates(['Url'],
                                         keep='first',
                                         ignore_index=True)
input_ = input_.sort_values(by=['Url'],
                            ascending=[True],
                            ignore_index=True)
input_['No'] = [i+1 for i in range(len(input_))]

input_.to_excel(f'./{param_date}/9.url.xlsx', index=False)

input_
```
