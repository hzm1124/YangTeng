```python
param_date = '20240706'
```

```python
import os

os.makedirs(f'./{param_date}/0.error')
os.makedirs(f'./{param_date}/4.model_2')
os.makedirs(f'./{param_date}/5.type_1')
```

```python
import pandas as pd

input_ = pd.read_excel(f'./{param_date}/3.model.xlsx',
                       header=0,
                       dtype=str).fillna('')

input_['Year'] = input_['Year'].astype(int)
input_['Make No'] = input_['Make No'].astype(int)
input_['Model No'] = input_['Model No'].astype(int)

input_
```

```python
input_2 = input_[['Make']].drop_duplicates(keep='first', ignore_index=True)
input_2 = input_2.sort_values(by=['Make'],
                              ascending=[True],
                              ignore_index=True)

input_2['Make No'] = [_+1 for _ in range(len(input_2))]
input_2['File Make'] = [_.strip().replace('/', '\\') for _ in input_2['Make'].tolist()]

input_2
```

```python
from tqdm import tqdm

for i in tqdm(range(len(input_2)), desc='Progress', ncols=77):
    input_3 = input_[input_['Make'] == input_2.loc[i, 'Make']].copy()
    input_3 = input_3.sort_values(by=['Year', 'Make No', 'Model No'],
                                      ascending=[False, True, True],
                                      ignore_index=True)

    input_3.to_excel(f'''./{param_date}/4.model_2/make_{input_2.loc[i, 'Make No']}.{input_2.loc[i, 'File Make']}.xlsx''', index=False)

print('Done ~')
```
