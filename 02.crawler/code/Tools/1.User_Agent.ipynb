```python
list_win, list_mac = [], []
```

```python
import requests

from bs4 import BeautifulSoup

from parsel import Selector

import re
```

```python
for times in range(10):
    for page in range(1, 10 + 1):
        url = f'https://useragents.io/explore/devices/types/desktop/maker/google-inc-f3d?p={page}'
    
        attempt = 0
        while True:
            attempt += 1
            print(f'{times}: {url} - {attempt}')
            try:
                response = requests.get(url, allow_redirects=False)
                break
            except:
                continue
    
        soup = BeautifulSoup(response.text, 'html5lib')
    
        selector = Selector(text=str(soup))
    
        list_text = [text.strip() for text in selector.xpath('//table[@class="table"]/tbody/tr/td/a/text()').getall()]
    
        for text in list_text:
            if text.startswith('Mozilla/5.0 (Windows NT 10.0; Win64; x64') and text.endswith('Safari/537.36') and re.search(r' Chrome/1[0-9][0-9]\.', text):
                if text not in list_win:
                    list_win.append(text)
            if text.startswith('Mozilla/5.0 (Macintosh; Intel Mac OS X') and text.endswith('Safari/537.36') and re.search(r' Chrome/1[0-9][0-9]\.', text):
                if text not in list_mac:
                    list_mac.append(text)
```
