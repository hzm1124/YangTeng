```python
import pandas as pd

input_ = pd.read_excel('./2.url.xlsx',
                       header=0,
                       dtype=str).fillna('')

input_['1st Category No'] = input_['1st Category No'].astype(int)
input_['2nd Category No'] = input_['2nd Category No'].astype(int)
input_['3rd Category No'] = input_['3rd Category No'].astype(int)
input_['Page'] = input_['Page'].astype(int)
input_['No'] = input_['No'].astype(int)

input_
```

```python
output = input_.drop_duplicates(['Url'],
                                keep='first',
                                ignore_index=True)

output['No'] = [_+1 for _ in range(len(output))]
output[['No', 'Url']].to_excel('./3.url_2.xlsx', index=False)

output
```

```python
output = input_.drop_duplicates(['1st Category No', '1st Category', '1st Category Url', '2nd Category No', '2nd Category', '2nd Category Url', '3rd Category No', '3rd Category', '3rd Category Url'],
                                keep='last',
                                ignore_index=True)

list_column = ['SKU Count' if _ == 'No' else _ for _ in output.columns]
output.columns = list_column
output[['1st Category No',
        '1st Category',
        '1st Category Url',
        '2nd Category No',
        '2nd Category',
        '2nd Category Url',
        '3rd Category No',
        '3rd Category',
        '3rd Category Url',
        'SKU Count']].to_excel('./3.catalog.xlsx', index=False)

output
```
