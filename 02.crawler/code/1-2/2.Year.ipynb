```python
item = {'make_url': 'https://www.rockauto.com/en/catalog/',
 'make_no': 178,
 'make': 'MAZDA',
 'make_market': 'US,CA,MX',
 'year_url': 'https://www.rockauto.com/en/catalog/MAZDA'}

item
```

```python
import json
import requests

import sys
sys.path.append('../00.Tools')
from configuration import generate_header, generate_proxy
```

```python
data = {'func': 'navnode_fetch',
        'payload': json.dumps({'jsn': {'tab': 'catalog',
                                       'make': item['make'],
                                       'nodetype': 'make'}}),
        'api_json_request': '1'}
response = requests.post('https://www.rockauto.com/catalog/catalogapi.php',
                         data=data,
                         headers=generate_header(),
                         proxies=generate_proxy(),
                         timeout=(10, 15),
                         allow_redirects=False)

response.status_code
```

```python
dict_ = response.json()

dict_
```

```python
'redirect_to_url' in dict_ or dict_['jsn']['tab'].strip() != 'catalog' or dict_['jsn']['make'].strip() != item['make'].strip() or dict_['jsn']['nodetype'].strip() != 'make' or dict_['jsn']['idepth'] != 0
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(dict_['html_fill_sections']['navchildren[]'], 'html5lib')

with open('./2.year.html', 'w', encoding='utf-8') as file:
    file.write(str(soup))

print(soup.prettify())
```
