```python
import io,os,urllib,gevent,time,random,xlsxwriter
from PIL import Image
from io import BytesIO
from urllib.request import urlopen
from gevent import monkey
monkey.patch_all()
from gevent.queue import Queue
import pandas as pd
fixed_size=266
df_new=pd.read_excel(r'D:/raybestos,group1,parking+brake+cable,1696.xlsx',sheet_name='Sheet1')
url_list=df_new
work=Queue()
rockname='uiu'
tt=time.strftime('%m月%d日%H_%M_%S', time.localtime())
wb = xlsxwriter.Workbook(r'D:/{}商品{}数据.xlsx'.format(tt,rockname))
worksheet = wb.add_worksheet("图片") 
worksheet.write_string(0,1,'图片地址')
worksheet.write_string(0,2,'图片')
print('开始')
for url_sku in zip(url_list.index,url_list['主图']):
    work.put_nowait(url_sku)
def crawler():
    while not work.empty():
        url_sku=work.get_nowait()
        index_1=url_sku[0]+1
        i=url_sku[1]
        url1 =i
#         print(work.qsize(),url1)
        try:
            if url1 !='-':
                image_data = io.BytesIO(urllib.request.urlopen(url1).read())
                img = Image.open(image_data)
                x = float(fixed_size / img.size[0])
                y = float(fixed_size / img.size[1])
                worksheet.write_string(index_1,1,"{}".format(url1))
                worksheet.insert_image(row=index_1,col=2,filename=url1,options={'x_scale': x,
                                                                                'y_scale': y,
                                                                                'image_data': image_data,
                                                                                'positioning': 1})
                print(str(index_1)+"    "+url1+' 写入完成')
                time.sleep(random.random())
            else:
                worksheet.write_string(index_1,1,"{}".format(url1))
                worksheet.write_string(index_1,2,"-")
        except:
            worksheet.write_string(index_1,1,"{}".format(url1))
            worksheet.write_string(index_1,2,"-")
            
task_list=[]
for x in range(100):
    task=gevent.spawn(crawler)
    task_list.append(task)
gevent.joinall(task_list)

wb.close()
print(time.strftime('%m月%d日%H:%M:%S', time.localtime()))

#未更改项
wb2 = xlsxwriter.Workbook(r'D:/{}商品{}未更改数据.xlsx'.format(tt,rockname))
worksheet = wb2.add_worksheet("图片") 
worksheet.write_string(0,1,'图片地址')
worksheet.write_string(0,2,'图片')
df_new=url_list
url_list=pd.read_excel(r'D:/{}商品{}数据.xlsx'.format(tt,rockname))
print(len(url_list[url_list['图片'].isin(['-'])]['图片地址']))
for url_sku in zip(url_list[url_list['图片'].isin(['-'])].index,url_list[url_list['图片'].isin(['-'])]['图片地址']):
    work.put_nowait(url_sku)
def crawler():
    while not work.empty():
        url_sku=work.get_nowait()
        index_1=url_sku[0]
        i=url_sku[1]
        url1 =i
#         print(work.qsize(),url1)
        try:
            if url1 !='-':
                image_data = io.BytesIO(urllib.request.urlopen(url1).read())
                img = Image.open(image_data)
                x = float(fixed_size / img.size[0])
                y = float(fixed_size / img.size[1])
                worksheet.write_string(index_1,1,"{}".format(url1))
                worksheet.insert_image(row=index_1,col=2,filename=url1,options={'x_scale': x,
                                                                                'y_scale': y,
                                                                                'image_data': image_data,
                                                                                'positioning': 1})
                print(str(index_1)+"    "+url1+' 写入完成')
                time.sleep(random.random())
            else:
                worksheet.write_string(index_1,1,"{}".format(url1))
                worksheet.write_string(index_1,2,"-")
        except:
            worksheet.write_string(index_1,1,"{}".format(url1))
            worksheet.write_string(index_1,2,"-")
            
task_list=[]
for x in range(50):
    task=gevent.spawn(crawler)
    task_list.append(task)
gevent.joinall(task_list)     
wb2.close()
print(time.strftime('%m月%d日%H:%M:%S', time.localtime()))
```
