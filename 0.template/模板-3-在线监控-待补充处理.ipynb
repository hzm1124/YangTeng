```python
import pandas as pd
```

```python
df_data = pd.read_excel(complement, header=0)

df_data
```

```python
import re
```

```python
def process_list(list_sku, test=False):
    list_sku_amount = ['1'] * len(list_sku)
    
    if test:
        print(f'list_sku_amount: {list_sku_amount}')
    
    syntax = re.compile('\*[2-9]$')
    for i in range(len(list_sku)):
        result = syntax.search(list_sku[i])
        if result != None:
            list_sku_amount[i] = str(result.group().split('*')[1])
    
    if test:
        print(f'list_sku_amount: {list_sku_amount}')
    
    list_pt_sku = []
    list_new_sku = [''] * 6
    for i in range(len(list_sku)):
        if list_sku[i].startswith('US'):
            list_pt_sku.append(re.sub(r'^US', '', list_sku[i]))
            list_new_sku[i] = re.sub(r'\*[2-9]$', '', list_sku[i])
        else:
            list_pt_sku.append(list_sku[i])
            list_new_sku[i] = 'US' + re.sub(r'\*[2-9]$', '', list_sku[i])

    if test:
        print(f'list_pt_sku: {list_pt_sku}')
        print(f'list_new_sku: {list_new_sku}')
    
    note = ''
    syntax = re.compile('^[0-9]+(\*[2-9])?$')
    for i in list_pt_sku:
        result = syntax.search(i)
        if result != None:
            note = 'error'
            break
        
    return '+'.join(list_pt_sku), '+'.join(list_sku_amount), list_new_sku[0], list_new_sku[1], list_new_sku[2], list_new_sku[3], list_new_sku[4], list_new_sku[5], note

# process_list(['US04342*2', 'USBHH02837*3', 'USBHH02838'], test=True)
```

```python
list_pt_sku = []
list_amount = []
dict_sku = {'sku_1':[], 'sku_2':[], 'sku_3':[], 'sku_4':[], 'sku_5':[], 'sku_6':[]}
list_note = []
for i in range(len(df_data)):
    list_sku = df_data.loc[i, '处理后SKU'].split('+')
    pt_sku, amount, sku_1, sku_2, sku_3, sku_4, sku_5, sku_6, note = process_list(list_sku)
    list_pt_sku.append(pt_sku)
    list_amount.append(amount)
    dict_sku['sku_1'].append(sku_1)
    dict_sku['sku_2'].append(sku_2)
    dict_sku['sku_3'].append(sku_3)
    dict_sku['sku_4'].append(sku_4)
    dict_sku['sku_5'].append(sku_5)
    dict_sku['sku_6'].append(sku_6)
    list_note.append(note)
```

```python
df_data['PT SKU'] = list_pt_sku
df_data['数量'] = list_amount
df_data['sku_1'] = dict_sku['sku_1']
df_data['sku_2'] = dict_sku['sku_2']
df_data['sku_3'] = dict_sku['sku_3']
df_data['sku_4'] = dict_sku['sku_4']
df_data['sku_5'] = dict_sku['sku_5']
df_data['sku_6'] = dict_sku['sku_6']
df_data['note'] = list_note

df_data
```

```python
import datetime

writer = pd.ExcelWriter(out.replace('NOW()', datetime.datetime.now().strftime("%Y%m%d_%H%M%S")))
df_data.to_excel(writer, index=False)
writer.close()
```
