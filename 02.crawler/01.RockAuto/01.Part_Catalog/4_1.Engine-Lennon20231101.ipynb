```python
import pandas as pd
```

```python
df = pd.read_excel('./3.model.xlsx',
                   header=0,
                   dtype=str).fillna('')

list_make = df['Make'].tolist()
list_year = df['Year'].tolist()
list_model = df['Model'].tolist()

print('总数量：' + str(len(list_model)))
list_model
```

```python
df = pd.DataFrame()

df
```

```python
from urllib.parse import quote
import requests
import json
```

```python
i = 0

make = list_make[0]
year = list_year[0]
model = list_model[0]

data = 'func=navnode_fetch&payload=' + quote('{"jsn":{"tab":"catalog","make":"') + make.replace(' ', '+') + quote('","year":"') + year + quote('","model":"') + model.replace('+', '%2B').replace(' ', '+').replace('&', '%26').replace('/', '%2F') + quote('","nodetype":"model"}}') + '&api_json_request=1&sctchecked=1&scbeenloaded=true&curCartGroupID='

resp = requests.post('https://www.rockauto.com/catalog/catalogapi.php',
                     data=data,
                     headers={'Content-Type': 'application/x-www-form-urlencoded; charset=UTF-8'}).text

resp = json.loads(resp)

resp
```

```python
from bs4 import BeautifulSoup
```

```python
html = resp['html_fill_sections']['navchildren[]']

soup = BeautifulSoup(html, 'lxml')

with open('./engine.html', 'w', encoding='UTF-8') as file:
    file.write(str(soup))

print(soup.prettify())
```

```python
from lxml import etree
```

```python
html = etree.HTML(str(soup))

list_engine = html.xpath('//td[@class="nlabel"]/a/text()')

list_engine
```

```python
list_url = html.xpath('//td[@class="niconspace ncollapsedicon"]/a/@href')

list_car_code = [_.split(',')[-1] for _ in list_url]

list_car_code
```

```python
list_url = ['https://www.rockauto.com'+_ for _ in list_url]

list_url
```

```python
df_temp = pd.DataFrame({'status': 'ok',
                        'Make': make,
                        'Year': year,
                        'Model': model,
                        'Engine': list_engine,
                        'Car_Code': list_car_code,
                        'Url': list_url})

df = pd.concat([df, df_temp], ignore_index=True).fillna('')

df
```
