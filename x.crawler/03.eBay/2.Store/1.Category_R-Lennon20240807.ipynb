```python
param_url = 'https://www.ebay.com/str/carpartswholesale'
```

```python
dict_param = {}
list_param = param_url.split('?')[1].split('&') if '?' in param_url else []
for param in list_param:
    dict_param[param.split('=')[0]] = param.split('=')[1]

dict_param
```

```python
dict_param.pop('_ipg', '')
dict_param['orig_cvip'] = 'true'

if param_url.startswith('https://www.ebay.com/'):
    dict_param['_ul'] = 'US'
    dict_param['_stpos'] = '91710'
elif param_url.startswith('https://www.ebay.de/'):
    dict_param['_ul'] = 'DE'
    dict_param['_stpos'] = '10115'
elif param_url.startswith('https://www.ebay.co.uk/'):
    dict_param['_ul'] = 'GB'
    dict_param['_stpos'] = 'BS81QU'
elif param_url.startswith('https://www.ebay.com.au/'):
    dict_param['_ul'] = 'AU'
    dict_param['_stpos'] = '2019'
elif param_url.startswith('https://www.ebay.ca/'):
    dict_param['_ul'] = 'CA'
    dict_param['_stpos'] = 'M5S2E8'

dict_param
```

```python
param_url = param_url.split('?')[0] + '?'
for key, value in dict_param.items():
    param_url += f'{key}={value}&'

param_url
```

```python
list_dict_category = json_['o']['w'][4][2]['page']['modules']['STORE_CATEGORIES_MODULE']['categories']

list_dict_category
```

```python
import pandas as pd

output = pd.DataFrame()
for i, dict_category in zip(range(len(list_dict_category)), list_dict_category):
    type_ = dict_category['_type'].strip()

    # = = = = = = = = = = = = = = =

    list_dict_category_2 = dict_category['categories'] if 'categories' in dict_category else []

    list_type_2 = [dict_category_2['_type'].strip() for dict_category_2 in list_dict_category_2]
    list_label_2 = [dict_category_2['label']['textSpans'][0]['text'].strip() for dict_category_2 in list_dict_category_2]
    
    list_url_2 = []
    list_dict_param_2 = [dict_category_2['action']['params'] for dict_category_2 in list_dict_category_2]
    for j, dict_param_2 in zip(range(len(list_dict_param_2)), list_dict_param_2):
        new_dict_param = {}
        for key, value in dict_param.items():
            new_dict_param[key] = value
        for key, value in dict_param_2.items():
            new_dict_param[key] = value

        if param_url.startswith('https://www.ebay.com/'):
            url_2 = f'https://www.ebay.com/sch/i.html?'
        elif param_url.startswith('https://www.ebay.de/'):
            url_2 = f'https://www.ebay.de/sch/i.html?'
        elif param_url.startswith('https://www.ebay.co.uk/'):
            url_2 = f'https://www.ebay.co.uk/sch/i.html?'
        elif param_url.startswith('https://www.ebay.com.au/'):
            url_2 = f'https://www.ebay.com.au/sch/i.html?'
        elif param_url.startswith('https://www.ebay.ca/'):
            url_2 = f'https://www.ebay.ca/sch/i.html?'

        for key, value in new_dict_param.items():
            url_2 += f'{key}={value}&'

        url_2 += f'_ipg=240&_udlo=&_udhi='
        
        list_url_2.append(url_2)
    
    # = = = = = = = = = = = = = = =

    label = dict_category['label']['textSpans'][0]['text'].strip()

    # = = = = = = = = = = = = = = =

    new_dict_param = {}
    for key, value in dict_param.items():
        new_dict_param[key] = value
    for key, value in dict_category['action']['params'].items():
        new_dict_param[key] = value
    
    if param_url.startswith('https://www.ebay.com/'):
        url = f'https://www.ebay.com/sch/i.html?'
    elif param_url.startswith('https://www.ebay.de/'):
        url = f'https://www.ebay.de/sch/i.html?'
    elif param_url.startswith('https://www.ebay.co.uk/'):
        url = f'https://www.ebay.co.uk/sch/i.html?'
    elif param_url.startswith('https://www.ebay.com.au/'):
        url = f'https://www.ebay.com.au/sch/i.html?'
    elif param_url.startswith('https://www.ebay.ca/'):
        url = f'https://www.ebay.ca/sch/i.html?'

    for key, value in new_dict_param.items():
        url += f'{key}={value}&'

    url += f'_ipg=240&_udlo=&_udhi='

    # = = = = = = = = = = = = = = =

    df_temp = pd.DataFrame({'1st Category No': i+1,
                            '1st Category Type': type_,
                            '1st Category': label,
                            '1st Category Url': url,
                            '2nd Category No': [j+1 for j in range(len(list_label_2))] if list_label_2 else ['-'],
                            '2nd Category Type': list_type_2 if list_type_2 else ['-'],
                            '2nd Category': list_label_2 if list_label_2 else ['-'],
                            '2nd Category Url': list_url_2 if list_url_2 else ['-']})
    
    output = pd.concat([output, df_temp], ignore_index=True).fillna('')

# = = = = = = = = = = = = = = =

output = output.sort_values(by=['1st Category No', '2nd Category No'],
                            ascending=[True, True],
                            ignore_index=True)

output.to_excel('./1.category.xlsx', index=False)

output
```
