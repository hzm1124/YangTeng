```python
import pandas as pd

input_ = pd.read_excel(f'./4.vehicle_1.xlsx',
                       header=0,
                       dtype=str).fillna('')

input_
```

```python
input_2 = input_[['Part Number', 'OEM']].drop_duplicates(keep='first', ignore_index=True)
input_2 = input_2.sort_values(by=['Part Number'],
                              ascending=[True],
                              ignore_index=True)

input_2
```

```python
from tqdm import tqdm

for i in tqdm(range(len(input_2)), desc='Progress', ncols=77):
    df_temp = input_[input_['Part Number'] == input_2.loc[i, 'Part Number']].copy().reset_index(drop=True)

    # = = = = = = = = = = = = = = =

    dict_vehicle = {}
    for j in range(len(df_temp)):
        if '-' in df_temp.loc[j, 'Year']:
            year, year_2 = int(df_temp.loc[j, 'Year'].split('-')[0]), int(df_temp.loc[j, 'Year'].split('-')[1])
        else:
            year, year_2 = int(df_temp.loc[j, 'Year']), int(df_temp.loc[j, 'Year'])

        if df_temp.loc[j, 'Make'] not in dict_vehicle.keys():
            dict_vehicle[df_temp.loc[j, 'Make']] = {df_temp.loc[j, 'Model']: [year] if year == year_2 else [year, year_2]}
        elif df_temp.loc[j, 'Model'] not in dict_vehicle[df_temp.loc[j, 'Make']].keys():
            dict_vehicle[df_temp.loc[j, 'Make']][df_temp.loc[j, 'Model']] = [year] if year == year_2 else [year, year_2]
        else:
            if year not in dict_vehicle[df_temp.loc[j, 'Make']][df_temp.loc[j, 'Model']]:
                dict_vehicle[df_temp.loc[j, 'Make']][df_temp.loc[j, 'Model']].append(year)
            if year_2 not in dict_vehicle[df_temp.loc[j, 'Make']][df_temp.loc[j, 'Model']]:
                dict_vehicle[df_temp.loc[j, 'Make']][df_temp.loc[j, 'Model']].append(year_2)

    # = = = = = = = = = = = = = = =

    list_vehicle = []
    for make in dict_vehicle.keys():
        for model in dict_vehicle[make].keys():
            year_begin = min(dict_vehicle[make][model])
            year_end = max(dict_vehicle[make][model])
            list_vehicle.append(f'{make} {model} {year_begin}' if year_begin == year_end else f'{make} {model} {year_begin}-{year_end}')

    input_2.loc[i, 'Vehicle'] = '\n'.join(sorted(list_vehicle))

# = = = = = = = = = = = = = = =

input_2.to_excel('./5.vehicle.xlsx', index=False)

input_2
```
