```python
import pandas as pd

input_ = pd.read_excel('./4.auth.xlsx',
                       header=0,
                       dtype=str).fillna('')

input_
```

```python
from tqdm import tqdm
import json

for i in tqdm(range(len(input_)), desc='Progress', ncols=77):
    dict_vehicle = json.loads(input_.loc[i, 'Json_Info'])
    for value in dict_vehicle.values():
        input_.loc[i, list(value.keys())[0]] = list(value.values())[0]

input_.to_excel('./8.info.xlsx', index=False)

input_
```
