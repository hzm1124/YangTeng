```python
import pandas as pd
```

```python
df_menu = pd.read_excel('./item.xlsx')
list_product = df_menu['item'].to_list()

print(len(list_product))
list_product
```

```python
df = pd.DataFrame(columns=['Item',
                           '链接',
                           '图片链接',
                           'status'])
```

```python
import requests
import json
```

```python
i = 0

resp = requests.get('https://ecat.spectrapremium.com/en/parts/'+str(list_product[i])).text

resp
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(resp, 'lxml')

print(soup.prettify())
```

```python
with open('./product.html', 'w', encoding='utf-8') as file:
    file.write(str(soup))
```

```python
from lxml import etree
```

```python
html = etree.HTML(str(soup))

pic = 'https://ecat.spectrapremium.com' + html.xpath('//a[@id="magic-zoom-plus"]')[0].get('href')

pic
```

```python
df_temp = pd.DataFrame({'Item': list_product[i],
                        '链接': 'https://ecat.spectrapremium.com/en/parts/' + list_product[i],
                        '图片链接': pic,
                        'status': ['ok']})
```

```python
list_specification = html.xpath('//div[@class="part__nom-spec-valeur-spec"]')
for specification in list_specification:
    if specification.xpath('./dt[@class="part__nom-spec"]')[0].text == 'Thickness':
        df_temp['Thickness'] = specification.xpath('./dd[@class="part__valeur-spec"]')[0].text
        
df_temp
```

```python
df = pd.concat([df, df_temp], ignore_index=True).fillna('')

df
```
