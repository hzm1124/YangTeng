```python
param_category = 'thermostat,coolant-housing'
param_subcategory = 'thermostat-assembly-failsafe,thermostat-assembly,power-sport-thermostat,fail-safe-thermostat-w-housing,standard-failsafe,egr-thermostat,hd-thermostat-w-housing,standard-naked-thermostat,hd-thermostat,ultrastat,thermostat-kit-failsafe,thermostat-w-housing,thermostat,thermostat-kit,water-pump-thermostat-assembly,high-flow-thermostat,thermostat-kit-ultrastat,economy-thermostat,water-outlet-coolant-housing'
param_len = [str(_+1) for _ in range(369)]

print('总数量：' + str(len(param_len)))
print()

from gevent import monkey
monkey.patch_all(thread=False)
from gevent.queue import Queue
import gevent

import sys
sys.path.append('../../00.module')
import UA
import Proxy

# = = = = = = = = = = = = = = =

import pandas as pd

# = = = = = = = = = = = = = = =

df = pd.DataFrame()

# = = = = = = = = = = = = = = =

import requests
import json

# = = = = = = = = = = = = = = =

import time
from datetime import datetime

# = = = = = = = = = = = = = = =

work = Queue()
for len_ in param_len:
    work.put_nowait(len_)

def crawler():
    global df
    
    while not work.empty():
        len_ = work.get_nowait()
        
        # = = = = = = = = = = = = = = =
        
        status = 'error'
        for _ in range(31):
            try:
                data = {'action': 'get_paginated_parts',
                        'categories': param_category,
                        'subcategories': param_subcategory,
                        'page': len_}

                resp = requests.post('https://motorad.com/wp-admin/admin-ajax.php',
                                     data=data,
                                     headers=UA.get_User_Agent_Requests()).text

                resp = json.loads(resp)
                
                # = = = = = = = = = = = = = = =
                
                list_part_number = [_['part_number'].strip() for _ in resp['parts']]
                
                # = = = = = = = = = = = = = = =
                
                df_temp = pd.DataFrame({'status': 'ok',
                                        'Page': int(len_),
                                        'Part_Number': list_part_number})
                
                status = 'ok'
                
                break
                
            except:
                time.sleep(1)
                continue

        if status == 'error':
            df_temp = pd.DataFrame([{'status': 'error',
                                     'Page': int(len_)}])

        df = pd.concat([df, df_temp], ignore_index=True).fillna('')

        print(len_ +  '  <->  [' + status + '] - 剩余数量：' + str(work.qsize()) + ' - 当前时间：' + datetime.now().strftime('%H:%M:%S'))

# = = = = = = = = = = = = = = =

list_task = []
for _ in range(20):
    task = gevent.spawn(crawler)
    list_task.append(task)
gevent.joinall(list_task)

df = df.drop_duplicates(ignore_index=True)

print()
print('爬虫输出中')
df.to_excel('./menu_' + datetime.now().strftime('%Y%m%d_%H%M%S') + '.xlsx', index=False)
print()
print('爬虫结束')
```
