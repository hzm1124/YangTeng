```python
import pandas as pd

output_correct = pd.DataFrame()
output_error = pd.DataFrame()
```

```python
input_ = pd.read_excel('./menu.xlsx',
                       header=0,
                       dtype=str).fillna('')

print('总数量：' + str(len(input_)))
input_['Part_Number'].tolist()
```

```python
crawler_status = 'error'

crawler_status
```

```python
import requests

import sys
sys.path.append('../00.Tools')
import UA
import Proxy
```

```python
a = 0

url = 'https://www.rockauto.com/en/parts/' + input_.loc[a, 'Manufacturer'].replace('+', '%2B').replace(',', '%2C') + ',' + input_.loc[a, 'Part_Number'] + ',Lennon,' + input_.loc[a, 'Type_Code']

resp = requests.get(url,
                    headers=UA.get_User_Agent_Requests(),
                    proxies=Proxy.get_Proxy_Requests()).text

resp
```

```python
from bs4 import BeautifulSoup
from lxml import etree
```

```python
soup = BeautifulSoup(resp, 'lxml')
html = etree.HTML(str(soup))

with open('./part.html', 'w', encoding='UTF-8') as file:
    file.write(str(soup))

print(soup.prettify())
```

```python
manufacturer = html.xpath('//select[@id="manufacturer_partsearch_007"]/option[@selected]/text()')[0].strip()

manufacturer
```

```python
category = html.xpath('//select[@id="partgroup_partsearch_007"]/option[@selected]/text()')[0].strip()

category
```

```python
type_ = html.xpath('//select[@id="parttype_partsearch_007"]/option[@selected]/text()')[0].strip()

type_
```

```python
no_parts_found = html.xpath('//span[@class="navlabellink nvoffset nnormal"]/text()')

no_parts_found
```

```python
manufacturer != input_.loc[a, 'Manufacturer'] or category != input_.loc[a, 'Category'] or type_ != input_.loc[a, 'Type'] or (len(no_parts_found) != 0 and no_parts_found[0].strip() == 'No Parts Found')
```

```python
list_part = html.xpath('//tbody[contains(@class, "listing-inner") and (contains(@class, "altrow-a-1") or contains(@class, "altrow-a-0"))]')

list_part
```

```python
len(list_part)
```

```python
import json
```

```python
list_json = [json.loads(part.xpath('./descendant::input[contains(@id, "listing_data_supplemental")]/@value')[0]) for part in list_part]

list_json
```

```python
list_json_2 = [json.loads(part.xpath('./descendant::input[contains(@id, "listing_data_essential")]/@value')[0]) for part in list_part]

list_json_2
```

```python
index_part = -1
for i in range(len(list_json)):
    if list_json[i]['catalogname'].strip() == input_.loc[a, 'Manufacturer'] and list_json[i]['partnumber'].strip() == input_.loc[a, 'Part_Number'] and list_json_2[i]['parttype'].strip() == input_.loc[a, 'Type_Code']:
        if index_part != -1:
            continue
        index_part = i
        
index_part
```

```python
list_note_1 = list_part[index_part].xpath('./descendant::div[contains(@class, "listing-text-row-moreinfo-truck")]/span[contains(@class, "span-link-underline-remover")]/text()')
note_1 = '' if len(list_note_1) == 0 else ';'.join([note_1.strip() for note_1 in list_note_1])

note_1
```

```python
list_note_2 = list_part[index_part].xpath('./descendant::span[contains(@title, "Please make sure")]/span/text()')
note_2 = '' if len(list_note_2) == 0 else ';'.join([note_2.strip() for note_2 in list_note_2])

note_2
```

```python
src = list_part[index_part].xpath('./descendant::img[contains(@id, "inlineimg_thumb")]')
src = '' if len(src) == 0 else 'https://www.rockauto.com' + src[0].xpath('./@src')[0].strip().replace('__ra_m.', '.')

src
```

```python
oe = list_part[index_part].xpath('./descendant::span[contains(@title, "Replaces these Alternate")]')
oe = '' if len(oe) == 0 else oe[0].xpath('./text()')[0].strip().replace(', ', ';')
oe = ';'.join(sorted(list(set(oe.split(';'))), key=oe.split(';').index))

oe
```

```python
part_key = list_json_2[index_part]['partkey'].strip()

part_key
```

```python
df_temp = pd.DataFrame([{'序号': int(input_.loc[a, 'No.']),
                         'Manufacturer': input_.loc[a, 'Manufacturer'],
                         'Category': input_.loc[a, 'Category'],
                         'Type': input_.loc[a, 'Type'],
                         'Type_Code': input_.loc[a, 'Type_Code'],
                         'Part_Number': input_.loc[a, 'Part_Number'],
                         'Vehcile_Json': '',
                         'Vehcile': '',
                         'OE': oe,
                         'Note_1': note_1,
                         'Note_2': note_2,
                         'Pic': '',
                         'Url': url,
                         'Info': 'https://www.rockauto.com/en/moreinfo.php?pt=' + input_.loc[a, 'Type_Code'] + '&pk=' + part_key,
                         'Src': src,
                         'Part_Key': part_key,
                         'JOIN_TCPK': input_.loc[a, 'Type_Code'] + ';' + part_key}])

output_correct = pd.concat([output_correct, df_temp], ignore_index=True).fillna('')

len(output_correct)
```

```python
crawler_status = 'ok'

crawler_status
```

```python
if crawler_status == 'error':
    df_temp = pd.DataFrame([{'No.': int(input_.loc[a, 'No.']),
                             'Manufacturer': input_.loc[a, 'Manufacturer'],
                             'Category': input_.loc[a, 'Category'],
                             'Type': input_.loc[a, 'Type'],
                             'Type_Code': input_.loc[a, 'Type_Code'],
                             'Part_Number': input_.loc[a, 'Part_Number']}])
    
    output_error = pd.concat([output_error, df_temp], ignore_index=True).fillna('')

len(output_error)
```

```python
output_correct = output_correct.drop_duplicates(ignore_index=True)
output_correct = output_correct.sort_values(by=['序号'], ascending=[True]).reset_index(drop=True)
output_correct.to_excel('./test_part.xlsx', index=False)

output_correct
```

```python
if len(output_error) != 0:
    output_error = output_error.drop_duplicates(ignore_index=True)
    output_error = output_error.sort_values(by=['No.'], ascending=[True]).reset_index(drop=True)
    output_error.to_excel('./test_part_error.xlsx', index=False)

output_error
```
