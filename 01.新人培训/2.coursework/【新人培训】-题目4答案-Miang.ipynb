```python
import pandas as pd

input_ = pd.read_excel('./answer_1.xlsx',
                       header=0,
                       dtype=str).fillna('')

input_['US_MRP_SKU'] = ['US' + sku for sku in input_['SKU'].tolist()]

input_
```

```python
df_base = pd.read_excel('./2.WEEK2—case题目0428.xlsx',
                        sheet_name='附1_US基础数据',
                        header=0,
                        dtype=str).fillna('')

for i in range(12):
    df_base[df_base.columns[i+1]] = df_base[df_base.columns[i+1]].astype(int)
df_base['周销'] = df_base['周销'].astype(float)

df_base
```

```python
df_base['销量'] = df_base.iloc[:, 1:13].sum(axis=1)

df_base
```

```python
input_ = pd.merge(left=input_,
                  right=df_base[['MRP SKU', '销量']],
                  how='left',
                  left_on='US_MRP_SKU',
                  right_on='MRP SKU').fillna('')

input_
```

```python
output = input_[['品类', '销量']].groupby(['品类'], as_index=False).sum()
output = output.sort_values(by=['销量'],
                            ascending=[False],
                            ignore_index=True)

output
```

```python
end = 10
while output.loc[end, '销量'] == output.loc[end-1, '销量']:
    end += 1

output = output[:end].copy()

output
```

```python
output['No'] = [_+1 for _ in range(len(output))]

output = output[['No', '品类', '销量']]

output
```

```python
from tqdm import tqdm

with pd.ExcelWriter('./answer_4.xlsx') as writer:
    output.to_excel(writer, sheet_name='Summary', index=False)
    
    for i in tqdm(range(len(output)), desc='Progress', ncols=77):
        df_temp = input_[input_['品类'] == output.loc[i, '品类']].reset_index(drop=True)
    
        # = = = = = = = = = = = = = = =
    
        df_temp = df_temp.sort_values(by=['销量'],
                                      ascending=[False],
                                      ignore_index=True)
    
        # = = = = = = = = = = = = = = =
    
        end = 10
        while df_temp.loc[end, '销量'] == df_temp.loc[end-1, '销量']:
            end += 1
        
        df_temp = df_temp[:end].copy()
    
        # = = = = = = = = = = = = = = =
    
        df_temp['No'] = [_+1 for _ in range(len(df_temp))]
    
        df_temp = df_temp[['品类', 'No', 'SKU', '销量']]
    
        # = = = = = = = = = = = = = = =

        df_temp.to_excel(writer, sheet_name=f'''品类{i+1}.{output.loc[i, '品类']}''', index=False)

print('Done ~')
```
