```python
import pandas as pd

output_correct, output_error = pd.DataFrame(), pd.DataFrame()
```

```python
input_ = pd.read_excel('./1.page.xlsx',
                       header=0,
                       dtype=str).fillna('')

input_['Page'] = input_['Page'].astype(int)

print(f'总数量：{len(input_)}')
input_['Page Url'].tolist()
```

```python
a = 0

a
```

```python
crawler_status = 'error'

crawler_status
```

```python
import requests
import json

import sys
sys.path.append('../00.Tools')
from crawler_configuration import get_header, get_proxy

b = 0
while True:

    while True:
        b += 1
        print(b)
        try:
            resp = requests.get(input_.loc[a, 'Page Url'],
                                headers=get_header(),
                                proxies=get_proxy(),
                                timeout=(10, 10))
            break
        except:
            continue

    if resp.status_code == 200:
        break

resp
```

```python
from bs4 import BeautifulSoup
from lxml import etree

soup = BeautifulSoup(resp.text, 'lxml')
html = etree.HTML(str(soup))

with open('./menu.html', 'w', encoding='UTF-8') as file:
    file.write(str(soup))

print(soup.prettify())
```

```python
list_url = html.xpath('//h4[@class="card-title NoMargin"]/a/@href')

list_url
```

```python
len(list_url) == 0
```

```python
df_temp = pd.DataFrame({'Page': input_.loc[a, 'Page'],
                        'Page Url': input_.loc[a, 'Page Url'],
                        'No': [i+1 for i in range(len(list_url))],
                        'Url': list_url})

output_correct = pd.concat([output_correct, df_temp], ignore_index=True).fillna('')

output_correct
```

```python
crawler_status = 'ok'

crawler_status
```

```python
if crawler_status == 'error':
    df_temp = pd.DataFrame([{'Page': input_.loc[a, 'Page'],
                             'Page Url': input_.loc[a, 'Page Url']}])

    output_error = pd.concat([output_error, df_temp], ignore_index=True).fillna('')

output_error
```

```python
if not output_correct.empty:
    output_correct = output_correct.sort_values(by=['Page', 'No'],
                                                ascending=[True, True],
                                                ignore_index=True)
    output_correct = output_correct.drop_duplicates(['Url'],
                                                    keep='first',
                                                    ignore_index=True)
    output_correct['No'] = [i+1 for i in range(len(output_correct))]
    output_correct.to_excel('./test_url.xlsx', index=False)

output_correct
```

```python
if not output_error.empty:
    output_error = output_error.sort_values(by=['Page'],
                                            ascending=[True, True],
                                            ignore_index=True)
    output_error.to_excel('./test_url_error.xlsx', index=False)

output_error
```
