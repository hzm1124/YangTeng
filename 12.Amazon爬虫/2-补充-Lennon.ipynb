```python
import os
```

```python
directory = list(os.walk('./'))[0][2]

list_file = []
for file in directory:
    if file.startswith('crawler-'):
        list_file.append(file)

print('list_file：' + str(list_file))
```

```python
import pandas as pd
```

```python
list_df = []
for file in list_file:
    list_df.append(pd.read_excel(file, header=0))
df = pd.concat(list_df, ignore_index=True).fillna('')
df.to_excel('./content_temp.xlsx', index=False)
df = df.drop_duplicates(ignore_index=True)
df.to_excel('./content.xlsx', index=False)

df
```
