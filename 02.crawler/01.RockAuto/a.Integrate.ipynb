```python
import os
```

```python
list_dir = list(os.walk('./01.Part_Catalog/data_20240206'))[0][1]
if '.DS_Store' in list_dir:
    list_dir.remove('.DS_Store')
if '0.null.txt' in list_dir:
    list_dir.remove('0.null.txt')
    
list_file_1 = ['./01.Part_Catalog/data_20240206/' + dir_ + '/7.part_number.xlsx' for dir_ in list_dir]
    
list_file_1.sort()
    
print(len(list_file_1))
list_file_1
```

```python
list_dir = list(os.walk('./02.Part_Number_Search/data_20240207'))[0][1]
if '.DS_Store' in list_dir:
    list_dir.remove('.DS_Store')
if '0.null.txt' in list_dir:
    list_dir.remove('0.null.txt')
    
list_dir = ['./02.Part_Number_Search/data_20240207/' + dir_ + '/4.part_number' for dir_ in list_dir]
    
list_file_2 = []
for dir_ in list_dir:
    list_file = list(os.walk(dir_))[0][2]
    list_file_2 += [dir_ + '/' + file for file in list_file]
    
list_file_2.sort()
    
print(len(list_file_2))
list_file_2
```

```python
list_file = list_file_1 + list_file_2 + ['./03.Tools_Universal_Parts/4.part_number.xlsx']

len(list_file)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame()

df
```

```python
for file in list_file:
    df_temp = pd.read_excel(file,
                            header=0,
                            dtype=str).fillna('')
    df_temp = df_temp[['JOIN_MPN', 'Manufacturer', 'Part_Number']]
    df = pd.concat([df, df_temp], ignore_index=True).fillna('')
    df = df.drop_duplicates(ignore_index=True)
    
df.to_excel('./a.integrate.xlsx', index=False)
```
