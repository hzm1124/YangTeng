```python
import pandas as pd
```

```python
df_total = pd.read_excel('./2.WEEK2—case题目0428.xlsx',
                         sheet_name='总表',
                         header=0,
                         usecols='A:B',
                         dtype=str).fillna('')

df_total
```

```python
df_stock = pd.read_excel('./2.WEEK2—case题目0428.xlsx',
                         sheet_name='附3_库存表',
                         header=0,
                         usecols='A:I',
                         dtype=str).fillna('')

df_stock['在途'] = df_stock['在途'].astype(int)
df_stock['可售'] = df_stock['可售'].astype(int)

df_stock
```

```python
df_stock['PTSKU'] = [sku[2:-2] if sku.endswith('-N') else sku[2:-2] if sku.endswith('-Y') else sku[2:] for sku in df_stock['商品编码 4-25'].tolist()]

df_stock
```

```python
df_stock = df_stock[['PTSKU', '可售', '在途']]

df_stock
```

```python
df_stock = df_stock.groupby([df_stock['PTSKU']], as_index=False).sum()

df_stock
```

```python
df_total = pd.merge(left=df_total,
                    right=df_stock,
                    how='left',
                    left_on='SKU',
                    right_on='PTSKU').fillna('')

df_total
```

```python
df_total = df_total[['SKU', '可售', '在途']]

df_total.columns = ['SKU', '在库', '在途']

df_total.to_excel('./answer_3.xlsx', index=False)

df_total
```
