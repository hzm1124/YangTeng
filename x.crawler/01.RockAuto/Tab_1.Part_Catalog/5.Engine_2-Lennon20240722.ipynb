```python
param_date = '20240715'
```

```python
import os

os.makedirs(f'./{param_date}/5.engine_2')
os.makedirs(f'./{param_date}/6.group_1')
os.makedirs(f'./{param_date}/6.group_1/0.error')
```

```python
import pandas as pd

input_ = pd.read_excel(f'./{param_date}/4.engine.xlsx',
                       header=0,
                       dtype=str).fillna('')

input_['Make No'] = input_['Make No'].astype(int)
input_['Year'] = input_['Year'].astype(int)
input_['Model No'] = input_['Model No'].astype(int)
input_['Engine No'] = input_['Engine No'].astype(int)

print(f'总数量：{len(input_)}')
input_['Engine'].tolist()
```

```python
input_2 = input_[['Make']].copy()
input_2 = input_2.drop_duplicates(ignore_index=True)
input_2 = input_2.sort_values(by=['Make'],
                              ascending=[True],
                              ignore_index=True)
input_2['Make No'] = [i+1 for i in range(len(input_2))]
input_2['File Make'] = [group.replace('/', '%2F').replace('\\', '%5C').strip() for group in input_2['Make'].tolist()]

input_2
```

```python
from tqdm import tqdm

for i in tqdm(range(len(input_2)), desc='Progress', ncols=77):
    output = input_[input_['Make'] == input_2.loc[i, 'Make']].reset_index(drop=True)
    output['Make No'] = input_2.loc[i, 'Make No']
    output = output.sort_values(by=['Model', 'Engine', 'Year'],
                                ascending=[True, True, False],
                                ignore_index=True)

    # = = = = = = = = = = = = = = =

    output.loc[0, 'Model No'], output.loc[0, 'Engine No'] = 1, 1
    for j in range(1, len(output)):
        if output.loc[j, 'Model'] == output.loc[j-1, 'Model']:
            output.loc[j, 'Model No'] = output.loc[j-1, 'Model No']
        else:
            output.loc[j, 'Model No'] = output.loc[j-1, 'Model No'] + 1

        if output.loc[j, 'Model No'] == output.loc[j-1, 'Model No']:
            if output.loc[j, 'Engine'] == output.loc[j-1, 'Engine']:
                output.loc[j, 'Engine No'] = output.loc[j-1, 'Engine No']
            else:
                output.loc[j, 'Engine No'] = output.loc[j-1, 'Engine No'] + 1
        else:
            output.loc[j, 'Engine No'] = 1

    # = = = = = = = = = = = = = = =

    output = output[['Make No', 'Make', 'Model No', 'Model', 'Engine No', 'Engine', 'Car Code', 'Year']]
    output.to_excel(f'''./{param_date}/5.engine_2/make_{input_2.loc[i, 'Make No']}.{input_2.loc[i, 'File Make']}.xlsx''', index=False)

print('Done ~')
```
