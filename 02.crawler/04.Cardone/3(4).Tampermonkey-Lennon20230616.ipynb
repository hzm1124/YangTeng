```python
import os
```

```python
list_file = list(os.walk('./oe'))[0][2]
if '.DS_Store' in list_file:
   list_file.remove('.DS_Store')

len(list_file)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns=['Part_Number', 'OE_Numbers'])

df
```

```python
import json
```

```python
for file in list_file:
    print(file)
    with open('./oe/' + file) as f:
        part = json.loads(f.read()[1:-1])
                
        df_temp = pd.DataFrame([{'Part_Number': part['Part_Number'],
                                 'OE_Numbers': part['OE_Numbers']}])
        df = pd.concat([df, df_temp], ignore_index=True)
```

```python
df.to_excel('./oe.xlsx', index=False)

df
```
