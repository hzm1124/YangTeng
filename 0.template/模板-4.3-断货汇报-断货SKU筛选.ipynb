# 0. 配置参数

- `calculate`：what the fuck

```python
calculate = '../../file/计算表-【每日断货汇报】US-Ebay 20230207.xlsx'

out = '../../file/Lennon_断货汇报_筛选好的断货SKU_NOW().xlsx'
```

```python
import pandas as pd
```

```python
df = pd.read_excel(calculate, sheet_name='附_1统计源（含新品、清仓、新品成长）', header=3, usecols=[i for i in range(1, 13)])

df
```

```python
df = df[(df['在库']==0) & (df['在途']==0) & (df['Level']!='普通')][['MRP SKU', 'Level', '目标\n周销', '在途']]

df
```

```python
import datetime

writer = pd.ExcelWriter(out.replace('NOW()', datetime.datetime.now().strftime("%Y%m%d_%H%M%S")))
df.to_excel(writer, index=False)
writer.close()
```
