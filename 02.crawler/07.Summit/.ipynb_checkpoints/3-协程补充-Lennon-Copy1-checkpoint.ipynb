```python
import pandas as pd
```

```python
df_menu = pd.read_excel('./menu-LS_Oil_Pan.xlsx', header=0)

df_menu
```

```python
df_crawler = pd.read_excel('./crawler-LS_Oil_Pan.xlsx', header=0)

df_crawler
```
