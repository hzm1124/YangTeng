```python
import pandas as pd

df_menu = pd.read_excel('./part_out.xlsx', header=0)
list_info = df_menu['Info'].to_list()

print(len(list_info))
list_info
```

```python
df = pd.DataFrame(columns=['Info',
                           'Src',
                           'status'])
```

```python
from urllib.parse import quote
```

```python
i = 0

url = list_info[i]

url
```

```python
import requests
```

```python
resp = requests.get(url).text

resp
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(resp, 'lxml')

print(soup.prettify())
```

```python
with open('./oe.html', 'w', encoding='utf-8') as file:
    file.write(str(soup))
```

```python
from lxml import etree
```

```python
html = etree.HTML(str(soup))

list_src = html.xpath('//img[@alt="Part image"]/@src')
if len(list_src) == 0:
    list_src = ['']

list_src
```

```python
df_temp = pd.DataFrame({'Info': list_info[i],
                        'Src': list_src,
                        'status': ['ok']})

df = pd.concat([df, df_temp], ignore_index=True).fillna('')

df
```
