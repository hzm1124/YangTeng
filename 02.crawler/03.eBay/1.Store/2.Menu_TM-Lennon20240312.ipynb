```python
import pandas as pd
```

```python
df_category = pd.read_excel('./category.xlsx',
                            header=0,
                            dtype=str).fillna('')

df_category
```

```python
from tqdm import tqdm
import os
import json
```

```python
for i in tqdm(range(len(df_category)), desc='Progress', ncols=77):
    list_file = list(os.walk('./category/' + df_category.loc[i, 'Category_No.']))[0][2]
    if '.DS_Store' in list_file:
        list_file.remove('.DS_Store')
    if '0.null.txt' in list_file:
        list_file.remove('0.null.txt')

    list_file.sort()
    
    df_menu = pd.DataFrame()
    for file in list_file:
        with open('./category/' + df_category.loc[i, 'Category_No.'] + '/' + file, 'r', encoding='UTF-8') as f:
            df_temp = pd.DataFrame(json.loads(f.read()))
            df_temp['Category_No.'] = int(df_category.loc[i, 'Category_No.'])
            df_temp['Category'] = df_category.loc[i, 'Category']
            df_menu = pd.concat([df_menu, df_temp], ignore_index=False).fillna('')
            
    df_menu = df_menu.drop_duplicates(ignore_index=True)
    df_menu = df_menu.sort_values(by=['Page', 'No.'], ascending=[True, True]).reset_index(drop=True)
    df_menu = df_menu.drop_duplicates(['Item_Number'], keep='first').reset_index(drop=True)
    df_menu['No.'] = [i+1 for i in range(len(df_menu))]
    
    df_menu.to_excel('./menu/' + str(i+1) + '.' + df_category.loc[i, 'Category'].replace('/', '_') + '.xlsx', index=False)
            
df_menu
```
