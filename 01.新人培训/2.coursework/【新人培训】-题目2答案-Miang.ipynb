```python
import pandas as pd

input_ = pd.read_excel('./2.WEEK2—case题目0428.xlsx',
                       sheet_name='总表',
                       header=0,
                       usecols='A:B',
                       dtype=str).fillna('')

input_['US_MRP_SKU'] = ['US' + sku for sku in input_['SKU'].tolist()]

input_
```

```python
df_base = pd.read_excel('./2.WEEK2—case题目0428.xlsx',
                        sheet_name='附1_US基础数据',
                        header=0,
                        dtype=str).fillna('')

for i in range(12):
    df_base[df_base.columns[i+1]] = df_base[df_base.columns[i+1]].astype(int)
df_base['周销'] = df_base['周销'].astype(float)

df_base
```

```python
input_ = pd.merge(left=input_,
                  right=df_base,
                  how='left',
                  left_on='US_MRP_SKU',
                  right_on='MRP SKU').fillna('')

input_
```

```python
input_['层级'] = ['TOP1' if sale >= 10 else 'TOP10' if sale >= 7 else 'TOP20' if sale >= 3 else '普通' for sale in input_['周销'].tolist()]

input_[['SKU', '仓库', '周销', '层级']].to_excel('./answer_2.xlsx', index=False)

input_
```
