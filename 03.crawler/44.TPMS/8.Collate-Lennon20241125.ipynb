```python
import pandas as pd

from tqdm import tqdm
import json
```

```python
df_input = pd.read_excel('./file/7.part.xlsx',
                         header=0,
                         dtype=str).fillna('')

df_input['No'] = df_input['No'].astype(int)

df_input
```

```python
list_series, list_column, list_column_price, list_column_info, list_column_description = [], df_input.columns.tolist(), [], [], []
for _, series in tqdm(df_input.iterrows(), total=len(df_input), desc='Progress', ncols=77):
    dict_price, dict_dict, dict_dict_2 = json.loads(series['Json_Price']), json.loads(series['Json_Info']), json.loads(series['Json_Description'])
    dict_ = {k: v for dict_ in dict_dict.values() for k, v in dict_.items()}
    dict_2 = {k: v for dict_ in dict_dict_2.values() for k, v in dict_.items()}

    # = = = = = = = = = = = = = = =

    for key, value in dict_price.items():
        key = f'Price-{int(key) + 1}'

        if key not in list_column_price:
            list_column_price.append(key)

        series[key] = value

    # = = = = = = = = = = = = = = =

    for key, value in dict_.items():
        if key in list_column:
            key = f'{key} (Info)'

        if key not in list_column_info:
            list_column_info.append(key)

        series[key] = value

    # = = = = = = = = = = = = = = =

    for key, value in dict_2.items():
        key = f'[Description] {key}'

        if key not in list_column_description:
            list_column_description.append(key)

        series[key] = value

    list_series.append(series)

# = = = = = = = = = = = = = = =

df_output = pd.DataFrame(list_series).sort_values(by=['No'],
                                                  ascending=[True],
                                                  ignore_index=True)
df_output[list_column[:5] + sorted(list_column_info) + sorted(list_column_price) + list_column[8:9] + sorted(list_column_description) + list_column[5:8] + list_column[9:]].to_excel('./file/8.part.xlsx', index=False)

print()
print('Done ~')
```
