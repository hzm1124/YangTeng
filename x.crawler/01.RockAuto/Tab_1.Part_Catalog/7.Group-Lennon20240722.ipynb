```python
import os

os.makedirs(f'./{param_date}/7.group')
os.makedirs(f'./{param_date}/8.type_1')
os.makedirs(f'./{param_date}/8.type_1/0.error')
```

```python
list_file = sorted(list(os.walk(f'./{param_date}/6.group_1'))[0][2])
while '.DS_Store' in list_file:
    list_file.remove('.DS_Store')
while '0.null.txt' in list_file:
    list_file.remove('0.null.txt')

list_file
```

```python
from tqdm import tqdm
import re

for file in tqdm(list_file, desc='Progress', ncols=77):
    os.rename(f'./{param_date}/6.group_1/{file}',
              f'''./{param_date}/6.group_1/{re.sub(r'-[0-9]{8}_[0-9]{6}.xlsx$', '.xlsx', file)}''')

print('Done ~')
```

```python
list_file = sorted(list(os.walk(f'./{param_date}/6.group_1'))[0][2])
while '.DS_Store' in list_file:
    list_file.remove('.DS_Store')
while '0.null.txt' in list_file:
    list_file.remove('0.null.txt')

list_file
```

```python
import pandas as pd

for file in tqdm(list_file, desc='Progress', ncols=77):
    input_ = pd.read_excel(f'./{param_date}/6.group_1/{file}',
                           header=0,
                           dtype=str).fillna('')

    input_['Make No'] = input_['Make No'].astype(int)
    input_['Model No'] = input_['Model No'].astype(int)
    input_['Engine No'] = input_['Engine No'].astype(int)
    input_['Year'] = input_['Year'].astype(int)
    input_['Group No'] = input_['Group No'].astype(int)

    list_output, index_start, index_end = [], 0, 10_000
    while True:
        if index_end < len(input_):
            while input_.loc[index_end, 'Model No'] == input_.loc[index_end-1, 'Model No']:
                index_end += 1
                if index_end == len(input_):
                    break
            if index_end != len(input_):
                list_output.append(input_[index_start:index_end].reset_index(drop=True))
                index_start, index_end = index_end, index_end + 10_000
        else:
            list_output.append(input_[index_start:].reset_index(drop=True))
            break

    if len(list_output) == 1:
        input_.to_excel(f'''./{param_date}/7.group/{file}''', index=False)
    else:
        for i, output in zip(range(len(list_output)), list_output):
            output.to_excel(f'''./{param_date}/7.group/{file.removesuffix('.xlsx')}-serial_{i+1}.xlsx''', index=False)

print('Done ~')
```
