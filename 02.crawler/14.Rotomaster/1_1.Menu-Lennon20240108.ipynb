```python
param_search_query = 'Turbocharger'
param_len = 12

list_url = ['https://rotomaster.com/search.php?section=product&_bc_fsnf=1&Part+Type=Turbocharger&limit=30&sort=alphaasc&search_query=' + param_search_query + '&page=' + str(_+1) for _ in range(param_len)]

print('总数量：' + str(len(list_url)))
list_url
```

```python
import pandas as pd
```

```python
df = pd.DataFrame()

df
```

```python
import requests
```

```python
i = 0

url = list_url[i]

resp = requests.get(url).text

resp
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(resp, 'lxml')

with open('./menu.html', 'w', encoding='UTF-8') as file:
    file.write(str(soup))

print(soup.prettify())
```

```python
from lxml import etree
```

```python
html = etree.HTML(str(soup))

list_url = html.xpath('//h4[@class="card-title"]/a/@href')

list_url
```

```python
df_temp = pd.DataFrame({'status': 'ok',
                        'Menu': url,
                        'Url': list_url})

df = pd.concat([df, df_temp], ignore_index=True).fillna('')

df
```
