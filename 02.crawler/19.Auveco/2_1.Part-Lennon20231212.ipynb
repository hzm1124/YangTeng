```python
import pandas as pd
```

```python
df = pd.read_excel('./menu.xlsx',
                   header=0,
                   dtype=str).fillna('')

list_url = df['Url'].tolist()

print('总数量：' + str(len(list_url)))
list_url
```

```python
import requests
```

```python
i = 1

url = list_url[i]

resp = requests.get(url).text

resp
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(resp, 'lxml')

with open('./part.html', 'w', encoding='UTF-8') as file:
    file.write(str(soup))

print(soup.prettify())
```

```python
from lxml import etree
```

```python
html = etree.HTML(str(soup))

title = html.xpath('//div[@class="span8"]/div[@class="row-fluid"]/h1/text()')[0].strip()

title
```

```python
item_number = html.xpath('//div[@class="span8"]/div[@class="row-fluid"]/p[@class="product-details-code"]/text()')[0].replace('Item No:', '').strip()

item_number
```

```python
upc = html.xpath('//div[@class="span8"]/div[@class="row-fluid"]/p[@class="product-details-upccode"]/text()')
if len(upc) == 0:
    upc = ''
else:
    upc = upc[0].replace('UPC Code:', '').strip()
    
upc
```

```python
src = html.xpath('//img[@id="product-detail-gallery-main-img"]/@src')[0].strip()

src
```

```python
kit = html.xpath('//div[@class="product-details-desc"]/descendant::table')
if len(kit) == 0:
    if len(html.xpath('//div[@class="product-details-desc"]/span[2]')) == 0:
        kit = ''
    else:
        kit = html.xpath('//div[@class="product-details-desc"]/span[2]/text()')[1].replace(' &', ',').strip().replace(', ', ';')
else:
    list_index = []
    list_span = kit[0].xpath('./tbody/tr[1]/td/span')
    for i in range(len(list_span)):
        header = ''.join(list_span[i].xpath('./text()'))
        if 'Auveco' in header and 'Part' in header and 'No.' in header:
            list_index.append(i)
    
    list_kit = []
    if len(kit[0].xpath('./tbody/tr[1]/td')) == len(kit[0].xpath('./tbody/tr[2]/td')):
        list_tr = kit[0].xpath('./tbody/tr')[1:]
    else:
        list_tr = kit[0].xpath('./tbody/tr')[2:]
    for tr in list_tr:
        list_td = tr.xpath('./td/span/text()')
        list_td = [_.replace('\xa0', '') for _ in list_td]
        for index in list_index:
            list_kit.append(list_td[index].strip())
    
    while '' in list_kit:
        list_kit.remove('')
        
    kit = ';'.join(list_kit)
        
kit
```

```python
list_li_1 = html.xpath('//li[@class="information-attribute"]')

list_li_1
```

```python
list_li_2 = html.xpath('//div[@class="row-fluid"]/div[@class="span6"]/ul[@class="unstyled"]/li/text()')

list_li_2
```

```python
df_temp = pd.DataFrame([{'status': 'ok',
                         '序号': '',
                         'Title': title,
                         'Item_Number': item_number,
                         'UPC': upc,
                         'Kit': kit,
                         'Pic': '',
                         'Url': url,
                         'Src': src}])

for li in list_li_1:
    df_temp.loc[0, li.xpath('./span[@class="attribute-title"]/text()')[0].replace(':', '').strip().lower()] = li.xpath('./span[@class="attribute-value"]/text()')[0].strip()

for li in list_li_2:
    df_temp.loc[0, li.split(':')[0].strip().upper()] = li.split(':')[1].strip()
    
df_temp
```
