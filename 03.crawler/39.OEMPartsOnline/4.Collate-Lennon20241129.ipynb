```python
import pandas as pd

from tqdm import tqdm
import json
```

# 1. Part

```python
print('Data Loading...')
directory_in = os.path.join(path_root, './file/3.part')
list_df, list_file = [], [file for file in sorted(list(os.walk(directory_in))[0][2]) if file not in ['.DS_Store'] and os.path.isfile(os.path.join(directory_in, file))]
for file in tqdm(list_file, desc='Progress', ncols=77):

    df_input = pd.read_excel(os.path.join(directory_in, file),
                             header=0,
                             dtype=str).fillna('')

    df_input['No'] = df_input['No'].astype(int)
    df_input['year'] = df_input['year'].astype(int)

    # = = =  = = =  = = =

    list_df.append(df_input)
print('Loading Done !')
print()

# = = =  = = =  = = =

print('Data Outputting...')
df_input = pd.concat(list_df, ignore_index=True)
df, df_2 = df_input.iloc[:, :11].drop_duplicates(ignore_index=True), df_input.iloc[:, 10:].sort_values(by=['JOIN_CPN', 'make', 'model', 'trims', 'engines', 'year'],
                                                                                                       ascending=[True, True, True, True, True, False],
                                                                                                       ignore_index=True)
df_2.to_excel(os.path.join(directory_out, './compatibility.xlsx'), index=False)
print('Output Finished !')
print()

# = = =  = = =  = = =

list_series, list_column, list_column_detail = [], df.columns.tolist(), []
for _, series in tqdm(df.iterrows(), total=len(df), desc='Progress', ncols=77):
    dict_dict = json.loads(series['Json_Detail'])
    dict_ = {k: v for dict_ in dict_dict.values() for k, v in dict_.items()}

    for key, value in dict_.items():
        if key in list_column:
            key = f'{key} (Detail)'

        if key not in list_column_detail:
            list_column_detail.append(key)

        series[key] = value

    list_series.append(series)

# = = =  = = =  = = =

print()
print('Data Outputting...')
df = pd.DataFrame(list_series).sort_values(by=['Catalog', 'Part Number'],
                                           ascending=[True, True],
                                           ignore_index=True)
df['No'] = range(1, len(df) + 1)
df[list_column[:8] + sorted(list_column_detail) + list_column[8:]].to_excel(os.path.join(directory_out, './part.xlsx'), index=False)
print('Output Finished !')
print()

df
```

# 2. Vehicle

```python
list_dict, grouped = [], df_2.groupby('JOIN_CPN')
for cpn, df in tqdm(grouped, desc='Progress', ncols=77):
    df = df.reset_index(drop=True)

    # = = =  = = =  = = =

    dict_vehicle = {}
    for i in range(len(df)):
        make, model, engine, year = df.loc[i, 'make'], df.loc[i, 'model'], df.loc[i, 'engines'], df.loc[i, 'year']

        if make not in dict_vehicle:
            dict_vehicle[make] = {model: {engine: [year]}}
        elif model not in dict_vehicle[make]:
            dict_vehicle[make][model] = {engine: [year]}
        elif engine not in dict_vehicle[make][model]:
            dict_vehicle[make][model][engine] = [year]
        elif year not in dict_vehicle[make][model][engine]:
            dict_vehicle[make][model][engine].append(year)

    # = = =  = = =  = = =

    list_vehicle, list_vehicle_2 = [], []
    for make, dict_model in dict_vehicle.items():
        for model, dict_engine in dict_model.items():
            list_year = [year for list_year in dict_engine.values() for year in list_year]
            year_begin, year_end = min(list_year), max(list_year)
            list_vehicle.append(f'{make} {model} {year_begin}' if year_begin == year_end else f'{make} {model} {year_begin}-{year_end}')

            for engine, list_year in dict_engine.items():
                year_begin, year_end = min(list_year), max(list_year)
                list_vehicle_2.append(f'{make} {model} {year_begin} {engine}'.strip() if year_begin == year_end else f'{make} {model} {year_begin}-{year_end} {engine}'.strip())

    # = = =  = = =  = = =

    list_dict.append({'JOIN_CPN': cpn,
                      'Vehicle W/o Engine': '\n'.join(sorted(list_vehicle)),
                      'Vehicle W/ Engine': '\n'.join(sorted(list_vehicle_2))})

# = = = = = = = = = = = = = = =

print()
print('Data Outputting...')
df = pd.DataFrame(list_dict).sort_values(by=['JOIN_CPN'],
                                         ascending=[True],
                                         ignore_index=True)
df.to_excel(os.path.join(directory_out, './vehicle.xlsx'), index=False)
print('Output Finished !')
print()

df
```
