```python
import os
```

```python
list_file = list(os.walk('./crawler'))[0][2]
if '.DS_Store' in list_file:
   list_file.remove('.DS_Store')

len(list_file)
```

```python
import json
import pandas as pd
```

```python
list_df = []
for file in list_file:
    print(file)
    f = open('./crawler/' + file)
    text = json.loads(f.read())
    text['Vehicle'] = '\n'.join(text['Vehicle'])
    list_df.append(pd.DataFrame([text]))
    f.close()
```

```python
df = pd.concat(list_df).fillna('').reset_index(drop=True)

df.to_excel('crawler.xlsx', index=False)

df
```
