```python
import pandas as pd
```

```python
df = pd.read_excel('./model_2.xlsx',
                   header=0,
                   dtype=str).fillna('')

print('总数量：' + str(len(df)))
df['Model_Id'].tolist()
```

```python
from tqdm import tqdm
```

```python
list_url = []
new_df = pd.DataFrame(columns=['No.', 'Product_Id', 'Model'])
for i in tqdm(range(len(df)), desc='Progress', ncols=77):
    list_model_id = df.loc[i, 'Model_Id'].split(';')
    list_model = df.loc[i, 'Vehicle_1'].split('\n')
    for model_id, model in zip(list_model_id, list_model):
        list_url.append('https://www.autoteiledirekt.de/ajax/product/related/vehicles?articleId=' + df.loc[i, 'Product_Id'] + '&makerId=' + df.loc[i, 'Maker_Id'] + '&modelId=' + model_id)
        
        df_temp = pd.DataFrame([{'Product_Id': df.loc[i, 'Product_Id'],
                                 'Model': model}])
        new_df = pd.concat([new_df, df_temp], ignore_index=True).fillna('')
        
new_df['No.'] = [i+1 for i in range(len(new_df))]
new_df.to_excel('./engine_1.xlsx', index=False)

list_url
```

```python
dict_url = {}
for i in range(len(list_url)):
    dict_url[str(i+1)] = list_url[i]
    
dict_url
```
