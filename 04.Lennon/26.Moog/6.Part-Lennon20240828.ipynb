```python
import os

os.makedirs('./file/6.part')
```

```python
import pandas as pd

input_ = pd.read_excel('./file/5.part.xlsx',
                       header=0,
                       dtype=str).fillna('')

input_['No'] = input_['No'].astype(int)

input_
```

```python
input_2 = input_.drop_duplicates(['Type'],
                                 keep='first',
                                 ignore_index=True)
input_2 = input_2.sort_values(by=['Type'],
                              ascending=[True],
                              ignore_index=True)

input_2['File Type'] = [_.strip().replace('/', '%2F').replace('\\', '%5C') for _ in input_2['Type'].tolist()]

input_2
```

```python
from tqdm import tqdm

for i in tqdm(range(len(input_2)), desc='Progress', ncols=77):
    output = input_[input_['Type'] == input_2.loc[i, 'Type']].reset_index(drop=True)
    output = output.sort_values(by=['No'],
                                ascending=[True],
                                ignore_index=True)

    output.to_excel(f'''./file/6.part/type_{i+1}.{input_2.loc[i, 'File Type']}.xlsx''', index=False)

print('Done ~')
```
