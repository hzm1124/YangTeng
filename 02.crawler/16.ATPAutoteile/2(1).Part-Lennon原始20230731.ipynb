```python
i = 0

browser.get(list_url[i])

time.sleep(5)
```

```python
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.common.by import By
```

```python
wait = WebDriverWait(browser, 60)
wait.until(EC.presence_of_element_located((By.XPATH, '//div[@class="oe-number"]')))
list_oe = browser.find_elements(by=By.XPATH, value='//div[@class="oe-number"]')
list_oe = [_.get_attribute('textContent').replace('\n', '').strip() for _ in list_oe]
list_oe = list(set(list_oe))
list_oe.sort()

oe = ';'.join(list_oe).replace(',;', ';').replace(' ', '').replace('.', '').replace('-', '')

oe
```
