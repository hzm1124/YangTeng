```python
import os

list_file = sorted(list(os.walk('./file/2.item'))[0][2])
while '.DS_Store' in list_file:
    list_file.remove('.DS_Store')
while '0.null.txt' in list_file:
    list_file.remove('0.null.txt')

list_file
```

```python
file = list_file[0]

file
```

```python
import pandas as pd

output_correct, output_error = pd.DataFrame(), pd.DataFrame()
```

```python
input_ = pd.read_excel(f'./file/2.item/{file}',
                       header=0,
                       dtype=str).fillna('')

input_['No'] = input_['No'].astype(int)

print(f'总数量：{len(input_)}')
input_['Item Number'].tolist()
```

```python
a = 0

a
```

```python
crawler_status = 'error'

crawler_status
```

```python
import requests

import sys
sys.path.append('../00.Tools')
from crawler_configuration import get_header, get_proxy

b = 0
while True:
    b += 1
    print(b)
    try:
        resp = requests.get(input_.loc[a, 'Description Url'],
                            headers=get_header(ua=False),
                            proxies=get_proxy(),
                            timeout=(10, 10))

        if resp.status_code == 200:
            break
    except:
        continue

resp
```

```python
from bs4 import BeautifulSoup
from lxml import etree

soup = BeautifulSoup(resp.text, 'lxml')
html = etree.HTML(str(soup))

with open('./description.html', 'w', encoding='UTF-8') as file:
    file.write(str(soup))

print(soup.prettify())
```

```python
list_div = html.xpath('//div')

list_div
```

```python
fitment = ''
for div in list_div:
    if div.xpath('./h4/text()') and div.xpath('./h4/text()')[0].strip() == 'FITMENT':
        list_text = [text.strip() for text in div.xpath('./following-sibling::div[1]/li/text()')]

while '' in list_text:
    list_text.remove('')

fitment = '\n'.join(list_text)

print(fitment)
```

```python
df_temp = pd.DataFrame([{'No': input_.loc[a, 'No'],
                         'Description Url': input_.loc[a, 'Description Url'],
                         'Fitment': fitment}])

output_correct = pd.concat([output_correct, df_temp], ignore_index=True).fillna('')

output_correct
```

```python
crawler_status = 'ok'

crawler_status
```

```python
if crawler_status == 'error':
    output_error = pd.concat([output_error, pd.DataFrame([input_.iloc[a]]).reset_index(drop=True)], ignore_index=True).fillna('')

output_error
```

```python
if not output_correct.empty:
    output_correct = output_correct.sort_values(by=['No'],
                                                ascending=[True],
                                                ignore_index=True)
    output_correct.to_excel('./test_description.xlsx', index=False)

output_correct
```

```python
if not output_error.empty:
    output_error = output_error.sort_values(by=['No'],
                                            ascending=[True],
                                            ignore_index=True)
    output_error.to_excel('./test_description_error.xlsx', index=False)

output_error
```
