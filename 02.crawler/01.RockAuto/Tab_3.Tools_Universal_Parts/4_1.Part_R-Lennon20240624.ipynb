```python
import requests

import sys
sys.path.append('../../00.Tools')
import UA
import Proxy
```

```python
a = 11

data = {'func': 'navnode_fetch',
        'payload': f'''{{"jsn":{{"tab":"extras","groupname":"{input_.loc[a, 'GroupName']}","subgroupname":"{input_.loc[a, 'SubGroupName']}","parttype":"{input_.loc[a, 'PartType']}","nodetype":"parttype"}}}}''',
        'api_json_request': '1'}

resp = requests.post('https://www.rockauto.com/catalog/catalogapi.php',
                     data=data,
                     headers=UA.get_User_Agent_Requests(),
                     proxies=Proxy.get_Proxy_Requests()).text

resp
```

```python
import json
```

```python
json_ = json.loads(resp)

json_
```

```python
from bs4 import BeautifulSoup
from lxml import etree
```

```python
soup = BeautifulSoup(json_['html_fill_sections']['navchildren[]'], 'lxml')
html = etree.HTML(str(soup))

with open('./part.html', 'w', encoding='UTF-8') as file:
    file.write(str(soup))

print(soup.prettify())
```

```python
list_tbody = html.xpath('//tbody[contains(@class, "listing-inner") and (contains(@class, "altrow-a-1") or contains(@class, "altrow-a-0"))]')

list_tbody
```

```python
list_list_note = [tbody.xpath('./tr[1]/td/div[@class="listing-text-row-moreinfo-truck"]/span[@class="span-link-underline-remover"]/text()') for tbody in list_tbody]
list_note = ['\n'.join([note.strip() for note in list_note]) for list_note in list_list_note]

list_note
```

```python
import re
```

```python
list_note_2 = [etree.tostring(tbody.xpath('./tr[1]/td/div[@class="listing-text-row"]')[0]).decode('utf-8') for tbody in list_tbody]
list_note_2 = [re.sub(r'<.*?>', '', note.strip()) for note in list_note_2]
list_note_2 = [re.sub(r' +', ' ', note.strip()) for note in list_note_2]

list_note_2
```

```python
list_json = [json.loads(tbody.xpath('./tr[1]/td[contains(@class, "listing-inner-content")]/input[contains(@id, "listing_data_essential[")]/@value')[0]) for tbody in list_tbody]

list_json
```

```python
list_parttype = [json_['parttype'].strip() for json_ in list_json]

list_parttype
```

```python
list_partkey = [json_['partkey'].strip() for json_ in list_json]

list_partkey
```

```python
list_info = []
list_list_info = [tbody.xpath('./tr[1]/td/div[@class="listing-text-row-moreinfo-truck"]/a[@class="ra-btn ra-btn-moreinfo"]/@href') for tbody in list_tbody]
for i in range(len(list_list_info)):
    if len(list_list_info[i]) == 0 or 'www.rockauto.com/en/moreinfo.php' not in list_list_info[i][0]:
        list_info.append(f'https://www.rockauto.com/en/moreinfo.php?pk={list_partkey[i]}&pt={list_parttype[i]}&Lennon=1')
    else:
        list_info.append(list_list_info[i][0])

list_info
```

```python
list_json = [json.loads(tbody.xpath('./tr[1]/td/input[contains(@id, "listing_data_supplemental[")]/@value')[0]) for tbody in list_tbody]

list_json
```

```python
list_partnumber = [json_['partnumber'].strip() for json_ in list_json]

list_partnumber
```

```python
list_catalogname = [json_['catalogname'].strip() for json_ in list_json]

list_catalogname
```

```python
list_sortgrouptext = [json_['sortgrouptext'].strip() for json_ in list_json]

list_sortgrouptext
```

```python
list_paramdesc = [json_['paramdesc'].strip() for json_ in list_json]

list_paramdesc
```

```python
list_list_oem = [tbody.xpath('./tr[1]/td/span[contains(@title, "Replaces these Alternate")]/text()') for tbody in list_tbody]
list_oem = [';'.join([oem_split.strip() for oem_split in oem.split(',')]) for oem in [','.join(list_oem) for list_oem in list_list_oem]]

list_oem
```

```python
list_list_src = [[] if len(list_src) == 0 else json.loads(list_src[0])['Slots'] for list_src in [tbody.xpath('./tr[1]/td/div/input[contains(@id, "jsninlineimg[")]/@value') for tbody in list_tbody]]

list_list_src
```

```python
list_json_src = []
for list_src in list_list_src:
    dict_src = {}
    for i in range(len(list_src)):
        dict_src[str(i)] = 'https://www.rockauto.com' + list_src[i]['ImageData']['Full'].strip()

    list_json_src.append(json.dumps(dict_src))

list_json_src
```

```python
df_temp = pd.DataFrame({'Tab No': int(input_.loc[a, 'Tab No']),
                        'Tab': input_.loc[a, 'Tab'],
                        'GroupName No': int(input_.loc[a, 'GroupName No']),
                        'GroupName': input_.loc[a, 'GroupName'],
                        'SubGroupName No': int(input_.loc[a, 'SubGroupName No']),
                        'SubGroupName': input_.loc[a, 'SubGroupName'],
                        'PartType No': int(input_.loc[a, 'PartType No']),
                        'PartType Name': input_.loc[a, 'PartType Name'],
                        'PartType': input_.loc[a, 'PartType'],
                        'Tab Url': input_.loc[a, 'Tab Url'],
                        'Part No': [i+1 for i in range(len(list_json))],
                        'SortGroupText': list_sortgrouptext,
                        'CatalogName': list_catalogname,
                        'PartNumber': list_partnumber,
                        'Part PartType': list_parttype,
                        'PartKey': list_partkey,
                        'JOIN_PTPKPNCN': [f'{parttype};{partkey};{partnumber};{catalogname}' for parttype, partkey, partnumber, catalogname in zip(list_parttype, list_partkey, list_partnumber, list_catalogname)],
                        'ParamDesc': list_paramdesc,
                        'Json_Vehicle': '',
                        'Vehicle': '',
                        'OEM': list_oem,
                        'Note 1': list_note,
                        'Note 2': list_note_2,
                        'Url': [f'''https://www.rockauto.com/en/parts/{catalogname},{partnumber},{input_.loc[a, 'PartType Name']},{part_parttype}''' for catalogname, partnumber, part_parttype in zip(list_catalogname, list_partnumber, list_parttype)],
                        'Info': list_info,
                        'Json_Src': list_json_src,
                        'Json_Specification': '',
                        'Json_Kit': ''})

output_correct = pd.concat([output_correct, df_temp], ignore_index=True).fillna('')

output_correct
```
