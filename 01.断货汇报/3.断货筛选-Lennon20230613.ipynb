```python
import os
import datetime
```

```python
list_file = list(os.walk('./'))[0][2]
for file in list_file:
    if '计算表' in file:
        calculate = file

print(f'calculate: \t{calculate}')
```

```python
import pandas as pd
```

```python
df = pd.read_excel(calculate,
                   sheet_name='附_1统计源（含新品、清仓、新品成长）',
                   header=3,
                   usecols=[2, 3, 5, 11, 12])

df
```

```python
df = df[(df['在库']==0) & (df['在途']==0) & (df['Level']!='普通')][['MRP SKU',
                                                                   'Level',
                                                                   '目标\n周销',
                                                                   '在途']]

df
```

```python
import datetime

writer = pd.ExcelWriter('Lennon_筛选好的断货SKU_' + datetime.datetime.now().strftime("%Y%m%d_%H%M%S") + '.xlsx')
df.to_excel(writer, index=False)
writer.close()
```
