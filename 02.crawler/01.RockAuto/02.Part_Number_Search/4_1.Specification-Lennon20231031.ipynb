```python
import pandas as pd
```

```python
df_menu = pd.read_excel('./crawler.xlsx',
                        header=0,
                        dtype=str).fillna('')

list_url = df_menu['Info'].tolist()

print('总数量：' + str(len(list_url)))
list_url
```

```python
df = pd.DataFrame()

df
```

```python
from selenium.webdriver import ChromeOptions
```

```python
option = ChromeOptions()
option.add_experimental_option('useAutomationExtension', False)
# option.add_argument('blink-settings=imagesEnabled=false')
option.add_argument('--disable-blink-features=AutomationControlled')
option.add_argument('--disable-dev-shm-usage')
option.add_argument('--disable-gpu')
option.add_experimental_option('excludeSwitches', ['enable-automation'])
# option.add_argument('--headless')
option.add_argument('--no-sandbox')
# option.add_argument('--proxy-server=' + Proxy.get_Proxy_selenium())
# option.add_argument('user-agent=' + UA.get_UA())
option.page_load_strategy = 'eager'
```

```python
from selenium.webdriver import Chrome
from selenium.webdriver.chrome.service import Service
```

```python
browser = Chrome(service=Service(str_chromedriver), options=option)

browser.maximize_window()
```

```python
import time
```

```python
url = list_url[0]

browser.get(url)

time.sleep(3)
```

```python
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.common.by import By
```

```python
wait = WebDriverWait(browser, 60)
wait.until(EC.presence_of_element_located((By.XPATH, '//section[contains(@aria-label, "More Information for")]')))
list_tr = browser.find_elements(by=By.XPATH, value='//table[contains(@class, "moreinfotable")]/tbody/tr')

list_tr
```

```python
df_temp = pd.DataFrame([{'status': 'ok',
                         'Info': url}])

df_temp
```

```python
for _ in list_tr[1:]:
    df_temp.loc[0, _.find_elements(by=By.XPATH, value='./td')[0].text.lower()] = _.find_elements(by=By.XPATH, value='./td')[1].text
  
browser.quit()

df = pd.concat([df, df_temp], ignore_index=True).fillna('')

df
```
