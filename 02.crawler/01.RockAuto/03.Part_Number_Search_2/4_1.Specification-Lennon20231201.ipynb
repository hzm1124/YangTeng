```python
df = pd.read_excel('./part.xlsx',
                   header=0,
                   dtype=str).fillna('')

list_manufacturer = df['Manufacturer'].tolist()
list_part_number = df['Part_Number'].tolist()
list_info = df['Info'].tolist()

print('总数量：' + str(len(list_info)))
list_info
```

```python
import requests
```

```python
import sys
sys.path.append('../../../00.module')
import UA
import Proxy

i = 0

manufacturer = list_manufacturer[i]
part_number = list_part_number[i]
info = list_info[i]

resp = requests.get(info,
                                     headers=UA.get_User_Agent_Requests(),
                                     proxies=Proxy.get_Proxy_Requests()).text

resp
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(resp, 'lxml')

with open('./info.html', 'w', encoding='UTF-8') as file:
    file.write(str(soup))

print(soup.prettify())
```

```python
from lxml import etree
```

```python
html = etree.HTML(str(soup))

html.xpath('//span[contains(@class, "listing-final-manufacturer")]/text()')[0] == manufacturer and html.xpath('//span[contains(@id, "vew_partnumber[1]")]/text()')[0] == part_number
```

```python
list_tr = html.xpath('//table[contains(@class, "moreinfotable")]/tr')

list_tr
```

```python
df_temp = pd.DataFrame([{'status': 'ok',
                         'Info': info}])

for tr in list_tr[1:]:
    df_temp.loc[0, tr.xpath('./td[1]/text()')[0].strip().lower()] = tr.xpath('./td[2]/text()')[0].strip()

df_temp
```
