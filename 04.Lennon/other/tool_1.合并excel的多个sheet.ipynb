```python
import pandas as pd

file = pd.ExcelFile('./fuck.xlsx')

file.sheet_names
```

```python
list_df = []
for sheet_name in file.sheet_names:
    df = pd.read_excel(file, sheet_name=sheet_name)
    list_df.append(df)

df = pd.concat(list_df).reset_index(drop=True)

df.to_excel('./fuck_new.xlsx', index=False)

df
```
