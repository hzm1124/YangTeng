```python
import pandas as pd

input_ = pd.read_excel('./2.part_number_1.xlsx',
                       header=0,
                       dtype=str).fillna('')

input_['Type No'] = input_['Type No'].astype(int)
input_['Type Code'] = input_['Type Code'].astype(int)
input_['SKU Count'] = input_['SKU Count'].astype(int)
input_['No'] = input_['No'].astype(int)

input_
```

```python
input_ = input_.drop_duplicates(['Part Number'],
                                keep='first',
                                ignore_index=True)
input_ = input_.sort_values(by=['Part Number'],
                            ascending=[True],
                            ignore_index=True)

input_['No'] = [i+1 for i in range(len(input_))]
input_['Url'] = [f'''https://www.standardbrand.com/en/ecatalog?part={_}&type=p&search=s''' for _ in input_['Part Number'].tolist()]
input_[['No', 'Part Number', 'Url']].to_excel('./3.part_number.xlsx', index=False)

input_
```
