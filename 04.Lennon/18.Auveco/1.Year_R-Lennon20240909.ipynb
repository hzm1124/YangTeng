```python
import requests

import sys
sys.path.append('../00.Tools')
from crawler_configuration import get_header, get_proxy

b = 0
while True:
    b += 1
    print(b)
    try:
        resp = requests.get('https://www.auveco.com/store_ss2/extensions/UnlockCommerce/YearMakeModel/1.0.0/Modules/YearMakeModel/SuiteScript2/YearMakeModel.Service.ss?action=initymm',
                            headers=get_header(),
                            proxies=get_proxy(),
                            timeout=(10, 10))
        
        if resp.status_code == 200:
            json_ = resp.json()

            list_year_code = [dict_['yearId'] if dict_['yearId'] else '' for dict_ in json_]
            while '' in list_year_code:
                list_year_code.remove('')
            list_year = [int(dict_['year'].strip()) if dict_['year'] else '' for dict_ in json_]
            while '' in list_year:
                list_year.remove('')
            if list_year:
                break
    except KeyboardInterrupt:
        break
    except:
        continue

json_
```

```python
import pandas as pd

output = pd.DataFrame({'Year': list_year,
                       'Year Code': list_year_code})

output = output.sort_values(by=['Year'],
                            ascending=[False],
                            ignore_index=True)
output.to_excel('./file/1.year.xlsx', index=False)

output
```
