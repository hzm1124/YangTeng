```python
# import gevent
from gevent import monkey
monkey.patch_all()
import gevent,time,random
from gevent.queue import Queue
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.common.exceptions import TimeoutException
from selenium.common.exceptions import WebDriverException

# driver.implicitly_wait(20)
chrome_options=Options()
chrome_options.add_argument('blink-settings=imagesEnabled=false')
user_ag='Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36'
chrome_options.add_argument('user-agent=%s'%user_ag)
# chrome_options.add_argument('--headless')

df_old=pd.read_excel('D:/Users/YangTeng/Desktop/培训文件/Amazon专项优化跟踪/daily work/2.6汇报优化SKU.xlsx',sheet_name='Sheet1')
url_list=df_old[['SKU','Asin']]
df_new=pd.DataFrame(columns=['SKU','ASIN','售价','星级','Review','QA 数量','几节小类类目','大类排名','小类排名','小类2排名','小类3排名','是否更改','更新时间'])

error_record={}
print('当前页面开始时间{}'.format(time.strftime('%m月%d日%H:%M:%S', time.localtime())))

work=Queue()
for url_sku in zip(url_list['SKU'],url_list['Asin']):
    work.put_nowait(url_sku)

def crawler():
    global df_new
    global error_record
    df_craw=pd.DataFrame(columns=['SKU','ASIN','售价','星级','Review','QA 数量','几节小类类目','大类排名','小类排名','小类2排名','小类3排名','是否更改','更新时间'])
    driver=webdriver.Chrome(options=chrome_options)
    driver.set_page_load_timeout(30)
    driver.get('https://www.amazon.com/dp/B07DR81W27')
    button=driver.find_element(By.XPATH,'//*[@id="glow-ingress-block"]')
    button.click()
    time.sleep(8)
    zip01=driver.find_element(By.XPATH,'//*[@id="GLUXZipUpdateInput"]')
    zip01.send_keys('91710')
    time.sleep(1)
    #设置
    button=driver.find_element(By.XPATH,'//*[@id="GLUXZipUpdate"]/span/input')
    button.click()
    time.sleep(3)
    #完成
    
    for dd in range(1,10):
        try:
            button=driver.find_element(By.XPATH,'/html/body/div[{}]/div/div/div[2]/span'.format(dd))
        except:
            pass
    button.click()
    time.sleep(3)  
    
    while not work.empty():
        url_sku=work.get_nowait()
        sku=url_sku[0]
        url=url_sku[1]
        for non in range(3):
            try:
                driver.get('https://www.amazon.com/dp/{}'.format(url))
                print('正在爬取https://www.amazon.com/dp/{},第{}轮'.format(url,(non+1)))
                try:
                    price=driver.find_element(By.XPATH,'//*[@id="corePrice_feature_div"]/div/span/span[2]').text.replace("\n",".").replace("。",".")
                except:
                    try:
                        pricehigh=driver.find_element(By.XPATH,'//*[@id="corePriceDisplay_desktop_feature_div"]/div[1]/span/span[2]/span[2]').text
                        pricelow=driver.find_element(By.XPATH,'//*[@id="corePriceDisplay_desktop_feature_div"]/div[1]/span/span[2]/span[3]').text
                        price=str('$'+pricehigh+'。'+pricelow).replace("\n",".").replace("。",".")
                    except:
                        price=driver.find_element(By.XPATH,'//*[@id="availability"]/span').text
                        if price=='Currently unavailable.':
                            price='-'
                star=0
                BSR=0
                for tt in range(1,4):
                    try:
                        item=driver.find_element(By.XPATH,'//*[@id="productDetails_detailBullets_sections1"]/tbody/tr[{}]/th'.format(tt)).text
                        if item=='Customer Reviews':
                            star=driver.find_element(By.XPATH,'//*[@id="productDetails_detailBullets_sections1"]/tbody/tr[{}]/td'.format(tt)).text.split('\n',1)[1]
                        elif item=='Best Sellers Rank':
                            BSR=driver.find_elements(By.XPATH,'//*[@id="productDetails_detailBullets_sections1"]/tbody/tr[{}]/td/span/span'.format(tt))
                    except:
                        pass       
                if star==0:
                    star='-'
                if BSR==0:
                    BSR='' 
                list_BSR=len(BSR)
                try:
                    review=driver.find_element(By.XPATH,'//*[@id="acrCustomerReviewText"]').text
                except:
                    review=0
                s=1
                list_item=[]
                for bs in BSR:
                    list_item.append(bs.text)
                    s=s+1
                try:
                    QA=driver.find_element(By.XPATH,'//*[@id="askATFLink"]/span').text
                except:
                    driver.get('https://www.amazon.com/ask/questions/asin/{}/ref=ask_dp_dpmw_ql_hza?isAnswered=true'.format(url))
                    try:
                        QA=driver.find_element(By.XPATH,'//*[@id="a-page"]/div[1]/div[5]/div[2]/span').text.split(' ')[3]
                    except:
                        QA=0
                print('https://www.amazon.com/dp/{}'.format(url),work.qsize(),"价格为:{} 星级为:{} 小节数量为:{} Review数量为:{} QA 数量为:{}".format(price,(str(star).split(' ',1)[0]),list_BSR,str(review).split(' ',1)[0],str(QA).split(' ',1)[0]))
                rank_big='-'
                rank1='-'
                rank2='-'
                rank3='-'
                try:
                    rank_big=list_item[0].strip('#').replace(",","").split(' ',1)[0]
                    rank1=list_item[1].strip('#').replace(",","").split(' ',1)[0]
                    rank2=list_item[2].strip('#').replace(",","").split(' ',1)[0]
                    rank3=list_item[3].strip('#').replace(",","").split(' ',1)[0]
                except:
                      pass
                end_time=time.strftime('%m月%d日%H:%M:%S', time.localtime())
                yn='已更改'
                break
            except TimeoutException:
                price=star=review=QA=list_BSR=rank_big=rank1=rank2=rank3='-'
                yn='未更改'
                pass
            except WebDriverException:
                price=star=review=QA=list_BSR=rank_big=rank1=rank2=rank3='-'
                yn='未更改'
                print('https://www.amazon.com/dp/{}'.format(url),work.qsize(),"价格为:{} 星级为:{} 小节数量为:{} Review数量为:{} QA 数量为:{}".format(price,(str(star).split(' ',1)[0]),list_BSR,str(review).split(' ',1)[0],str(QA).split(' ',1)[0]))
                error_time=time.strftime('%m月%d日%H_%M_%S', time.localtime())
                error_record[error_time]='WebDriverException'
                driver.get_screenshot_as_file('D:/Users/YangTeng/Desktop/培训文件/Amazon专项优化跟踪/daily work/错误图片/{}.png'.format(error_time))
                pass
            except Exception as e:
                print(repr(e))
                error_time=time.strftime('%m月%d日%H_%M_%S', time.localtime())
                error_record[error_time]=repr(e)
                driver.get_screenshot_as_file('D:/Users/YangTeng/Desktop/培训文件/Amazon专项优化跟踪/daily work/错误图片/{}.png'.format(error_time))        
                end_time=time.strftime('%m月%d日%H:%M:%S', time.localtime())
                price=star=review=QA=list_BSR=rank_big=rank1=rank2=rank3='-'
                yn='未更改'
                break
        df_more=pd.DataFrame([{'SKU':sku,
                        'ASIN':url,
                       '售价':price,
                       '星级':str(star).split(' ',1)[0],
                       'Review':str(review).split(' ',1)[0],
                       'QA 数量':str(QA).split(' ',1)[0],
                       '几节小类类目':list_BSR,
                       '大类排名':rank_big,
                       '小类排名':rank1,
                       '小类2排名':rank2,
                       '小类3排名':rank3,
                       '是否更改':yn,
                        '更新时间':end_time
                      }])
        df_craw=pd.concat([df_craw,df_more])
        time.sleep(random.uniform(2,3))
    df_new=pd.concat([df_new,df_craw],ignore_index=True)
    driver.close()
    driver.quit()

task_list=[]
for x in range(8):
    task=gevent.spawn(crawler)
    task_list.append(task)
gevent.joinall(task_list)

print(time.strftime('%m月%d日%H:%M:%S', time.localtime()))
print('queue爬取完毕')

#未更改项
driver=webdriver.Chrome(options=chrome_options)
driver.set_page_load_timeout(60)
driver.get('https://www.amazon.com/dp/B00ZYNPKOO')
button=driver.find_element(By.XPATH,'//*[@id="glow-ingress-block"]')
button.click()
time.sleep(3)
zip01=driver.find_element(By.XPATH,'//*[@id="GLUXZipUpdateInput"]')
zip01.send_keys('91710')
time.sleep(1)
#设置
button=driver.find_element(By.XPATH,'//*[@id="GLUXZipUpdate"]/span/input')
button.click()
time.sleep(3)
#完成
error_record={}
for dd in range(1,10):
    try:
        button=driver.find_element(By.XPATH,'/html/body/div[{}]/div/div/div[2]/span'.format(dd))
    except:
        pass
button.click()
time.sleep(3)  
# 开始爬取未更改项
print(len(df_new[df_new['是否更改'].isin(['未更改'])].index))
for h in df_new[df_new['是否更改'].isin(['未更改'])].index:
    i=df_new.loc[h,'ASIN']
    for non in range(3):
        try:
            print('https://www.amazon.com/dp/{}'.format(i))
            print('当前页面开始时间{}'.format(time.strftime('%H:%M:%S', time.localtime())))
            driver.get('https://www.amazon.com/dp/{}'.format(i))
            time.sleep(3)  
            try:
                price=driver.find_element(By.XPATH,'//*[@id="corePrice_feature_div"]/div/span/span[2]').text.replace("\n",".").replace("。",".")
            except:
                try:
                    pricehigh=driver.find_element(By.XPATH,'//*[@id="corePriceDisplay_desktop_feature_div"]/div[1]/span/span[2]/span[2]').text
                    pricelow=driver.find_element(By.XPATH,'//*[@id="corePriceDisplay_desktop_feature_div"]/div[1]/span/span[2]/span[3]').text
                    price=str('$'+pricehigh+'。'+pricelow).replace("\n",".").replace("。",".")
                except:
                    price=driver.find_element(By.XPATH,'//*[@id="availability"]/span').text
                    if price=='Currently unavailable.':
                        price='-'
            star=0
            BSR=0
            for tt in range(1,4):
                try:
                    item=driver.find_element(By.XPATH,'//*[@id="productDetails_detailBullets_sections1"]/tbody/tr[{}]/th'.format(tt)).text
                    if item=='Customer Reviews':
                        star=driver.find_element(By.XPATH,'//*[@id="productDetails_detailBullets_sections1"]/tbody/tr[{}]/td'.format(tt)).text.split('\n',1)[1]
                    elif item=='Best Sellers Rank':
                        BSR=driver.find_elements(By.XPATH,'//*[@id="productDetails_detailBullets_sections1"]/tbody/tr[{}]/td/span/span'.format(tt))
                except:
                    pass       
            if star==0:
                star='-'
            if BSR==0:
                BSR='' 
            list_BSR=len(BSR)
            try:
                review=driver.find_element(By.XPATH,'//*[@id="acrCustomerReviewText"]').text
            except:
                review=0
            s=1
            list_item=[]
            for bs in BSR:
                list_item.append(bs.text)
                s=s+1
            try:
                QA=driver.find_element(By.XPATH,'//*[@id="askATFLink"]/span').text
            except:
                driver.get('https://www.amazon.com/ask/questions/asin/{}/ref=ask_dp_dpmw_ql_hza?isAnswered=true'.format(i))
                try:
                    QA=driver.find_element(By.XPATH,'//*[@id="a-page"]/div[1]/div[5]/div[2]/span').text.split(' ')[3]
                except:
                    QA=0
            df_new.loc[h,'售价']=price
            df_new.loc[h,'星级']=str(star).split(' ',1)[0]
            df_new.loc[h,'几节小类类目']=list_BSR
            df_new.loc[h,'Review']=str(review).split(' ',1)[0]
            df_new.loc[h,'QA 数量']=str(QA).split(' ',1)[0]
            print("价格为:{}".format(price))
            print("星级为{}".format(str(star).split(' ',1)[0]))
            print("小节数量为{}".format(list_BSR))
            print("Review数量为{}".format(str(review).split(' ',1)[0]))
            print('QA 数量为:{}'.format(str(QA).split(' ',1)[0]))
            try:
                df_new.loc[h,'大类排名']=list_item[0].strip('#').replace(",","").split(' ',1)[0]
                df_new.loc[h,'小类排名']=list_item[1].strip('#').replace(",","").split(' ',1)[0]
                df_new.loc[h,'小类2排名']=list_item[2].strip('#').replace(",","").split(' ',1)[0]
                df_new.loc[h,'小类3排名']=list_item[3].strip('#').replace(",","").split(' ',1)[0]
            except:
                  pass
            end_time=time.strftime('%m月%d日%H:%M:%S', time.localtime())
            print('当前页面结束时间{}'.format(end_time))
            time.sleep(random.uniform(0,1))
            print('记录到第{}条'.format(h))
            df_new.loc[h,'是否更改']='已更改'
            df_new.loc[h,'更新时间']=end_time
            break
        except TimeoutException:
            price=star=review=QA=list_BSR=rank_big=rank1=rank2=rank3='-'
            yn='未更改'
            pass
        except WebDriverException:
            price=star=review=QA=list_BSR=rank_big=rank1=rank2=rank3='-'
            yn='未更改'
            error_time=time.strftime('%m月%d日%H_%M_%S', time.localtime())
            error_record[error_time]='WebDriverException'
            driver.get_screenshot_as_file('D:/Users/YangTeng/Desktop/培训文件/Amazon专项优化跟踪/daily work/错误图片/{}.png'.format(error_time))
            pass
        except Exception as e:
            print(repr(e))
            error_time=time.strftime('%m月%d日%H_%M_%S', time.localtime())
            error_record[error_time]=repr(e)
            driver.get_screenshot_as_file('D:/Users/YangTeng/Desktop/培训文件/Amazon专项优化跟踪/daily work/错误图片/{}.png'.format(error_time))        
            price=star=review=QA=list_BSR=rank_big=rank1=rank2=rank3='-'
            df_new.loc[h,'是否更改']='未更改'
            df_new.loc[h,'更新时间']=end_time
            break
driver.close()
driver.quit()
print('爬取完毕')
record_date=time.strftime('%m月%d日', time.localtime())
print(time.strftime('%m月%d日%H:%M:%S', time.localtime()))
df_new.to_excel('D:/Users/YangTeng/Desktop/培训文件/Amazon专项优化跟踪/daily work/{}优化SKU.xlsx'.format(record_date),sheet_name='Sheet1',index=False)
print('写入完毕')
print('原始数据条数{}'.format(df_old.shape[0]))
print('记录数据条数{}'.format(df_new.shape[0]))
```

# 图片转化

Readme：
速度设置的很高，所以没挂代理池，需要手动开vpn

## 最新版
    版本更新-20230313：设置了x/y_scale为固定大小

```python
import io,os,urllib,gevent,time,random,xlsxwriter
from PIL import Image
from io import BytesIO
from urllib.request import urlopen
from gevent import monkey
monkey.patch_all()
from gevent.queue import Queue
import pandas as pd
fixed_size=266
df_new=pd.read_excel('D:/Users/YangTeng/Desktop/培训文件/Amazon专项优化跟踪/daily work/爬取eBay/mirror_all.xlsx',sheet_name='Sheet1')
url_list=df_new
work=Queue()
tt=time.strftime('%m月%d日%H_%M_%S', time.localtime())
wb = xlsxwriter.Workbook(r'D:/{}商品数据.xlsx'.format(tt))
worksheet = wb.add_worksheet("图片") 
worksheet.write_string(0,1,'图片地址')
worksheet.write_string(0,2,'图片')
print('开始')
for url_sku in zip(url_list.index,url_list['主图']):
    work.put_nowait(url_sku)
def crawler():
    while not work.empty():
        url_sku=work.get_nowait()
        index_1=url_sku[0]+1
        i=url_sku[1]
        url1 =i
#         print(work.qsize(),url1)
        try:
            if url1 !='-':
                image_data = io.BytesIO(urllib.request.urlopen(url1).read())
                img = Image.open(image_data)
                x = float(fixed_size / img.size[0])
                y = float(fixed_size / img.size[1])
                scale=min(x,y)
                worksheet.write_string(index_1,1,"{}".format(url1))
                worksheet.insert_image(row=index_1,col=2,filename=url1,options={'x_scale': scale,
                                                                                'y_scale': scale,
                                                                                'image_data': image_data,
                                                                                'positioning': 1})
                print(str(index_1)+"    "+url1+' 写入完成')
                time.sleep(random.random())
            else:
                worksheet.write_string(index_1,1,"{}".format(url1))
                worksheet.write_string(index_1,2,"-")
        except:
            worksheet.write_string(index_1,1,"{}".format(url1))
            worksheet.write_string(index_1,2,"-")
            
task_list=[]
for x in range(100):
    task=gevent.spawn(crawler)
    task_list.append(task)
gevent.joinall(task_list)

wb.close()
print(time.strftime('%m月%d日%H:%M:%S', time.localtime()))
```
