```python
param_search_term = 'Trunk Lock Cylinder'
param_page_size = 1000
```

```python
import requests

import sys
sys.path.append('../00.Tools')
import UA
import Proxy
```

```python
resp = requests.get('https://parts.gmparts.com/wcs/resources/store/11705/Search/product?searchTerm=' + param_search_term + '&orderBy=2&pageNumber=1&pageSize=' + str(param_page_size) + '&catalogId=3074457345616677718',
                    headers=UA.get_User_Agent_Requests(),
                    proxies=Proxy.get_Proxy_Requests()).text

resp
```

```python
import json
```

```python
json_data = json.loads(resp)

json_data
```

```python
list_url = [part['seoUrl'].strip() for part in json_data['catalogEntryView']]

list_url
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({'No.': [i+1 for i in range(json_data['recordSetCount'])],
                   'Url_1': list_url})

df.to_excel('./menu.xlsx', index=False)

df
```
