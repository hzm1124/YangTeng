```python
from urllib.parse import urlsplit
```

```python
tuple_ = urlsplit('https://www.dormanproducts.com/gsearch.aspx?type=keyword&origin=keyword&parttype=Radiator%2520Fan%2520Assembly&start=0&num=100')

tuple_
```

```python
from urllib.parse import parse_qs
```

```python
dict_query = {key: value[0] for key, value in parse_qs(tuple_.query).items()}

dict_query
```

```python
dict_query['start'] = 0
dict_query['num'] = 100

dict_query
```

```python
from urllib.parse import urlunsplit
from urllib.parse import urlencode
```

```python
url = urlunsplit((tuple_.scheme, tuple_.netloc, tuple_.path, urlencode(dict_query), tuple_.fragment))

url
```

```python
import requests

import sys
sys.path.append('../00.Tools')
from configuration import generate_header, generate_proxy
```

```python
response = requests.get(url,
                        headers=generate_header(**{'Accept': 'text/html',
                                                   'Accept-Language': 'en',
                                                   'Cache-Control': 'no-cache',
                                                   'Pragma': 'no-cache',
                                                   'Sec-Fetch-Dest': 'document',
                                                   'Sec-Fetch-Mode': 'navigate',
                                                   'Sec-Fetch-Site': 'same-origin',
                                                   'Sec-Fetch-User': '?1',
                                                   'Upgrade-Insecure-Requests': '1'}),
                        proxies=generate_proxy(),
                        timeout=(10, 15),
                        allow_redirects=False)

response.status_code
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(response.text, 'html5lib')

with open('./1.page.html', 'w', encoding='utf-8') as file:
    file.write(str(soup))

print(soup.prettify())
```
