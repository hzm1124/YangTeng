```python
import pandas as pd

df = pd.DataFrame(columns=['Part Number',
                           'Category',
                           'OE',
                           '图片链接',
                           '链接',
                           'status'])
```

```python
df_menu = pd.read_excel('./url.xlsx', header=0)
list_url = df_menu['url'].to_list()

print(len(list_url))
list_url
```

```python
import requests
```

```python
i = 0
resp = requests.get(list_url[i]).text

resp
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(resp, 'lxml')

print(soup.prettify())
```

```python
with open('./product.html', 'w', encoding='utf-8') as file:
    file.write(str(soup))
```

```python
from lxml import etree
```

```python
html = etree.HTML(str(soup))

list_part_number = html.xpath('//span[@id="lblProductName"]')
list_part_number= [part_number.text for part_number in list_part_number]

list_part_number
```

```python
list_category = html.xpath('//span[@id="lblProductDesc"]')
list_category= [category.text for category in list_category]

list_category
```

```python
list_img = html.xpath('//img[@id="imgProductMain"]')
list_img = [img.get('src').replace('/icon/', '/medium/') for img in list_img]

list_img
```

```python
list_oe = html.xpath('//section[@id="productOE"]/descendant::*/th[@scope="row"]')
list_oe = ';'.join([oe.text for oe in list_oe])

list_oe
```

```python
df_temp = pd.DataFrame({'Part Number': list_part_number,
                        'Category': list_category,
                        'OE': list_oe,
                        '图片链接': list_img,
                        '链接': list_url[i],
                        'status': 'ok'})

df = pd.concat([df, df_temp], ignore_index=True).fillna('')

df
```
