```python
import pandas as pd

df_menu = pd.read_excel('./item.xlsx', header=0)
list_item = df_menu['Item'].to_list()

print(len(list_item))
list_item
```

```python
df = pd.DataFrame(columns=['Item',
                           'Fit',
                           'status'])
```

```python
import requests
```

```python
i = 0

resp = requests.get('https://vi.vipr.ebaydesc.com/ws/eBayISAPI.dll?ViewItemDescV4&item=' + str(list_item[i]) + '&t=0&category=42604&seller=lkqonline&excSoj=1&excTrk=1&lsite=100&ittenable=true&domain=ebay.com&descgauge=1&cspheader=1&oneClk=2&secureDesc=1').text

resp
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(resp, 'lxml')

print(soup.prettify())
```

```python
with open('./page.html', 'w', encoding='utf-8') as file:
    file.write(str(soup))
```

```python
from lxml import etree
```

```python
html = etree.HTML(str(soup))

description = ''.join(html.xpath('//div[@class="itemdescription"]/descendant-or-self::*/text()'))

description
```

```python
fit = description.strip().split('This will fit:')[1]

fit
```

```python
df_temp = pd.DataFrame({'Item': list_item[i],
                        'Fit': fit,
                        'status': ['ok']})

df = pd.concat([df, df_temp], ignore_index=True).fillna('')

df
```
