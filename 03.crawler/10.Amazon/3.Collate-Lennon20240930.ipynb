```python
import os
import shutil

import pandas as pd

from tqdm import tqdm
import json

# = = = = = = = = = = = = = = =

if os.path.exists('./file/3.collate'):
    shutil.rmtree('./file/3.collate')

os.makedirs('./file/3.collate')

# = = = = = = = = = = = = = = =

list_file = sorted(list(os.walk('./file/2.part'))[0][2])
while '.DS_Store' in list_file:
    list_file.remove('.DS_Store')
while '0.null.txt' in list_file:
    list_file.remove('0.null.txt')
    
# = = = = = = = = = = = = = = =

for file in list_file:

    print(file)

    # = = = = = = = = = = = = = = =
    
    input_ = pd.read_excel(f'./file/2.part/{file}',
                       header=0,
                       dtype=str).fillna('')

    input_['No'] = input_['No'].astype(int)
    
    # = = = = = = = = = = = = = = =

    list_column, list_column_overview, list_column_information = input_.columns.tolist(), [], []

    # = = = = = = = = = = = = = = =

    list_df = []
    for i in tqdm(range(len(input_)), desc='Progress', ncols=77):
        df_temp, dict_dict_overview, dict_dict_technical, dict_dict_additional, dict_dict_technical_2, dict_dict_additional_2, dict_dict_information = pd.DataFrame([input_.iloc[i]]).reset_index(drop=True), json.loads(input_.loc[i, 'Json_Overview']), json.loads(input_.loc[i, 'Json_Technical']), json.loads(input_.loc[i, 'Json_Additional']), json.loads(input_.loc[i, 'Json_Technical 2']), json.loads(input_.loc[i, 'Json_Additional 2']), json.loads(input_.loc[i, 'Json_Information'])

        # = = = = = = = = = = = = = = =

        dict_count = {}
        for key in dict_dict_technical:
            if key in dict_count:
                dict_count[key] += 1
            else:
                dict_count[key] = 1
        for key in dict_dict_additional:
            if key in dict_count:
                dict_count[key] += 1
            else:
                dict_count[key] = 1
        for key in dict_dict_technical_2:
            if key in dict_count:
                dict_count[key] += 1
            else:
                dict_count[key] = 1
        for key in dict_dict_additional_2:
            if key in dict_count:
                dict_count[key] += 1
            else:
                dict_count[key] = 1
        for key in dict_dict_information:
            if key in dict_count:
                dict_count[key] += 1
            else:
                dict_count[key] = 1

        dict_dict, dict_no = {}, {}
        for key, value in dict_dict_technical.items():
            if dict_count[key] == 1:
                dict_dict[key] = value
            else:
                if key in dict_no:
                    dict_no[key] += 1
                else:
                    dict_no[key] = 1
                dict_dict[f'{key}-{dict_no[key]}'] = value
        for key, value in dict_dict_additional.items():
            if dict_count[key] == 1:
                dict_dict[key] = value
            else:
                if key in dict_no:
                    dict_no[key] += 1
                else:
                    dict_no[key] = 1
                dict_dict[f'{key}-{dict_no[key]}'] = value
        for key, value in dict_dict_technical_2.items():
            if dict_count[key] == 1:
                dict_dict[key] = value
            else:
                if key in dict_no:
                    dict_no[key] += 1
                else:
                    dict_no[key] = 1
                dict_dict[f'{key}-{dict_no[key]}'] = value
        for key, value in dict_dict_additional_2.items():
            if dict_count[key] == 1:
                dict_dict[key] = value
            else:
                if key in dict_no:
                    dict_no[key] += 1
                else:
                    dict_no[key] = 1
                dict_dict[f'{key}-{dict_no[key]}'] = value
        for key, value in dict_dict_information.items():
            if dict_count[key] == 1:
                dict_dict[key] = value
            else:
                if key in dict_no:
                    dict_no[key] += 1
                else:
                    dict_no[key] = 1
                dict_dict[f'{key}-{dict_no[key]}'] = value

        # = = = = = = = = = = = = = = =
        
        for dict_ in dict_dict_overview.values():
            for key, value in dict_.items():
                key = f'[Overview] {key}'

                if key not in list_column_overview:
                    list_column_overview.append(key)

                if type(value) == dict:
                    list_str = []
                    for key_2, dict_2 in value.items():
                        list_str.append(f'{int(key_2)+1}. {list(dict_2.keys())[0]} ({list(dict_2.values())[0]})')
                    df_temp.loc[0, key] = '\n'.join(list_str)
                else:
                    df_temp.loc[0, key] = value

        # = = = = = = = = = = = = = = =

        for dict_ in dict_dict.values():
            for key, value in dict_.items():
                if key in list_column:
                    key = f'{key} (Information)'

                if key not in list_column_information:
                    list_column_information.append(key)

                if type(value) == dict:
                    list_str = []
                    for key_2, dict_2 in value.items():
                        list_str.append(f'{int(key_2)+1}. {list(dict_2.keys())[0]} ({list(dict_2.values())[0]})')
                    df_temp.loc[0, key] = '\n'.join(list_str)
                else:
                    df_temp.loc[0, key] = value

        list_df.append(df_temp)
    
    # = = = = = = = = = = = = = = =

    output = pd.concat(list_df, ignore_index=True).fillna('')

    output = output.sort_values(by=['No'],
                                ascending=[True],
                                ignore_index=True)

    output[list_column[:6] + sorted(list_column_information) + sorted(list_column_overview) + list_column[6:]].to_excel(f'./file/3.collate/{file}', index=False)

    # = = = = = = = = = = = = = = =

    print()

print('Done ~')
```
