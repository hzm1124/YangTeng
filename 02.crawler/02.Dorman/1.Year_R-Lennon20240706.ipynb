```python
param_date = '20240706'
```

```python
import requests
import json

import sys
sys.path.append('../00.Tools')
from crawler_configuration import get_header, get_proxy

data = {'cssClass': 'form-control',
        'dropdownType': '0',
        'selectedValue': '',
        'vehicleAttributes': {'Make': '',
                              'MakeID': 0,
                              'ModelID': 0,
                              'SkinID': '1',
                              'Year': 0}}

resp = requests.post('https://www.dormanproducts.com/ApplicationGuideService.asmx/BuildDropdown',
                     data=json.dumps(data),
                     headers=get_header(**{'Content-Type': 'application/json; charset=UTF-8'}),
                     proxies=get_proxy(),
                     timeout=(10, 10)).text

resp
```

```python
json_ = json.loads(resp.strip())

json_
```

```python
from bs4 import BeautifulSoup
from lxml import etree

soup = BeautifulSoup(json_['d'].strip(), 'lxml')
html = etree.HTML(str(soup))

with open('./year.html', 'w', encoding='UTF-8') as file:
    file.write(str(soup))

print(soup.prettify())
```

```python
list_year = [int(value.strip()) for value in html.xpath('//select[@id="yearDropdown"]/option/@value')[1:]]

list_year
```

```python
import pandas as pd

output = pd.DataFrame({'Year': list_year})
output = output.sort_values(by=['Year'],
                            ascending=[False],
                            ignore_index=True)

output.to_excel(f'./{param_date}/1.year.xlsx', index=False)

output
```
