```python
from gevent import monkey
monkey.patch_all()
import gevent,time,random
from gevent.queue import Queue
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.common.exceptions import TimeoutException
from selenium.common.exceptions import WebDriverException
from selenium.common.exceptions import NoSuchElementException
user_ag='Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36'
# driver.implicitly_wait(20)
chrome_options=webdriver.ChromeOptions()
chrome_options.add_argument('user-agent=%s'%user_ag)
chrome_options.add_argument('--proxy-server=http://a002.kdltps.com:15818') #ip代理池
chrome_options.add_argument('blink-settings=imagesEnabled=false')
chrome_options.add_argument('--headless')
print('开始爬取')
df_cat=pd.read_excel(r'D:11111.xlsx')
uo=0
ssr=0
file=1
for i in df_cat['品牌']:
    mau=df_cat.loc[ssr,'品牌']
    pag=df_cat.loc[ssr,'Part Group']
    pat=df_cat.loc[ssr,'Part Type']
    pat_new=pat.replace(", "," ")
    url_num=df_cat.loc[ssr,'number']
    print('爬取第{}类，{}  {}  {}  {}中'.format(ssr,mau,pag,pat,url_num))
    ssr+=1
    mr=0
    for long in range(15):
        try:
            h=0
            time.sleep(random.uniform(3,5))
            driver=webdriver.Chrome(options=chrome_options)
            driver.get('https://www.rockauto.com/en/parts/{},{},{}'.format(mau,pat_new,url_num))
            df=pd.DataFrame(columns=['OE'])
            time.sleep(random.uniform(3,5))
            number=driver.find_elements(By.XPATH,'''//*[@class="ranavnode"]''')
            time.sleep(random.uniform(3,5))
            for mmt in number[1:]:
                df.loc[h,'OE']=mmt.text
            #     print(mmt.text)
                h+=1
            driver.close()
            print(df.shape)
            if df.loc[0,'OE'] !='':
                break
        except:
            pass
        mr+=1
    if mr==15:
        continue
        
    #------------------------------------------

    record_date=time.strftime('%m月%d日', time.localtime())
    url_list=df['OE']
    df_new=pd.DataFrame(columns=['Item', '链接', '标题',
           'OEM Part Number','AIN','车型','主图'
                                 ,'配件NO','配件数量','配件类别','配件名称','配件图片','配件OE','配件AIN'
                                 , '是否更改','更新时间'])
    error_record={}
    end_time=time.strftime('%m月%d日%H:%M:%S', time.localtime())
    print('当前页面开始时间{}'.format(time.strftime('%m月%d日%H:%M:%S', time.localtime())))
    work=Queue()
    for url5 in df['OE']:
        work.put_nowait(url5)
    def crawler():
        global df_new
        global error_record
        global end_time
        df_craw=pd.DataFrame(columns=['Item', '链接', '标题',
           'OEM Part Number','AIN','车型','主图'
                                 ,'配件NO','配件数量','配件类别','配件名称','配件图片','配件OE','配件AIN'
                                 , '是否更改','更新时间'])
        driver=webdriver.Chrome(options=chrome_options)
        driver.set_page_load_timeout(60)
        while not work.empty():
            url_oe=work.get_nowait()
            for non in range(30):
                try:
                    driver.get('https://www.rockauto.com/en/parts/{},{},{},{}'.format(mau,url_oe,pat_new,url_num))
                    print('https://www.rockauto.com/en/parts/{},{},{},{}      ,第{}轮'.format(mau,url_oe,pat_new,url_num,(non+1)))
                    link='https://www.rockauto.com/en/parts/{},{},{},{}'.format(mau,url_oe,pat_new,url_num)
                    title='-'
                    try:
                        title=driver.find_element(By.CLASS_NAME,'listing-text-row-moreinfo-truck').text.replace("Info","")
                    except:
                        pass
                    pic='-'
                    try:                    
                        pic=driver.find_element(By.CLASS_NAME,'listing-inline-image.listing-inline-image-thumb.listing-inline-image-border').get_attribute("src").replace("t.jpg","p.jpg")
                    except:
                        pass
                    button01=driver.find_element(By.XPATH,'''//*[@title="Buyer's Guide"]''')
                    button01.click()
                    time.sleep(2)
                    car='-'
                    try:
                        car=driver.find_element(By.CLASS_NAME,'buyersguide-nested').text.replace("Please refer to catalog for application details.","")
                    except:
                        pass
                    try:
                        OE='-'
                        AIN='-'
                        info=driver.find_element(By.CLASS_NAME,'ra-btn.ra-btn-moreinfo').get_attribute("href")
                        driver.get(info)
                        try:
                            OEs=driver.find_elements(By.XPATH,'''//*[@aria-label="Alternate/OEM Part Number(s)"]''')
                            for oe in OEs:
                                if 'OEM / Interchange Numbers:' in oe.text:
                                    OE=oe.text.replace("OEM / Interchange Numbers: ","")
                                elif 'Alternate Inventory Numbers:' in oe.text:
                                    AIN=oe.text.replace("{Alternate Inventory Numbers: ","").replace("}","")
                        except:
                            pass
                    except NoSuchElementException:
                        pass
                    pj_geshu=[]
                    pj_leibie=[]
                    pj_name=[]
                    pj_oe_url=[]
                    pj_pic=[]
                    try:
                        button02=driver.find_elements(By.XPATH,'''//*[@class="listing-inner altrow-b-0"]''')
                        for i in button02:
                            pj_geshu.append(i.find_element(By.XPATH,'''./tr/td[1]''').text)
                            pj_leibie.append(i.find_element(By.XPATH,'''./tr/td[2]''').text)
                            pj_name.append(i.find_element(By.XPATH,'''./tr/td[3]''').text.replace(" Info",""))
                            try:
                                pj_oe_url.append(i.find_element(By.XPATH,'''./tr/td[3]/div[1]/a''').get_attribute("href"))
                            except:
                                pj_oe_url.append("")
                            try:
                                pj_pic.append(i.find_element(By.XPATH,'''./tr/td[4]/div/table/tbody/tr/td[2]/img''').get_attribute("src"))
                            except:
                                pj_pic.append("")
                    except:
                        pass
                    try:
                        button03=driver.find_elements(By.XPATH,'''//*[@class="listing-inner altrow-b-1"]''')
                        for i in button03:
                            pj_geshu.append(i.find_element(By.XPATH,'''./tr/td[1]''').text)
                            pj_leibie.append(i.find_element(By.XPATH,'''./tr/td[2]''').text)
                            pj_name.append(i.find_element(By.XPATH,'''./tr/td[3]''').text.replace(" Info",""))
                            try:
                                pj_oe_url.append(i.find_element(By.XPATH,'''./tr/td[3]/div[1]/a''').get_attribute("href"))
                            except:
                                pj_oe_url.append("")
                            try:
                                pj_pic.append(i.find_element(By.XPATH,'''./tr/td[4]/div/table/tbody/tr/td[2]/img''').get_attribute("src"))
                            except:
                                pj_pic.append("")
                    except:
                        pass
                    pj_num=[i for i in range(1,len(pj_name)+1)]
                    pj_OE_l=[]
                    pj_AIN_l=[]
                    for so in pj_oe_url:
                        pj_OE='-'
                        pj_AIN='-'
                        for ul in range(3):
                            try:
                                time.sleep(random.uniform(1,3))
                                driver.get(so)
                                try:
                                    pjs=driver.find_elements(By.XPATH,'''//*[@aria-label="Alternate/OEM Part Number(s)"]''')
                                    for oe in pjs:
                                        if 'OEM / Interchange Numbers:' in oe.text:
                                            pj_OE=oe.text.replace("OEM / Interchange Numbers: ","")
                                        elif 'Alternate Inventory Numbers:' in oe.text:
                                            pj_AIN=oe.text.replace("{Alternate Inventory Numbers: ","").replace("}","")
                                    if pj_OE !='-' or pj_AIN !='-':
                                        break                  
                                except:
                                    pass
                            except:
                                pass
                        pj_OE_l.append(pj_OE)
                        pj_AIN_l.append(pj_AIN)
                    end_time=time.strftime('%m月%d日%H:%M:%S', time.localtime())
                    yn='已更改'
                    break
                except TimeoutException:
                    title=pic=car=AIN=OE='-'
                    yn='未更改'
                    end_time=time.strftime('%m月%d日%H:%M:%S', time.localtime())
                    time.sleep(10)
                    pass
                except WebDriverException:
                    title=pic=car=AIN=OE='-'
                    yn='未更改'
                    end_time=time.strftime('%m月%d日%H:%M:%S', time.localtime())
                    error_time=time.strftime('%m月%d日%H_%M_%S', time.localtime())
                    error_record[error_time]='WebDriverException'
                    pass
                except Exception as e:
                    print(repr(e))
                    error_time=time.strftime('%m月%d日%H_%M_%S', time.localtime())
                    error_record[error_time]=repr(e)        
                    end_time=time.strftime('%m月%d日%H:%M:%S', time.localtime())
                    title=pic=car=AIN=OE='-'
                    yn='未更改'
                    break
            print(work.qsize(),url_oe,link,title,AIN,OE)
#             print(work.qsize(),"item为: {} 链接为: {} 标题为: {} AIN为: {} OE为: {}".format(url_oe,link,title,AIN,OE))
            df_more=pd.DataFrame()
            if pj_num==[]:
                df_more=pd.DataFrame([{'Item':url_oe,
                                       '配件NO':pj_num,
                                        '链接':link,
                                        '标题':title,
                                        'OEM Part Number':OE,
                                        'AIN':AIN,
                                        '车型':car,
                                       '主图':pic,
                                        '配件数量':pj_geshu,
                                        '配件类别':pj_leibie,
                                       '配件链接':pj_oe_url,
                                        '配件名称':pj_name,
                                        '配件图片':pj_pic,
                                        '配件OE':pj_OE_l,
                                        '配件AIN':pj_AIN_l,
                                       '是否更改':yn,
                                        '更新时间':end_time
                                      }])
            else:
                url_oe2=[]
                url_oe2=[url_oe for i in range(len(pj_num))]
                df_more['Item']=url_oe2
                df_more['配件NO']=pj_num
                df_more['链接']=link
                df_more['标题']=title
                df_more['OEM Part Number']=OE
                df_more['AIN']=AIN
                df_more['车型']=car
                df_more['主图']=pic
                df_more['配件数量']=pj_geshu
                df_more['配件类别']=pj_leibie
                df_more['配件名称']=pj_name
                df_more['配件链接']=pj_oe_url
                df_more['配件图片']=pj_pic
                df_more['配件OE']=pj_OE_l
                df_more['配件AIN']=pj_AIN_l
                df_more['是否更改']=yn
                df_more['更新时间']=end_time
            df_craw=pd.concat([df_craw,df_more])
        df_new=pd.concat([df_new,df_craw],ignore_index=True)
        driver.close()
        driver.quit()
    task_list=[]
    for x in range(8):
        task=gevent.spawn(crawler)
        task_list.append(task)
    gevent.joinall(task_list)
    
    print(time.strftime('%m月%d日%H:%M:%S', time.localtime()))
    print('queue爬取完毕')
    print('爬取未更改项')
    df_new2=df_new.set_index(["Item", "链接","标题",'OEM Part Number','AIN','车型','主图','更新时间','是否更改'])
    mau=mau.replace("/","")
    pag=pag.replace("/","")
    pat=pat.replace("/","").replace('''"''',"")
    try:
        df_new2.to_excel(r'D:\Users\YangTeng\Desktop\培训文件\Amazon专项优化跟踪\daily work\爬取数据打包\数据记录\RockAuto爬取\3{},{},{},{}.xlsx'.format(mau,pag,pat,url_num),sheet_name='Sheet1')
    except:
        df_new2.to_excel(r'D:\Users\YangTeng\Desktop\培训文件\Amazon专项优化跟踪\daily work\爬取数据打包\数据记录\RockAuto爬取\3{}.xlsx'.format(file),sheet_name='Sheet1')
        file+=1
    print('{},{},{},{}写入完毕'.format(mau,pag,pat,url_num))
```
