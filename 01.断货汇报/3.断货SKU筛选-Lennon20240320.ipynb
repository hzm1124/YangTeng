```python
import os
```

```python
list_file = list(os.walk('./'))[0][2]
if '.DS_Store' in list_file:
    list_file.remove('.DS_Store')
if '0.null.txt' in list_file:
    list_file.remove('0.null.txt')
    
for file in list_file:
    if '计算表' in file:
        calculate = './' + file
        
print('calculate:      ' + calculate)
```

```python
import pandas as pd
```

```python
df = pd.read_excel(calculate,
                   sheet_name='附_1统计源（含新品、清仓、新品成长）',
                   header=3,
                   names=['SKU', '周销', '层级', '状态', '在库', '在途'],
                   usecols='C, D, F, H, L, M',
                   dtype=str).fillna('')

df['周销'] = df['周销'].astype(float)
df['在库'] = df['在库'].astype(int)
df['在途'] = df['在途'].astype(int)

df
```

```python
from datetime import datetime
```

```python
df = df[(df['在库'] == 0) & (df['在途'] == 0) & (df['层级'] != '普通')].reset_index(drop=True)
df = df.sort_values(by=['层级', '周销', 'SKU'], ascending=[True, False, True])
df['序号'] = [i+1 for i in range(len(df))]
df = df[['序号', 'SKU', '层级', '周销']]

df.to_excel('./断货SKU_' + datetime.now().strftime("%Y%m%d_%H%M%S") + '.xlsx', index=False)

df
```
