```python
param_market = 'AU'
url = 'https://www.ebay.com.au/str/ozepartsyourpartspartner'
```

```python
import requests

import sys
sys.path.append('../../00.Tools')
import UA
import Proxy
```

```python
if param_market == 'US':
    url += '?_ul=US&_stpos=91710&orig_cvip=true'
elif param_market == 'AU':
    url += '?_ul=AU&_stpos=2019&orig_cvip=true'
    
resp = requests.get(url,
                    headers=UA.get_User_Agent_Requests(),
                    proxies=Proxy.get_Proxy_Requests()).text

resp
```

```python
import json
```

```python
json_data = json.loads(resp.split('$storenode_C=(window.$storenode_C||[]).concat(')[1].split(']}]})</script></body>')[0] + ']}]}')

json_data
```

```python
list_part = json_data['o']['w'][4][2]['page']['modules']['STORE_CATEGORIES_MODULE']['categories']

list_part
```

```python
list_category = [part['label']['textSpans'][0]['text'].strip() for part in list_part]

list_category
```

```python
list_part = [part['action']['params'] for part in list_part]

list_part
```

```python
import os
```

```python
list_url = []
if param_market == 'US':
    for i in range(len(list_part)):
        os.makedirs('./category/' + str(i+1))
        
        url = 'https://www.ebay.com/sch/i.html?'
        for name in list_part[i]:
            url += name.strip() + '=' + list_part[i][name].strip() + '&'
        list_url.append(url + '_ul=US&_stpos=91710&orig_cvip=true&&_sop=16&&_ipg=240')
if param_market == 'AU':
    for i in range(len(list_part)):
        os.makedirs('./category/' + str(i+1))
        
        url = 'https://www.ebay.com.au/sch/i.html?'
        for name in list_part[i]:
            url += name.strip() + '=' + list_part[i][name].strip() + '&'
        list_url.append(url + '_ul=AU&_stpos=2019&orig_cvip=true&&_sop=16&&_ipg=240')
    
list_url
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({'Category_No.': [i+1 for i in range(len(list_part))],
                   'Category': list_category,
                   'Url': list_url})

df.to_excel('./category.xlsx', index=False)

df
```
