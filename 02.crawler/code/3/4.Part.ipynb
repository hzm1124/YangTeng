```python
import os
```

```python
list_file = [file for file in sorted(list(os.walk('./4.part'))[0][2]) if os.path.isfile(f'./4.part/{file}') and file.endswith('.xlsx')]

list_file
```

```python
import pandas as pd
```

```python
list_df = []
for i, file in enumerate(list_file):
    print(f'*  Lennon Log - Info - Step: Load ({i + 1} / {len(list_file)})')
    df = pd.read_excel(f'./4.part/{file}',
                       header=0,
                       dtype=str).fillna('')

    df['no'] = df['no'].astype(int)
    df['default_no'] = df['default_no'].astype(int)

    list_df.append(df.copy())

df = pd.concat(list_df, ignore_index=True)

df = df.sort_values(['item_number'],
                    ascending=True,
                    ignore_index=True)
df['no'] = range(1, len(df) + 1)
df.to_excel('./4.part.xlsx', index=False)

df
```
