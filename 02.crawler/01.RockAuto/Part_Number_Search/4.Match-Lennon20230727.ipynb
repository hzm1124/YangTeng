```python
import pandas as pd
```

```python
df = pd.read_excel('./menu.xlsx').fillna('')

df
```

```python
for i in range(len(df)):
    if str(df.loc[i, 'OE']).replace('-', '').strip() in str(df.loc[i, 'Alternate_OE_Part_Numbers']):
        df.loc[i, 'Match'] = True
    else:
        df.loc[i, 'Match'] = False
        
df
```

```python
df.to_excel('match.xlsx', index=False)
```
