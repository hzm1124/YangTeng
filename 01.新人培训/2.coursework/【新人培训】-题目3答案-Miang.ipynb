```python
import pandas as pd

input_ = pd.read_excel('./2.WEEK2—case题目0428.xlsx',
                       sheet_name='总表',
                       header=0,
                       usecols='A:B',
                       dtype=str).fillna('')

input_
```

```python
df_stock = pd.read_excel('./2.WEEK2—case题目0428.xlsx',
                         sheet_name='附3_库存表',
                         header=0,
                         usecols='A:I',
                         dtype=str).fillna('')

df_stock['在途'] = df_stock['在途'].astype(int)
df_stock['可售'] = df_stock['可售'].astype(int)

df_stock
```

```python
df_stock['PTSKU'] = [sku.removeprefix('US')[:-2] if sku.endswith('-N') or sku.endswith('-Y') else sku.removeprefix('US') for sku in df_stock['商品编码 4-25'].tolist()]
df_stock = df_stock[['PTSKU', '在途', '可售']].groupby(['PTSKU'], as_index=False).sum()

df_stock
```

```python
input_ = pd.merge(left=input_,
                  right=df_stock,
                  how='left',
                  left_on='SKU',
                  right_on='PTSKU').fillna('')

input_.columns = input_.columns.tolist()[:-1] + ['在库']

input_[['SKU', '仓库', '在库', '在途']].to_excel('./answer_3.xlsx', index=False)

input_
```
