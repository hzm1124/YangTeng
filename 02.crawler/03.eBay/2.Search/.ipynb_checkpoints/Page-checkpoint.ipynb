```python
url = 'https://www.ebay.com/sch/i.html?_dkr=1&iconV2Request=true&_blrs=recall_filtering&_ssn=am-autoparts&store_name=amautoparts&_oac=1&_nkw=seal%20kit'
```

```python
if url.startswith('https://www.ebay.com/'):
    param_market = 'US'
```

```python
dict_param = {}
list_param = url.split('?')[1].split('&') if '?' in url else []
for param in list_param:
    dict_param[param.split('=')[0]] = param.split('=')[1]

dict_param
```

```python
if param_market == 'US':
    dict_param.pop('_pgn', '')
    if not '_ul' in dict_param.keys():
        dict_param['_ul'] = 'US'
    if not '_stpos' in dict_param.keys():
        dict_param['_stpos'] = '91710'
    if not 'orig_cvip' in dict_param.keys():
        dict_param['orig_cvip'] = 'true'
    if not '_sop' in dict_param.keys():
        dict_param['_sop'] = '16'
    if not '_ipg' in dict_param.keys():
        dict_param['_ipg'] = '240'
    if not '_udlo' in dict_param.keys():
        dict_param['_udlo'] = ''
    if not '_udhi' in dict_param.keys():
        dict_param['_udhi'] = ''

dict_param
```

```python
url = url.split('?')[0] + '?'
for key, value in dict_param.items():
    url += f'{key}={value}&'

url = url[:-1]

url
```

```python
import requests
import json

import sys
sys.path.append('../../00.Tools')
from crawler_configuration import get_header, get_proxy

resp = requests.get(url,
                    proxies=get_proxy()).text
                    # ).text

resp
```

```python
from bs4 import BeautifulSoup
from lxml import etree

soup = BeautifulSoup(resp.strip(), 'lxml')
html = etree.HTML(str(soup))

with open('./page.html', 'w', encoding='UTF-8') as file:
    file.write(str(soup))

print(soup.prettify())
```

```python
count = int(html.xpath('//h1[@class="srp-controls__count-heading"]/span[1]/text()')[0].strip().replace(',', ''))

count
```

```python
import math

page = math.ceil(count / 240) + 1

page
```

```python
import pandas as pd

output = pd.DataFrame({'Page': [_+1 for _ in range(page)],
                       'Page Url': [f'{url}&_pgn={_+1}' for _ in range(page)]})

output.to_excel(f'./1.page.xlsx', index=False)

output
```
