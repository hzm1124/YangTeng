```python
list_file = sorted(list(os.walk('./5.type_1'))[0][2])
while '.DS_Store' in list_file:
    list_file.remove('.DS_Store')
while '0.null.txt' in list_file:
    list_file.remove('0.null.txt')

list_file
```

```python
import pandas as pd

output = pd.DataFrame()
for file in tqdm(list_file, desc='Progress', ncols=77):
    input_ = pd.read_excel(f'./5.type_1/{file}',
                           header=0,
                           dtype=str).fillna('')

    input_['Make No'] = input_['Make No'].astype(int)
    input_['Model No'] = input_['Model No'].astype(int)
    input_['Year'] = input_['Year'].astype(int)
    input_['Type No'] = input_['Type No'].astype(int)

    output = pd.concat([output, input_], ignore_index=True).fillna('')

output = output.drop_duplicates(['Type', 'Type Url'],
                                keep='first',
                                ignore_index=True)
output = output.sort_values(by=['Type'],
                            ascending=[True],
                            ignore_index=True)

output['Type No'] = [i+1 for i in range(len(output))]
output[['Type No', 'Type', 'Type Url']].to_excel('./6.type.xlsx', index=False)

output
```
