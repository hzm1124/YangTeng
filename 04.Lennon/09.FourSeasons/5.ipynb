```python
import os

list_file = sorted(list(os.walk('./file/4.src'))[0][2])
while '.DS_Store' in list_file:
    list_file.remove('.DS_Store')
while '0.null.txt' in list_file:
    list_file.remove('0.null.txt')

list_file
```

```python
from tqdm import tqdm
import pandas as pd
from datetime import datetime

input_ = pd.DataFrame()
for file in tqdm(list_file, desc='Progress', ncols=77):
    df_temp = pd.read_excel(f'./file/4.src/{file}',
                            header=0,
                            dtype=str).fillna('')
    
    df_temp['No'] = df_temp['No'].astype(int)
    df_temp['rowId'] = df_temp['rowId'].astype(int)
    df_temp['qtyEach'] = df_temp['qtyEach'].astype(int)
    df_temp['maxRows'] = df_temp['maxRows'].astype(int)

    input_ = pd.concat([input_, df_temp], ignore_index=True).fillna('')

list_column = input_.columns.tolist()
list_column.remove('No')
list_column.remove('Json_Src')
input_ = input_.drop_duplicates(list_column,
                                keep='first',
                                ignore_index=True)
input_ = input_.sort_values(by=['partLongDesc_en', 'Part Number'],
                            ascending=[True, True],
                            ignore_index=True)

input_['No'] = [_+1 for _ in range(len(input_))]

input_.to_excel(f'''./file/【网站爬虫】FourSeasons-{datetime.now().strftime('%Y%m%d')}.xlsx''', index=False)

input_
```
