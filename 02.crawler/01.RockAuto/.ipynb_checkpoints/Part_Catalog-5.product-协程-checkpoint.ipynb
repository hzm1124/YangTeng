# 0. 配置参数

- `site_code`：品类在网站的内部编码

```python
from gevent import monkey
monkey.patch_all()
from gevent.queue import Queue
import gevent

# = = = = = = = = = = = = = = = = = =

import pandas as pd
import requests
from bs4 import BeautifulSoup
from lxml import etree
import re
import time

# = = = = = = = = = = = = = = = = = =

site_code= ',drivetrain,cv+axle,2288'

proxies = {'http': 'http://t18126899091866:kh8hjqtl@j933.kdltps.com:15818',
           'https': 'http://t18126899091866:kh8hjqtl@j933.kdltps.com:15818'}

# = = = = = = = = = = = = = = = = = =

df = pd.DataFrame(columns=['make',
                           'year',
                           'model',
                           'engine',
                           'url',
                           'manufacturer',
                           'part_number',
                           'info',
                           'note',
                           'status'])

# = = = = = = = = = = = = = = = = = =

df_menu = pd.read_excel('./engine.xlsx', header=0)
list_make = df_menu['make'].to_list()
list_year = df_menu['year'].to_list()
list_model = df_menu['model'].to_list()
list_engine = df_menu['engine'].to_list()
list_url = df_menu['url'].to_list()
print('总数量：' + str(len(list_url)))
print()

# = = = = = = = = = = = = = = = = = =

work = Queue()
for make, year, model, engine, url in zip(list_make, list_year, list_model, list_engine, list_url):
    work.put_nowait((make, year, model, engine, url))

# = = = = = = = = = = = = = = = = = =

def crawler():
    global df
    
    while not work.empty():
        make, year, model, engine, url = work.get_nowait()
        
        for i in range(31):
            try:
                resp = requests.get(url + site_code).text

                # = = = = = = = = = = = = = = = = = =
                
                soup = BeautifulSoup(resp, 'lxml')
                
                # = = = = = = = = = = = = = = = = = =

                html = etree.HTML(str(soup))
                
                list_row = html.xpath('//div[@class="listing-text-row-moreinfo-truck"]')
                
                # = = = = = = = = = = = = = = = = = =

                list_manufacturer = [row.xpath('.//span[@class="listing-final-manufacturer"]/text()')[0] for row in list_row]

                # = = = = = = = = = = = = = = = = = =

                list_part_number = [row.xpath('.//span[@title="Buyer\'s Guide"]/text()')[0] for row in list_row]

                # = = = = = = = = = = = = = = = = = =

                list_info = []
                for j in range(len(list_row)):
                    info = list_row[j].xpath('.//a[@class="ra-btn ra-btn-moreinfo"]')
                    if len(info) == 0:
                        list_info.append('')
                    else:
                        list_info.append(info[0].get('href'))

                # = = = = = = = = = = = = = = = = = =
                
                list_row = html.xpath('//div[@class="listing-text-row"]')
                
                # = = = = = = = = = = = = = = = = = =
                
                list_note = [re.sub(' +', ' ', ' '.join(row.xpath('./descendant-or-self::*/text()'))).strip() for row in list_row]
                
                # = = = = = = = = = = = = = = = = = =

                df_temp = pd.DataFrame({'make': make,
                                        'year': int(year),
                                        'model': model,
                                        'engine': engine,
                                        'url': url,
                                        'manufacturer': list_manufacturer,
                                        'part_number': list_part_number,
                                        'info': list_info,
                                        'note': list_note,
                                        'status': 'ok'})

                break

            except:
                if i == 30:
                    df_temp = pd.DataFrame({'make': make,
                                            'year': int(year),
                                            'model': model,
                                            'engine': engine,
                                            'url': url,
                                            'manufacturer': '',
                                            'part_number': '',
                                            'info': '',
                                            'text': '',
                                            'status': ['error']})
                time.sleep(0.3)
                continue

        # = = = = = = = = = = = = = = = = = =

        df = pd.concat([df, df_temp], ignore_index=True).fillna('')
        
        print(make + ' - ' + str(year) + ' - ' + model + ' - ' + engine +  '  <->  [' + str(i) + '] - 剩余：' + str(work.qsize()))
        
# = = = = = = = = = = = = = = = = = =

list_task = []
for _ in range(25):
    task = gevent.spawn(crawler)
    list_task.append(task)
gevent.joinall(list_task)

df.to_excel('./product.xlsx', index=False)
print()
print('搞定')
```
