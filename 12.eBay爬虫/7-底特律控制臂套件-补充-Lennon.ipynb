# 1. 处理空行

```python
import pandas as pd
```

```python
df = pd.read_excel('./crawler-detroit_20230424_205220.xlsx', header=0)

df
```

```python
for i in range(len(df)):
    print(i)
    try:
        list_row = df.loc[i, '内容'].split('\n')

        for j in range(len(list_row)):
            list_row[j] = list_row[j].strip()
            if list_row[j][-5:] == 'lamps' or list_row[j][-4:] == 'oots' or list_row[j][-4:] == 'bely':
                list_row[j] += ' - N/A'

        while True:
            if '' in list_row:
                list_row.remove('')
            else:
                break

        df.loc[i, '内容_new'] = '\n'.join(list_row)
    except:
        pass
```

```python
df.to_excel('./Lennon4.xlsx', index=False)

df
```

# 2. 分行

```python
import pandas as pd
```

```python
df = pd.read_excel('./Lennon4.xlsx', header=0)

df
```

```python
df_list = []
for i in range(len(df)):
    print(i)
    try:
        list_kit = df.loc[i, '内容_new'].split('\n')
        for j in range(len(list_kit)):
            df_temp = pd.DataFrame({'Item': [df.loc[i, 'Item']],
                                    '序号': [str(j+1)],
                                    '内容_new': [df.loc[i, '内容_new']],
                                    'kit': [list_kit[j]]})
            df_list.append(df_temp)
    except:
        pass
```

```python
pd.concat(df_list, ignore_index=True).to_excel('./Lennon4.xlsx', index=False)
```

# 3. 分列

```python
import pandas as pd
```

```python
df = pd.read_excel('./Lennon4.xlsx', header=0)

df
```

```python
import re
```

```python
for i in range(len(df)):
    print(i)
    try:
        amount = re.match(r'^[1-9][xX] ?', df.loc[i, 'kit'])[0]
        number = df.loc[i, 'kit'].split('-')[-1]
        name = df.loc[i, 'kit'].replace(amount, '').replace(number, '').strip()[:-1].strip()
        amount = amount.strip()
        if amount[-1] == 'x' or amount[-1] == 'X':
            amount = amount[:-1]
        number = number.strip()
        
        df.loc[i, 'amount'] = amount
        df.loc[i, 'name'] = name
        df.loc[i, 'number'] = number
    except:
        pass
```

```python
df.to_excel('./Lennon4.xlsx', index=False)

df
```
