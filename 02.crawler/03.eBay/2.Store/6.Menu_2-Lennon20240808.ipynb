```python
import os

os.makedirs('./6.menu_2')
os.makedirs('./7.item_1')
os.makedirs('./7.item_1/0.error')
```

```python
import pandas as pd

input_ = pd.read_excel('./5.menu.xlsx',
                       header=0,
                       dtype=str).fillna('')

input_['1st Category No'] = input_['1st Category No'].astype(int)
input_['2nd Category No'] = ['-' if _ == '-' else int(_) for _ in input_['2nd Category No'].tolist()]
input_['No'] = input_['No'].astype(int)

input_
```

```python
input_ = input_.drop_duplicates(['Item Number'],
                                keep='first',
                                ignore_index=True)
input_ = input_.sort_values(by=['1st Category No', '2nd Category No', 'No'],
                            ascending=[True, True, True],
                            ignore_index=True)

input_['No'] = [_+1 for _ in range(len(input_))]

list_output, index_start, index_end = [], 0, 3_000
while True:
    if index_end < len(input_):
        list_output.append(input_[index_start:index_end].reset_index(drop=True))
        index_start, index_end = index_end, index_end + 3_000
    else:
        list_output.append(input_[index_start:].reset_index(drop=True))
        break

if len(list_output) == 1:
    input_[['No',
            'Item Number',
            'Url']].to_excel('./6.menu_2/menu_2.xlsx', index=False)
else:
    for i, output in zip(range(len(list_output)), list_output):
        output[['No',
                'Item Number',
                'Url']].to_excel(f'./6.menu_2/menu_2-serial_{i+1}.xlsx', index=False)

print('Done ~')
```
