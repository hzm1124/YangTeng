```python
html.xpath('//table[@class="table-fixed rounded-lg"]/tbody/tr')
```

```python
resp = html.xpath('//div[@class="body-overlay"]/script/text()')[0].split('window.currentPart = ')[1][:-2].strip()
resp = json.loads(resp)

resp
```

```python
dict_padb = resp['padb']

dict_padb
```

```python
dict_attribute = resp['attributes']

dict_attribute
```

```python
list_oe = []
oe = resp['interchange'].split(',')
for _ in oe:
    if _.strip() != 'NI':
        list_oe.append(_.strip())

oe = ';'.join(list_oe)

oe
```

```python
df_temp = pd.DataFrame([{'status': 'ok',
                         '序号': '',
                         'Part_Number': part_number,
                         'Category': resp['major'].strip(),
                         'Type': resp['minor'].strip(),
                         'Description': resp['long_desc'].strip(),
                         'Vehicle': '',
                         'OE': oe,
                         'Pic': '',
                         'Url': 'https://motorad.com/part/' + part_number,
                         'Src': 'https://saasphoto.com/share/PaSfVF/motorad-web-images/' + resp['angle_view'].strip()}])

if dict_attribute != False:
    for key_value in dict_attribute:
        df_temp.loc[0, key_value.strip().lower()] = dict_attribute[key_value].strip()

if dict_padb != False:
    for key_value in dict_padb:
        df_temp.loc[0, key_value.strip().upper()] = dict_padb[key_value].strip()
    
df = pd.concat([df, df_temp], ignore_index=True).fillna('')

df
```
