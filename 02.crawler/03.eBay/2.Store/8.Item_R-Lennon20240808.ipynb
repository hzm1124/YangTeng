```python
list_file = sorted(list(os.walk('./7.item_1'))[0][2])
while '.DS_Store' in list_file:
    list_file.remove('.DS_Store')
while '0.null.txt' in list_file:
    list_file.remove('0.null.txt')

list_file
```

```python
import pandas as pd

output = pd.DataFrame()
for file in tqdm(list_file, desc='Progress', ncols=77):
    input_ = pd.read_excel(f'./7.item_1/{file}',
                           header=0,
                           dtype=str).fillna('')

    input_['No'] = input_['No'].astype(int)

    output = pd.concat([output, input_], ignore_index=True).fillna('')

output = output.sort_values(by=['No'],
                            ascending=[True],
                            ignore_index=True)

output.to_excel('./8.item.xlsx', index=False)

output
```
