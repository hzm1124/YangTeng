## 2.2. error修正

```python
import pandas as pd
import os
```

```python
df = pd.read_excel('./menu.xlsx',
                   header=0,
                   dtype=str).fillna('')

df_file = pd.DataFrame()

list_file = list(os.walk('./part'))[0][2]
if '.DS_Store' in list_file:
    list_file.remove('.DS_Store')
if '0.null.txt' in list_file:
    list_file.remove('0.null.txt')
    
for i in range(len(list_file)):
    df_file.loc[i, 'No.'] = list_file[i].split('.txt')[0]
    df_file.loc[i, 'Crawler'] = 'y'
    
df = pd.merge(left=df,
              right=df_file,
              how='left',
              left_on='No.',
              right_on='No.').fillna('')

df = df[df['Crawler'] != 'y'].reset_index(drop=True)

with open('./part.txt', 'w', encoding='UTF-8') as f:
    f.write('let list_no = ["' + '","'.join(df['No.'].tolist()) + '"];\nlet list_url = ["' + '","'.join(df['Url'].tolist()) + '"];\n')

print('Done ~')
```

## 4.2. error修正

```python
import pandas as pd
import os
```

```python
df = pd.read_excel('./make.xlsx',
                   header=0,
                   dtype=str).fillna('')

df['Code'] = [part_code + ';' + make_code for part_code, make_code in zip(df['Part_Code'].tolist(), df['Make_Code'].tolist())]

df_file = pd.DataFrame()

list_file = list(os.walk('./model'))[0][2]
if '.DS_Store' in list_file:
    list_file.remove('.DS_Store')
if '0.null.txt' in list_file:
    list_file.remove('0.null.txt')
    
for i in range(len(list_file)):
    df_file.loc[i, 'Code'] = list_file[i].split('.txt')[0]
    df_file.loc[i, 'Crawler'] = 'y'
    
df = pd.merge(left=df,
              right=df_file,
              how='left',
              left_on='Code',
              right_on='Code').fillna('')

df = df[df['Crawler'] != 'y'].reset_index(drop=True)

with open('./model.txt', 'w', encoding='UTF-8') as f:
    f.write('let list_part_code = ["' + '","'.join(df['Part_Code'].tolist()) + '"];\nlet list_make_code = ["' + '","'.join(df['Make_Code'].tolist()) + '"];\nlet list_url = ["' + '","'.join(df['Model_Url'].tolist()) + '"];\n')

print('Done ~')
```
