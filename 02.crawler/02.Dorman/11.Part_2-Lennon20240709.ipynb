```python
param_date = '20240706'
```

```python
import os

os.makedirs(f'./{param_date}/11.part_2')
os.makedirs(f'./{param_date}/12.vehicle_1')
```

```python
import pandas as pd

input_ = pd.read_excel(f'./{param_date}/10.part.xlsx',
                       header=0,
                       dtype=str).fillna('')

input_['No'] = input_['No'].astype(int)

input_
```

```python
import math
from tqdm import tqdm

num = math.ceil(len(input_) / 3_000)
for i in tqdm(range(num), desc='Progress', ncols=77):
    input_[i*3000:i*3_000+3_000].to_excel(f'./{param_date}/11.part_2/part_{i+1}-fragment_{i+1}.xlsx', index=False)

print('Done ~')
```
