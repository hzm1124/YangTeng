```python
param_date = '20240706'
```

```python
import os

list_file = sorted(list(os.walk(f'./{param_date}/5.type_1'))[0][2])
while '.DS_Store' in list_file:
    list_file.remove('.DS_Store')
while '0.null.txt' in list_file:
    list_file.remove('0.null.txt')

list_file
```

```python
from tqdm import tqdm
import re

for file in tqdm(list_file, desc='Progress', ncols=77):
    os.rename(f'./{param_date}/5.type_1/{file}',
              f'''./{param_date}/5.type_1/{re.sub(r'-[0-9]{8}_[0-9]{6}.xlsx$', '.xlsx', file)}''')

print('Done ~')
```

```python
list_file = sorted(list(os.walk(f'./{param_date}/5.type_1'))[0][2])
while '.DS_Store' in list_file:
    list_file.remove('.DS_Store')
while '0.null.txt' in list_file:
    list_file.remove('0.null.txt')

list_file
```

```python
import pandas as pd

output = pd.DataFrame()
for file in tqdm(list_file, desc='Progress', ncols=77):
    input_ = pd.read_excel(f'./{param_date}/5.type_1/{file}',
                           header=0,
                           dtype=str).fillna('')

    output = pd.concat([output, input_], ignore_index=True).fillna('')

output = output[['Type', 'Type Url']].drop_duplicates(keep='first', ignore_index=True)
output = output.sort_values(by=['Type'],
                            ascending=[True],
                            ignore_index=True)

output['Type No'] = [i+1 for i in range(len(output))]
output.to_excel(f'./{param_date}/6.type.xlsx', index=False)

output
```
