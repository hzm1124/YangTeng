```python
from lxml import etree
import re
import html
```

```python
dict_key, list_tr = {}, selector.xpath('//table[@class="moreinfotable"]/tbody/tr')
for tr in list_tr:
    list_td = tr.xpath('./td')
    if not list_td:
        continue
    list_text = [etree.tostring(td.root).decode('utf-8') for td in list_td]
    list_text = [re.sub(r'<.*?>', '', text) for text in list_text]
    list_text = [re.sub(r' +', ' ', text) for text in list_text]
    list_text = [html.unescape(text).strip() for text in list_text]

    key = list_text[0]

    if key in dict_key:
        dict_key[key] += 1
    else:
        dict_key[key] = 1

dict_specification, dict_count = {}, {}
for tr in list_tr:
    list_td = tr.xpath('./td')
    if not list_td:
        continue
    list_text = [etree.tostring(td.root).decode('utf-8') for td in list_td]
    list_text = [re.sub(r'<.*?>', '', text) for text in list_text]
    list_text = [re.sub(r' +', ' ', text) for text in list_text]
    list_text = [html.unescape(text).strip() for text in list_text]

    key, value = list_text[0], '' if len(list_text) == 1 else '\n'.join(list_text[1:])

    if dict_key[key] == 1:
        dict_specification[key] = 'nan' if value == '' else value
    else:
        if key in dict_count:
            dict_count[key] += 1
        else:
            dict_count[key] = 1
        dict_specification[f'{key}_{dict_count[key]}'] = 'nan' if value == '' else value

dict_specification
```

```python
import json
from urllib.parse import urljoin
```

```python
dict_kit, list_tr = {}, selector.xpath('//section[@aria-label="Components that come in this kit"]/table/tbody/tr/td/table/tbody[contains(@class, "listing-inner") and (contains(@class, "altrow-b-0") or contains(@class, "altrow-b-1"))]/tr')
for i, tr in enumerate(list_tr):
    dict_kit[str(i)] = {}

    dict_kit[str(i)]['kit_quantity'] = int(tr.xpath('./td[@class="pinched" and contains(@style, "vertical-align")]/text()').get().strip())

    type_ = tr.xpath('./td[@class="taleft pinched"]/text()').get().strip()
    dict_kit[str(i)]['kit_type'] = type_

    dict_ = json.loads(tr.xpath('./td[contains(@id, "listing_data_essential[")]/@value').get())
    type_code, part_code = dict_['parttype'], dict_['partkey']
    dict_kit[str(i)]['kit_type_code'], dict_kit[str(i)]['kit_part_code'] = type_code, part_code

    dict_ = json.loads(tr.xpath('./td[contains(@id, "listing_data_essential[")]/input[contains(@id, "listing_data_supplemental[")]/@value').get())
    part_number, manufacturer = dict_['partnumber'], dict_['catalogname']
    dict_kit[str(i)]['kit_part_number'], dict_kit[str(i)]['kit_manufacturer'] = part_number, manufacturer

    dict_kit[str(i)]['kit_url'] = f'https://www.rockauto.com/en/parts/{manufacturer},{part_number},{type_},{type_code}&Lennon=1'

    dict_kit[str(i)]['kit_join_mpntcpc'] = f'{manufacturer};{part_number};{type_code};{part_code}'

    list_a = tr.xpath('./td[contains(@id, "listing_data_essential[")]/div[@class="listing-text-row-moreinfo-truck"]/a[@class="ra-btn ra-btn-moreinfo"]')
    info_url = list_a[0].xpath('./@href').get() if list_a else ''
    info_url = info_url if info_url.startswith('https://www.rockauto.com/en/moreinfo.php?') else f'https://www.rockauto.com/en/moreinfo.php?pk={part_code}&pt={type_code}&Lennon=1'
    dict_kit[str(i)]['kit_info_url'] = info_url

    dict_src, j, list_value = {}, 0, tr.xpath('./td[@class="pinched" and not(contains(@style, "vertical-align"))]/div/input[contains(@id, "jsninlineimg[")]/@value').getall()
    for value in list_value:
        list_src = json.loads(value).get('Slots', None)
        for src in list_src:
            dict_src[str(j)] = urljoin(item['url'], src['ImageData']['Full'].strip())
            j += 1
    dict_kit[str(i)]['kit_dict_src'] = dict_src

dict_kit
```
